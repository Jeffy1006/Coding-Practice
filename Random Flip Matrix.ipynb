```python
import random
Solution:
    def __init__(self, n_rows, n_cols):
        self.c = n_cols
        self.end = n_rows * n_cols - 1
        self.d = {}
        self.start = 0
    
    def flip(self):
        rand = random.randint(self.start, self.end)
        res = self.d.get(rand, rand)
        self.d[rand] = self.d.get(self.start, self.start) # swap the generated random element with the first element
        self.start += 1
        return divmod(res, self.c)
    
    def reset(self):
        self.d = {}
        self.start = 0
```
