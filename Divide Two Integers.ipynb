```python
def divideTwoInt(dividend, divisor):
    MAX_INT = 2**31 - 1
    MIN_INT = -2**31 
    HALF_MIN_INT = MIN_INT // 2
    
    if dividend == MIN_INT and divisor == -1:
        return MAX_INT
    
    negatives = 2
    if dividend > 0:
        negatives -= 1
        dividend = -dividend
    if divisor > 0:
        negatives -= 1
        divisor = -divisor
        
    quotient = 0
    
    while divisor >= dividend:
        powerOfTwo = -1
        value = divisor
        while value >= HALF_MIN_INT and value + value >= dividend:
            value += value
            powerOfTwo += powerOfTwo
        quotient += powerOfTwo
        dividend -= value
    
    return -quotient if negatives != 1 else quotient
```

```python
def divideTwoInt(dividend, divisor):
    MAX_INT = 2**31 - 1
    MIN_INT = -2**31 
    HALF_MIN_INT = MIN_INT // 2
    
    if dividend == MIN_INT and divisor == -1:
        return MAX_INT
    
    negatives = 2
    if dividend > 0:
        negatives -= 1
        dividend = -dividend
    if divisor > 0:
        negatives -= 1
        divisor = -divisor
        
    highest_double = divisor
    highest_power_of_two = -1
    while highest_double >= HALF_MIN_INT and highest_double + highest_double >= dividend:
        highest_power_of_two += highest_power_of_two
        highest_double += highest_double
    
    quotient = 0
    while divisor >= dividend:
        if highest_double >= dividend:
            quotient += highest_power_of_two
            dividend -= highest_double
        highest_power_of_two >>= 1
        highest_double >>= 1
        
    return quotient if negatives != 1 else -quotient
```
