```python
class SparseVector:
    def __init__(self, nums):
        self.vector = {}
        for i in range(len(nums)):
            if nums[i] != 0:
                self.vector[i] = nums[i]
        

    # Return the dotProduct of two sparse vectors
    def dotProduct(self, vec):
        res = 0
        for i, v in vec.vector.items():
            if i in self.vector:
                res += v * self.vector[i]
        return res
        
```

```python
nums1 = [1,0,0,2,3] 
nums2 = [0,3,0,4,0]
v1 = SparseVector(nums1)
v2 = SparseVector(nums2)
ans = v1.dotProduct(v2)
ans
```

```python
nums1 = [0,1,0,0,0]
nums2 = [0,0,0,0,2]
v1 = SparseVector(nums1)
v2 = SparseVector(nums2)
ans = v1.dotProduct(v2)
ans
```

```python
nums1 = [0,1,0,0,2,0,0]
nums2 = [1,0,0,0,3,0,4]
v1 = SparseVector(nums1)
v2 = SparseVector(nums2)
ans = v1.dotProduct(v2)
ans
```
