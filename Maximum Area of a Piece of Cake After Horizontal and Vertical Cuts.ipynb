```python
def maxArea(self, h, w, horizontalCuts, verticalCuts) -> int:
    horizontalCuts = [0] + sorted(horizontalCuts) + [h]
    verticalCuts = [0] + sorted(verticalCuts) + [w]
    max_h = max(j - i for i, j in zip(horizontalCuts, horizontalCuts[1:]))
    max_w = max(j - i for i, j in zip(verticalCuts, verticalCuts[1:]))
    return (max_h * max_w) % ((10**9) + 7)
```
