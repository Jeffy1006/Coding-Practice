```python
def integerBreak(n):
    mem = {}
    
    def helper(n):
        if n <= 2:
            return 1
        if n in mem:
            return mem[n]
        
        ans = 1 * (n - 1)
        for i in range(2, n):
            product = i * max(n - 1, integerBreak(n - i))
            ans = max(ans, product)
        mem[n] = ans
        return ans
    
    return helper(n)
```

```python
def integerBreak(n):
    dp = [0] * (n + 1)
    dp[1] = 1
    
    for i in range(2, n + 1):
        for j in range(1, (i//2) + 1):
            dp[i] = max(dp[i], (max(j, dp[j]) * max(i - j, dp[i - j])))
    return dp[n]
```
