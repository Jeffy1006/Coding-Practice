```python
def isPalin(head):
    if not head or not head.next:
        return True
    
    slow, fast = head, head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    head2 = slow.next
    slow.next = None
    
    prev, cur = None, head2
    while cur:
        temp = cur.next
        cur.next = prev
        prev = cur
        cur = temp
    head2 = prev
    
    cur1 = head
    cur2 = head2
    while cur2:
        if cur2.val != cur1.val:
            return False
    
    return True
```
