```python
class Nodes():
    def __init__(self, data):
        self.data = data
        self.next = None
    
```

```python
class LL():
    def __init__(self):
        self.head = None
    
    def start(self):
        return self.head
    
    def push(self, data):
        newNode = Nodes(data)
        newNode.next = self.head
        self.head = newNode
    
    def rdpii(self):
        if self.head is None or self.head.next is None:
            return
        
        h1 = Nodes(0)
        h1.next = self.head
        tmp = h1
        cur = self.head
        
        while cur != None and cur.next != None:
            if cur.data == cur.next.data:
                x = cur.data
                while cur != None and cur.data == x:
                    cur = cur.next
                tmp.next = cur
            else:
                tmp = cur
                cur = cur.next
        
        self.head = h1.next
        
    
    def printll(self):
        cur = self.head
        while cur is not None:
            print(cur.data, end = " ")
            cur = cur.next
```

```python
ll = LL()
for i in range(5):
    ll.push(2)
print("Before")
ll.printll()
print()
print("After")
ll.rdpii()
ll.printll()
```
