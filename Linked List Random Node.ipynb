```python
from random import random
class Solution:
    def __init__(self, head):
        self.head = head

    def getRandom(self):
        cur_node = self.head
        counter = 1
        choice = -1
        
        while cur_node:
            if random.random() < 1 / counter:
                choice = cur_node.val
            cur_node = cur_node.next
            counter += 1
        return choice
```
