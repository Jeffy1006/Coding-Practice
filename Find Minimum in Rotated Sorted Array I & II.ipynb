```python
def findMin(nums):
    
    n = len(nums)
    i, j = 0, n - 1
    minNum = float('inf')
    
    while i <= j:
        if i == j:
            minNum = min(minNum, nums[i])
            break
        elif j - i == 1:
            minNum = min(minNum, min(nums[i],nums[j]))
            break
        else:
            mid = (i + j) // 2
            minNum = min(minNum, nums[mid])
            if nums[mid] > nums[j]:
                i = mid + 1
            else:
                j = mid - 1

    return minNum  
```

```python
def findMin(nums):
    
    l, r = 0, len(nums) - 1
    
    while l < r:
        mid = (l + r) // 2
        if nums[mid] > nums[r]:
            l = mid + 1
        else:
            r = mid
    
    return nums[l]
```

```python
def findMinII(nums):
    
    l, r = 0, len(nums) - 1
    
    while l < r:
        mid = (l + r) // 2
        if nums[mid] > nums[r]:
            l = mid + 1
        elif nums[mid] < nums[r]:
            r = mid
        else:
            r -= 1
    
    return nums[l]
```
