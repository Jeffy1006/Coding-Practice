```python
def getMoneyAmount(n):
    dp = [[0] * (n + 1) for _ in range(n + 2)]
    for length in range(2, n + 1):
        for start in range(1, n - length + 2):
            end = start + length - 1
            curMin = float("inf")
            #print("start :{}, end :{}".format(start, end))
            for pivot in range(start, end + 1):
                print("start :{}, end :{}, pivot:{}".format(start, end, pivot))
                curMin = min(curMin, pivot + max(dp[start][pivot - 1], dp[pivot + 1][end]))
            dp[start][end] = curMin
    return dp[1][n]
```
