```python
# TC: O(nlgn)
# SC: O(n)
def findMinArrowShots(points):
    if not points: return 0
    points.sort(key = lambda x: x[1])
    right = points[0][1]
    ans = 1
    for point in points:
        if point[0] > right:
            right = point[1]
            ans += 1
    return ans
```

```python
findMinArrowShots([[10,16],[2,8],[1,6],[7,12]])
```
