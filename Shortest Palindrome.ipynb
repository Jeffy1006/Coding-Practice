```python
def shortestPalindrome(s):
    
    def isPalindrome(w):
            i, j = 0, len(w) - 1
            while i < j:
                if w[i] != w[j]:
                    return False
                i += 1
                j -= 1
            return True
    
    if isPalindrome(s):
        return s
    
    for i in range(len(s)-1, -1,-1):
        temp = list(s[i:])
        temp.reverse()
        temp = "".join(temp) + s
        if isPalindrome(temp):
            return temp
```
