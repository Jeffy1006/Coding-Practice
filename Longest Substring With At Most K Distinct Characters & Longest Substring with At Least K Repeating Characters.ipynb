```python
from collections import defaultdict
def longestSubstring(input, k):
    if len(input) == 0 or k == 0: return 0
    startIdx, endIdx = 0, 0
    seen = defaultdict(int)
    result = 0
    
    while endIdx < len(input):
        if input[endIdx] not in seen:
            k -= 1
        seen[input[endIdx]] = endIdx
        while k < 0:
            if seen[input[startIdx]] == startIdx:
                k += 1
                # keep dict at k keys, otherwise k won't minus 1 because many have been seen
                del seen[input[startIdx]]
            startIdx += 1
        
        result = max(result, endIdx - startIdx + 1)
        endIdx += 1
    
    return result
```

```python
import collections
def lengthOfLongestSubstringKDistinct(s, k):
    counter = defaultdict(int)
    left = ans = 0
    for right, c in enumerate(s):
        counter[c] += 1
        while len(counter) > k:
            counter[s[left]] -= 1
            if counter[s[left]] == 0:
                del counter[s[left]]
            left += 1
        ans = max(ans, right - left + 1)
    return ans
```

```python
from collections import Counter
def longestSubstring(s, k):
    maxUnique = len(Counter(s))
    res = 0
    
    for curUnique in range(1, maxUnique + 1):
        countMap = [0] * 26
        windowStart = windowEnd = idx = unique = countAtLeastK = 0
        while windowEnd < len(s):
            if unique <= curUnique:
                idx = s[windowEnd] - 'a'
                if countMap[idx] == 0:
                    unique += 1
                countMap += 1
                if countMap[idx] == k:
                    countAtLeastK += 1
                windowEnd += 1
            else:
                idx = s[windowStart] - 'a'
                if countMap[idx] == k:
                    countAtLeastK -= 1
                countMap -= 1
                if countMap[idx] == 0:
                    unique -= 1
                windowStart += 1 
            if unique == curUnique and unique == countAtLeastK:
                res = max(res, windowEnd - windowStart)
    
    return res
```
