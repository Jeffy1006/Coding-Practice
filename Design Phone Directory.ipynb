```python
class PhoneDirectory:
    def __init(self, maxNumbers):
        available = set()
        for i in range(maxNumbers):
            available.add(i)
    
    def get(self):
        tmp = -1
        for i in self.available:
            tmp = i
            break
        self.available.remove(tmp)
        return tmp
    
    def check(number):
        return number in self.available
```
