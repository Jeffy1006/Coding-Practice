```python
def eraseOverlapIntervals(intervals):
    intervals.sort()
    n = len(intervals)
    count = 0
    left, right = 0, 1
    while right < n:
        if intervals[left][1] <= intervals[right][0]:
            left = right
            right += 1
        elif intervals[left][1] <= intervals[right][1]:
            count += 1
            right += 1
        elif intervals[left][1] > intervals[right][1]:
            count += 1
            left = right
            right += 1
    return count
```
