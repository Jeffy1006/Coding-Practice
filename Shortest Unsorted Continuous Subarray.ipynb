```python
def findUnsortedSubarray(nums):
    i, j = 0, len(nums) - 1
    while i < j: 
        if nums[i] > nums[i + 1]: break
        i += 1
    if i == j:
        return 0
    while j > 0:
        if nums[j] < nums[j - 1]: break
        j -= 1
    min_num = min(nums[i: j + 1])
    max_num = max(nums[i: j + 1])
    x, y = 0, len(nums) - 1
    while x < i: 
        if nums[x] > min_num: break
        x += 1
    while y > j:
        if nums[y] < max_num: break
        y -= 1
    return y - x + 1
```

```python
findUnsortedSubarray([2,3,4,1,5,6])
```
