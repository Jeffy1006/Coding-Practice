```python
 def jobScheduling(startTime, endTime, profit):
        n = len(profit)
        interval = [(0, 0, 0)]
        interval.extend(sorted(zip(startTime, endTime, profit)))
        dp = [0] * (n + 1)
        
        for i in range(n + 1):
            for j in range(i):
                if interval[i][0] >= interval[j][1]:
                    dp[i] = max(dp[i], dp[j] + interval[i][2])

        return max(dp)
```

```python
import bisect
def jobScheduling(startTime, endTime, profit):
    jobs = sorted(zip(startTime, endTime, profit), key=lambda v: v[1])
    dp = [[0, 0]]
    print(jobs)
    for s, e, p in jobs:
        i = bisect.bisect(dp, [s + 1]) - 1
        print('{} {} {}'.format(dp, i, s))
        if dp[i][1] + p > dp[-1][1]:
            dp.append([e, dp[i][1] + p])
    return dp[-1][1]
```

```python
startTime = [1,2,3,3] 
endTime = [3,4,5,6]
profit = [50,10,40,70]
jobScheduling(startTime, endTime, profit)
```
