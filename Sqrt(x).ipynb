```python
def mySqrt(x):
    if x < 2: return x
    
    l, r = 1, x
    while l < r:
        mid = (l + r) // 2
        val = mid * mid
        if val ==  x:
            return mid
        elif val < x:
            l = mid + 1
        else:
            r = mid
    
    return l - 1
```

```python
mySqrt(10)
```
