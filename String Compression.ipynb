```python
def compressI(chars):
    ans = 0
    left, right = 0, 0
    while right < len(chars):
        while right < len(chars) and chars[right] == chars[left]:
            right += 1
        ans += 1 + (len(str(right - left)) if right - left > 10 else 1)
        left = right
    return ans
```

```python
def compressII(chars):
    rptr, wptr = 0, 0
    while rptr < len(chars):
        ch, f = chars[rptr], 0
        while rptr < len(chars) and chars[rptr] == ch:
            rptr, f = rptr + 1, f + 1
        chars[wptr], wptr = ch, wptr + 1
        if f > 1:
            for c in str(f):
                chars[wptr], wptr = c, wptr + 1
    return chars[:wptr]
```

```python
chars = ['a', 'a', 'b', 'b', 'c', 'c', 'c']
print(compressI(chars))
print(compressII(chars))
```
