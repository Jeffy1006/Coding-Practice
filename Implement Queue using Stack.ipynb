```python
class Queue
    def __init__(self):
        self.stack1 = []
        self.stack2 = []
        self.size = 0
    
    def enque(self, val):
        
        while self.stack1:
            self.stack2.push(self.stack1.pop())
        
        self.stack1.push(val)
        self.size += 1
        
        while self.stack2:
            self.stack1.push(self.stack2.pop())
        
    def deque(self):
        if self.size == 0:
            return None
            
        self.size -= 1
        return self.stack1.pop()
        
    def peak(self):
        return self.stack1[-1]
    
    def empty(self):
        return True if self.size == 0 else False 
    
```
