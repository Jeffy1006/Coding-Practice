```python
def increasingTripletSubsequence(nums):
    if len(nums) < 3: return False
    idx1 = idx2 = float("inf")
    for i in range(len(nums)):
        # <= instead of < to absorb duplicate
        if nums[i] <= idx1:
            idx1 = nums[i]
        elif nums[i] <= idx2:
            idx2 = nums[i]
        else:
            return True
    return False
```
