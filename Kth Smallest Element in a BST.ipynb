```python
def kSmallest(root, k):
    n = 0
    stack = []
    cur = root
    
    while True:
        while cur:
            stack.append(cur)
            cur = cur.left
        cur = stack.pop()
        n += 1
        if n == k:
            return cur.val
        cur = cur.right
```

```python
def kSmallest(root, k):
    n, res = 0, None
    
    def inorder(node):
        nonlocal n
        if not node: 
            return
        inorder(node.left)
        n += 1
        if n == k:
            res = node.val
            return
        inorder(node.right)
    
    inorder(root)
    return res
```
