```python
def nextGreaterElement(nums1, nums2):
    d = {}
    st = []
    ans = []
    
    for x in nums2:
        while st and st[-1] < x:
            d[st.pop()] = x
        st.append(x)
    
    for x in nums1:
        ans.append(d.get(x, -1)) 
    return ans
```

```python
def nextGreaterElementII(nums):
    n = len(nums)
    stack, res = [], [-1] * n
    for i in range(2 * n):
        while stack and nums[stack[-1]] < nums[i % n]:
            res[stack.pop()] = nums[i % n]
        stack.append(i % n)
    return res
```

```python
def nextGreaterElementII(n):
    s = list(str(n))
    i = len(s) - 1
    while i > 0 and s[i] < s[i - 1]: i -= 1
    if i == 0: return -1
    
    nxt = min(c for c in s[i:] if c > s[i - 1])
    j = s.index(nxt, i)
    s[i - 1], s[j] = s[j], s[i - 1]
    s[i:] = s[i:][::-1]
    return int(''.join(s))
```

```python
def nextGreaterElementII(n):
    s = list(str(n))
    i = len(s) - 1
    while i > 0 and s[i] < s[i - 1]: i -= 1
    if i == 0: return -1
    
    j = i
    while j + 1 <= len(s) and s[j + 1] > s[i - 1]: j += 1
    s[i - 1], s[j] = s[j], s[i - 1]
    s[i:] = s[i:][::-1]
    ret = int(''.join(s))
    return ret if ret < 1<<31 else -1
```
