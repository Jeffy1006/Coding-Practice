```python
from collections import defaultdict
def subarraysWithKDistinct(nums, k):
    return atMostK(nums, k) - atMostK(nums, k - 1)
    
def atMostK(nums, k):
    count = defaultdict(int)
    res = i = 0
    for j in range(len(nums)):
        if count[nums[j]] == 0: 
            k -= 1
        count[nums[j]] += 1
        while k < 0:
            count[nums[i]] -= 1
            if count[nums[i]] == 0:
                k += 1
            i += 1
        res += j - i + 1
    return res           
```

```python
subarraysWithKDistinct([1,2,1,3,4], 3)
```
