```python
def trppingRainWaterII(height):
    if not height: return 0
    
    l, r = 0, len(height) - 1
    leftMax, rightMax = height[l], height[r]
    res = 0
    
    while l < r:
        if leftMax < rightMax:
            l += 1
            leftMax = max(leftMax, height[l])
            res += leftMax - height[l]
        else:
            r -= 1
            rightMax = max(rightMax, height[r])
            res += rightMax - height[r]
    
    return res
```

```python
import heapq
def trppingRainWaterII(matrix):
    m, n = len(matrix), len(matrix[0])
    visited = [[False] * n for _ in range(m)]
    pq = []
    for i in range(n):
        heapq.heappush(pq, (matrix[0][i], 0, i))
        heapq.heappush(pq, (matrix[m - 1][i], m - 1, i))
        visited[0][i] = True
        visited[m - 1][i] = True
    for i in range(1, m - 1):
        heapq.heappush(pq, (matrix[i][0], i, 0))
        heapq.heappush(pq, (matrix[i][n - 1], i, n - 1))
        visited[i][0] = True
        visited[i][n - 1] = True
    
    ans = 0
    direction = [0, 1, 0, -1, 0]
    MAX = float("-inf")
    while pq:
        tmp = heapq.heappop(pq)
        if 0 < tmp[1] < m - 1 and 0 < tmp[2] < n - 1 and tmp[0] < MAX:
            ans += MAX - tmp[0]
        MAX = max(MAX, tmp[0])
        for i in range(1, len(direction)):
            x = tmp[1] + direction[i]
            y = tmp[2] + direction[i - 1]
            if 0 < x < m - 1 and 0 < y < n - 1 and not visited[x][y]:
                heapq.heappush(pq, (matrix[x][y], x, y))
                visited[x][y] = True
    return ans
```
