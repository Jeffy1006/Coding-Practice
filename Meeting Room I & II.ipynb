```python
def canAttendMeetings(intervals):
    intervals.sort()
    for i in range(len(intervals) - 1):
        if intervals[i][1] > intervals[i + 1][0]:
            return False
    
    return True
```

```python
def canAttendMeetings(intervals):
    intervals.sort()
    lastEnd = -1
    for inter in intervals:
        if inter[0] < lastEnd:
            return False
        lastEnd = inter[1]
    
    return True
```

```python
import heapq
def canAttendMeetingsII(intervals):
    intervals.sort()
    res = []
    heapq.heappush(res, intervals[0][1])
    
    for i in range(1, len(intervals)):
        if intervals[i][0] >= res[0]:
            heapq.heappushpop(res, intervals[i][1])
        else:
            heapq.heappush(res, intervals[i][1])
    
    return len(res)
```

```python
def canAttendMeetingsII(intervals):
    start = sorted([i.start for i in intervals])
    end = sorted([i.end for i in intervals])
    
    res = count = 0
    s = e = 0
    while s < len(intervals):
        if start[s] < end[e]:
            s += 1
            count += 1
        else:
            e += 1
            count -= 1
        # is this necessary, try break after start run out
        res = max(res, count)
    
    return res
```
