```python
def carFleet(target, pos, speed):
    time = [float(target - p) / s for p, s in sorted(zip(pos, speed))]
    res = cur = 0
    for t in time[::-1]:
        if t > cur:
            res += 1
            cur = t
    return res
```

```python
target = 12
position = [10,8,0,5,3,11]
speed = [2,4,1,1,3,3]
carFleet(target, position, speed)
```

```python
target = 10
position = [3]
speed = [3]
carFleet(target, position, speed)
```
