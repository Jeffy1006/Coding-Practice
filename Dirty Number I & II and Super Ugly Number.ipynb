```python
def dirtyNumber(n):
    
    while n != 1:
        if n % 2 == 0:
            n /= 2
        elif n % 3 == 0:
            n /= 3
        elif n % 5 == 0:
            n /= 5
        else:
            return False
    
    return True
```

```python
def dirtyNumberII(n):
    count = 0
    i = 1
    while count != n:
        if dirtyNumber(i):
            count += 1
        i += 1
    
    return i - 1
```

```python
def dirtyNumberII(n):
    count = 0
    hashmap = set()
    
    def helper(num):
        nonlocal hashmap
        temp = num
        while temp != 1:
            if temp < num:
                if temp in hashmap: return True
                else: return False
                
            if temp % 2 == 0: temp /= 2
            elif temp % 3 == 0: temp /= 3
            elif temp % 5 == 0: temp /= 5
            else: return False
    
        return True
    
    i = 1
    while count != n:
        if helper(i):
            hashmap.add(i)
            count += 1
        i += 1
    
    return i - 1
```

```python
def dirtyNumberII(n):
    count = 0
    hashmap = set([1, 2, 3, 5])
    
    i = 1
    while count != n:
        if i == 1 or i == 2 or i ==3 or i == 5:
            count += 1
            i += 1
            continue
            
        if i % 2 == 0 and i / 2 in hashmap:
                hashmap.add(i)
                count += 1
        elif i % 3 == 0 and i / 3 in hashmap:
                hashmap.add(i)
                count += 1
        elif i % 5 == 0 and i / 5 in hashmap:
                hashmap.add(i)
                count += 1
        i += 1

    return i - 1
```

```python
def superUglyNumber(n, primes):
    count = 0
    hashmap = set([1] + primes)
    
    i = 1
    while count != n:
        if i in hashmap:
            count += 1
            i += 1
        else:
            for j in primes:
                if i % j == 0 and i / j in hashmap:
                    hashmap.add(i)
                    count += 1
                    break
            i += 1

    return i - 1
```

```python
import heapq
def superUglyNumber(n, primes):
    ugly_number = [1]
    pq = [[p, 0, p] for p in primes]
    seen = {1}
    while n - 1:
        num = pq[0][0]
        ugly_number.append(num)
        while pq[0][0] == num:
            t = heapq.pop(pq)
            t[1] += 1
            t[0] = ugly_number[t[1]] * t[2]
            heapq.push(pq, t)
        n -= 1
    return ugly_number[-1]
```
