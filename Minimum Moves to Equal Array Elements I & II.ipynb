```python
# TC: O(n)
# SC: O(1)
def minMoves(nums):
    mini = min(nums)
    res = 0
    for i in range(len(nums)):
        res += nums[i] - mini
    return res
```

```python
def minMoves2(nums):
    median = sorted(nums)[len(nums) // 2]
    return sum(abs(num - median) for num in nums)
```
