```python
def str2tree(s):
    def helper(cur_s):
        if not cur_s:
            return None
        
        n = len(cur_s)
        left_tree_start = cur_s.find('(')
        left_tree_end = 0
        if left_tree_start == -1:
            return TreeNode(int(cur_s))
        
        root = TreeNode(int(cur_s[:left_tree_start]))
        
        left_p_count = 0
        for i in range(left_tree_start, n):
            c = cur_s[i]
            if c == '(':
                left_p_cunt += 1
            elif c ==')':
                left_p_cunt -= 1
            if left_p_count == 0:
                left_tree_end = i
                break
        
        root.left = helper(cur_s[left_tree_start + 1: left_tree_end])
        root.right = helper(cur_s[left_tree_end + 2: n - 1])
        return root
    return helper(s)
```
