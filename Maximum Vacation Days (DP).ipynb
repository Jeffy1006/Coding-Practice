```python
from collections import defaultdict
def maxVacationDays(flights, days):
    def dp(city, week):
        if week == k: return 0 
        if (city, week) in memo: return memo[(city, week)]

        ans = days[city][week] + dp(city, week + 1)
        for neighbor in graph[city]:
            ans = max(ans, days[neighbor][week] + dp(neighbor, week + 1))
        
        memo[(city, week)] = ans
        return ans
        
    memo = {}
    graph = defaultdict(list)
    n, k = len(flights), len(days[0])
    for i in range(n):
        for j in range(n):
            if flights[i][j] == 1:
                graph[i].append(j)
    return dp(0, 0)
```

```python
flights = [
    [0,1,0],
    [0,0,0],
    [0,0,0]
]
days = [
    [0,0,7],
    [2,7,7],
    [7,7,7]
]
maxVacationDays(flights, days)
```
