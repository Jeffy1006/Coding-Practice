```python
def findLength(nums1, nums2):
    m, n = len(nums1), len(nums2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    maximum = 0
    for i in range(m):
        for j in range(n):
            if nums1[i] == nums2[j]:
                dp[i + 1][j + 1] = dp[i][j] + 1
                maximum = max(maximum, dp[i + 1][j + 1])
    return maximum
```

```python
def findLength(nums1, nums2):
    maxlen, prev = 0, [0] * (len(nums2)+1)

    for i in range(len(nums1)):
        curr = [0] * (len(nums2)+1)  
          
        for j in range(len(nums2)):
            if nums1[i] == nums2[j]: 
                curr[j+1] = prev[j] + 1
                maxlen = max(maxlen, curr[j+1])
                    
        prev = curr
            
    return maxlen
```

```python
nums1 = [1,2,3,2,1] 
nums2 = [3,2,1,4,7]
findLength(nums1, nums2)
```
