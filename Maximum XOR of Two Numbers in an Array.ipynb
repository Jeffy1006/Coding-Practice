```python
def findMaximumXOR(nums):
    mask = MAX = 0
    for i in range(31, -1, -1):
        mask = mask | (1 << i)
        hashset = set()
        for num in nums:
            hashset.add(mask & num)
        
        tmp = MAX | (1 << i)
        for prefix in hashset:
            if (tmp ^ prefix) in hashset:
                MAX = tmp
                break
    return MAX
```
