```python
from functools import reduce
def maxRotateFunction(nums):
    n = len(nums)
    if n <= 1: return 0
    
    sum_nums = sum(nums)
    # first argument: function 
    # second argument: range of array length
    # third argument: start value of accumulator
    prev_F = reduce(lambda accumulator, it: accumulator + nums[it] * it, range(n), 0)
    
    ans = prev_F
    for i in range(n - 1, 0, -1):
        tmp = prev_F + sum_nums - (n * nums[i])
        ans = max(ans, tmp)
        prev_F = tmp
    return ans
```

```python
maxRotateFunction([4,3,2,6])
```
