```python
def uniquePaths(m, n):
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    dp[m - 1][n - 1] = 1

    for i in range(m - 1, -1, -1):
        for j in range(n - 1, -1, -1):
            dp[i][j] += dp[i + 1][j] + dp[i][j + 1]
    
    return dp[0][0]
```

```python
def uniquePaths(m, n):
    dp = [0] * (n + 1)
    dp[n - 1] = 1

    for _ in range(m):
        new = [0] * (n + 1)
        for i in range(n - 1, -1, -1):
            new[i] += new[i + 1] + dp[i]
        dp = new
    
    return dp[0]
```

```python
def uniquePaths(m, n):
    row = [1] * n
    
    for i in range(m - 1):
        newRow = [1] * n
        for j in range(n - 2, -1, -1):
            newRow[j] = newRow[j + 1] + row[j]
        row = newRow
    
    return row[0]
```
