```python
def exclusiveTime(n, logs):
    res = [0] * n
    logs = [log.split(':') for log in logs]
    for i in range(len(logs)):
        logs[i][0] = int(logs[i][0])
        logs[i][2] = int(logs[i][2])
        
    stack = []
    for log in logs:
        if not stack: 
            stack.append(log)
        elif log[1] == 'start':
            res[stack[-1][0]] += log[2] - stack[-1][2]
            stack.append(log)
        else:
            res[log[0]] += log[2] - stack[-1][2] + 1
            stack.pop()
            if stack:
                stack[-1][2] = log[2] + 1
    return res
```

```python
def exclusiveTime(n, logs):
    ans = [0] * n
    stack = []
    prev_time = 0

    for log in logs:
        fn, typ, time = log.split(':')
        fn, time = int(fn), int(time)

        if typ == 'start':
            if stack:
                ans[stack[-1]] += time - prev_time 
            stack.append(fn)
            prev_time = time
        else:
            ans[stack.pop()] += time - prev_time + 1
            prev_time = time + 1

    return ans
```

```python
n = 2 
logs = ["0:start:0","1:start:2","1:end:5","0:end:6"]
exclusiveTime(n, logs)
```
