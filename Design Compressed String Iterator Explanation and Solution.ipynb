```python
class StringIterator:
    def __init__(self, compressedString):
        self.index = 0
        self.count = 0
        self.s = compressedString
        self.ch = ' '
    
    def next(self):
        if not self.hasNext(): return ' '
        if self.count == 0:
            self.ch = self.s[self.index]
            self.index += 1
            while self.index < len(self.s) and self.s[self.index].isdigit():
                self.count = self.count * 10 + int(self.s[self.index])
                self.index += 1
        self.count -= 1
        return self.ch
    
    def hasNext(self):
        return self.index < len(self.s) or self.count > 0
```

```python
a = StringIterator("L1e2t1C1o1d1e1")
```
