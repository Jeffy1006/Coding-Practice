```python
import heapq
def getOrder(tasks):
    tasks = [(task[0], task[1], i) for i, task in enumerate(tasks)]
    heapq.heapify(tasks)
    available = []
    t = 0
    res = []
    
    while tasks or available:
        if available:
            process, index = heapq.heappop(available)
            res.append(index)
            t +=  process
        else:
            enqueue, process, index = heapq.heappop(tasks)
            res.append(index)
            t =  max(t, enqueue) + process
        
        while tasks and tasks[0][0] <= t:
            enqueue, process, index = heapq.heappop(tasks)
            heapq.heappush(available, (process, index))
        
    return res
```

```python
class Solution:
    def getOrder(self, tasks: List[List[int]]) -> List[int]:
        res = []
        tasks = sorted([(t[0], t[1], i) for i, t in enumerate(tasks)])
        i = 0
        h = []
        time = tasks[0][0]
        while len(res) < len(tasks):
            while (i < len(tasks)) and (tasks[i][0] <= time):
                heapq.heappush(h, (tasks[i][1], tasks[i][2])) # (processing_time, original_index)
                i += 1
            if h:
                t_diff, original_index = heapq.heappop(h)
                time += t_diff
                res.append(original_index)
            elif i < len(tasks):
                time = tasks[i][0]
        return res
```
