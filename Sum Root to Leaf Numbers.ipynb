```python
class TreeNode:
    def __init__(self, data):
        self.val = data
        self.left = None
        self.right = None
        
def sumRootToLeaf(root):
        if not root:
            return 0
        
        def dfs(node, cand):
            if not node:
                return 0
            if not node.left and not node.right:
                cand += str(node.val)
                return int(cand)
            
            cand += str(node.val)
            
            return dfs(node.left,cand) + dfs(node.right,cand)
        
        return dfs(root,'')
            
```

```python
a = TreeNode(4)
b = TreeNode(9)
c = TreeNode(0)
d = TreeNode(5)
e = TreeNode(1)
a.left, a.right = b, c
b.left, b.right = d, e
```

```python
s = sumRootToLeaf(a)
s
```
