```python
def findRotateSteps(ring, key):
    mem = {}
    n = len(ring)
    def helper(i, k):
        if k == '': return 0
        if (i, k) in mem:
            return mem[(i, k)]
        
        cw, ccw, cur = 1, 1, i
        while ring[cur % n] != k[0]:
            cur += 1
            cw += 1
        cw += helper(cur % n, k[1:])
        cur = i
        while ring[cur % n] != k[0]:
            cur -= 1
            ccw += 1
        ccw += helper(cur % n, k[1:])
        mem[(i, k)] = min(cw, ccw)
        return mem[(i, k)]

    return helper(0, key)
```

```python
def findRotateSteps(ring, key):
    mem = {}
    def helper(ring, key, idx):
        if idx == len(key): return 0
        if (ring, idx) in mem:
            return mem[(ring, idx)]
        
        res = float('inf')
        for i in range(len(ring)):
            if ring[i] == key[idx]:
                steps = min(i, len(ring) - i)
                res = min(res, steps + 1 + helper(ring[i:] + ring[0:i], key, idx + 1))
        mem[(ring, idx)] = res
        return res

    return helper(ring, key, 0)
```
