```python
def countPrime(num):
    if num < 2:
        return 0
    
    res = []
    if num % 2 == 0: num -= 1

    for i in range(num,2,-2):
        factor = i ** 0.5
        flag = True
        for j in range(3,factor + 1):
            if i % j == 0:
                flag = False
                break
        if flag:
            res.append(i)
    
    return res + [2]
```

```python
def countPrime(num):
    if num < 2:
        return 0
    
    res = [2]
    if num % 2 == 1: num += 1

    for i in range(3, num, 2):
        factor = i ** 0.5
        flag = True
        for j in range(3, factor + 1):
            if i % j == 0:
                flag = False
                break
        if flag:
            res.append(i)
    
    return res
```

```python
# TC: O(nlglgn)
def countPrime(n):
    primes = [True] * (n + 1)
    
    i = 2
    while i * i < len(primes):
        if primes[i]:
            j = i
            while j * i < len(primes):
                primes[i * j] = False
                j += 1
        i += 1
    
    return [i for i in range(2, n + 1) if primes[i]]
```
