```python
def twoSum(nums, target):
    hashmap = {}
    for i in range(len(nums)):
        diff = target - nums[i]
        if diff in hashmap:
            return (hashmap[diff], i)
        hashmap[nums[i]] = i
    return None
```

```python
twoSum([2,7,11,15], 9)
```

```python
twoSum([3,2,4], 6)
```

```python
twoSum([3,3], 6)
```
