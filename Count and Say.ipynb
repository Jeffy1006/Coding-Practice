```python
import itertools
def countAndSay(n):
    s = '1'
    for _ in range(n - 1):
        s = ''.join(str(len(list(group))) + digit for digit, group in itertools.groupby(s))
    
    return s
```

```python
def countAndSay(n):
    def helper(s):
        ans = ''
        count = 1
        for i in range(1, len(s)):
            if s[i] == s[i - 1]:
                count += 1
            else:
                ans = ans + str(count) + s[i - 1]
                count = 1
        return ans + str(count) + s[-1]
    
    s = '1'
    for _ in range(n - 1):
        s = helper(s)
    return s
```

```python
sample1 = []
for i in range(1, 10):
    sample1.append(countAndSay(i))
```

```python
sample2 = []
for i in range(1, 10):
    sample2.append(countAndSay(i))
```
