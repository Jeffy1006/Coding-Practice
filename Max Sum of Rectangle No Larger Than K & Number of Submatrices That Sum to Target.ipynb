```python
import bisect
def maxSumSunmatrix(matrix, k):
    def getMaxSum(arr):
        sortedPreSum = [0]
        ans = float('-inf')
        runningSum = 0
        for num in arr:
            runningSum += num
            targetSum = runningSum - k
            index = bisect.bisect_left(sortedPreSum, targetSum)
            if index < len(sortedPreSum):
                ans = max(ans, runningSum - sortedPreSum[index])
            if ans == k:
                break
            bisect.insort(sortedPreSum, runningSum)
        return ans
        
    m, n = len(matrix), len(matrix[0])
    preSums = []
    
    for i in range(m):
        rowPreSum = []
        for j in range(n):
            if j == 0:
                rowPreSum.append(matrix[i][j])
            else:
                rowPreSum.append(matrix[i][j] + rowPreSum[-1]) 
        preSums.append(rowPreSum)
    
    ans = float('-inf') 
    for c2 in range(n):
        for c1 in range(c2 + 1):
            curArr = []
            for i in range(m):
                rowSum = preSums[i][c2] - (preSum[i][c1 - 1] if c1 > 0 else 0)
                curArr.append(rowSum)
            ans = max(ans, getMaxSum(curArr))
            if ans == k:
                return k
    return ans
```
