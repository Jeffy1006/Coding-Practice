```python
def candy(ratings):
    res = [1]*len(ratings)
    for i in range(1, len(ratings)):
        if ratings[i] > ratings[i-1]:
            res[i] = res[i-1] + 1
    for i in range(len(ratings)-2, -1, -1):
        if ratings[i] > ratings[i+1]:
            res[i] = max(res[i], res[i+1] + 1)
    print(res)
    return sum(res)
```

```python
candy([1,0,2,3,4,2,1,3])
```
