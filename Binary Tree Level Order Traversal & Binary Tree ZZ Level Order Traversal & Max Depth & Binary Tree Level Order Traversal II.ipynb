```python
from collections import deque
def levelOrder(root):
    res = []
    
    q = deque([root])
    
    while q:
        level = []
        for i in range(len(q)):
            node = q.popleft()
            if node:
                level.append(node.val)
                q.append(node.left)
                q.append(node.right)
        if level:
            res.append(level)
    
    return res
```
