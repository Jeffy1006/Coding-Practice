```python
from collections import deque
def findMaxDepth(root):
    if not root: return 0
    queue = deque([root])
    depth = 0
    while queue:
        depth += 1
        for _ in range(len(queue)):
            node = queue.popleft()
            for child in node.children:
                queue.append(child)
    return depth
```

```python
def findMaxDepth(root):
    if not root: return 0
    depth = 0
    for child in root.children:
        depth = max(depth, findMaxDepth(child))
    return depth + 1
```
