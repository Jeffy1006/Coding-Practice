```python
def reorderLL(head):
    if not head: return None
    
    slow, fast = head. head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
    second = slow.next
    prev = slow.next = None
    
    while second:
        tmp = second.next
        second.next = prev
        prev = second
        second = tmp
    
    first, second = head, prev
    while second:
        tmp1, tmp2 = first.next, second.next
        first.next = second
        second.next = tmp1
        first = tmp1
        second = tmp2
    
    return head
```
