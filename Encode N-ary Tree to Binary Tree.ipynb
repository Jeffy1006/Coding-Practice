```python
def encode(root):
    if not root: return None
    val = root.val
    treeRoot = TreeNode(val)
    if not root.children: return treeRoot
    treeRoot.left = encode(root.children[0])
    cur = treeRoot.left
    for i in range(1, len(root.children)):
        cur.right = encode(root.children[i])
        cur = cur.right
    return TreeRoot

def decode(root):
    if not root: return None
    nRoot = Node()
    nRoot.val = root.val
    cur = root.left
    while cur:
        nRoot.children.append(decode(cur))
        cur = cur.right
    return nRoot
```
