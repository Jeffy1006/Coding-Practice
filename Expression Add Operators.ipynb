```python
def addOperators(num, target):
    solution = []
    
    def dfs(candidate, i, total, prev_add):
        if i == len(num) and total == target: solution.append(candidate)
        # len(num) + 1 because j will be excluded in list dissecting, 
        # so in this case len(num) will be excluded but not len(num) - 1 
        for j in range(i + 1, len(num) + 1):
            s = num[i:j]
            d = int(s)
            # omit leading 0's except "0"
            if num[i] == '0' and s != '0': continue
            # initial step
            if not candidate:
                dfs(s, j, d, d)
            else:
                dfs(candidate + "+" + s, j, total + d, d)
                dfs(candidate + "-" + s, j, total - d, -d)
                dfs(candidate + "*" + s, j, total - prev_add + prev_add * d, prev_add * d)
                
    
    dfs("",0,0,0)
    return solution
```
