```python
def readBinaryWatch(num):
    res = []
    for hour in range(12):
        for minute in range(60):
            if bin(hour).count('1') + bin(minute).count('1') == num:
                res.append("{}:{:02d}".format(hour, minute))   
    return res
```

```python
readBinaryWatch(1)
```
