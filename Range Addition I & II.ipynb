```python
def getModifiedArray(length, updates):
    ans = [0] * (length + 1)
    for i, j, k in updates:
        ans[i] += k
        ans[j + 1] -= k
    for i in range(1, length + 1):
        ans[i] += ans[i - 1]
    return ans[:-1]
```

```python
getModifiedArray(5, [[1,3,2],[2,4,3],[0,2,-2]])
```

```python
def maxCount(m, n, ops):
    if not ops: return m * n
    min_m, min_n = m, n
    for op in ops:
        min_m, min_n = min(min_m, ops[0]), min(min_n, ops[1])
    return min_m * min_n
```

```python
def maxCount(m, n, ops):
    if not ops: return m * n
    return min(op[0] for op in ops) * min(op[1] for op in ops)
```

```python
maxCount(3, 3, [[2,2],[3,3]])
```
