```python
def flipGame(s):
    res, i = [], 0
    
    while i < len(s):
        try:
            nextmove = s.index("++") if i == 0 else s.index("++", i)
        except ValueError:
            nextmove = -1
        
        if nextmove == -1:
            return res
        
        res.append(s[:nextmove] + "--" + s[nextmove + 2:])
        i = nextmove + 1
    
    return res
```

```python
def canWin(s):
    if not s or len(s) < 2: 
        return False
    
    for i in range(len(s) - 1):
        if s[i] == "+" and s[i + 1] == "+":
            nextState = s[:i] + "--" + s[i + 2:]
            if not canWin(nextState):
                return True   
    return False
```
