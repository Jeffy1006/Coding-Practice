```python
class Solution:
    def isMatch(self, s, p):
        dp = [[False for _ in range(len(p)+1)] for i in range(len(s)+1)]
        dp[0][0] = True
        # The below lopp set the beginning *s True since * can match "" 
        for j in range(1, len(p)+1):
            if p[j-1] != '*':
                break
            dp[0][j] = True
                
        for i in range(1, len(s)+1):
            for j in range(1, len(p)+1):
                if p[j-1] in {s[i-1], '?'}:
                    dp[i][j] = dp[i-1][j-1]
                elif p[j-1] == '*':
                    dp[i][j] = dp[i-1][j] or dp[i][j-1]
        return dp[-1][-1]
```

```python
# @return a boolean
def isMatch(s, p):
    length = len(s)
    if len(p) - p.count('*') > length:
        return False
    dp = [True] + [False]*length
    for i in p:
        if i != '*':
            for n in reversed(range(length)):
                dp[n+1] = dp[n] and (i == s[n] or i == '?')
        else:
            for n in range(1, length+1):
                dp[n] = dp[n-1] or dp[n]
        dp[0] = dp[0] and i == '*'
    return dp[-1]
```

```python
isMatch(s, p):
    spt, ppt = 0, 0
    match, startIdx = 0, -1
    
    while spt < len(s):
        if ppt < len(p) and (p[ppt] == '?' or p[ppt] == s[spt]):
            spt += 1
            ppt += 1
        elif ppt < len(p) and p[ppt] == '*':
            startIdx = ppt
            match = spt
            ppt += 1
        elif startIdx != -1:
            ppt = startIdx + 1
            match += 1
            spt = match
        else:
            return False
    
    while ppt < len(p) and p[ppt] == '*':
        ppt += 1
    
    return ppt = len(p)
```
