```python
import math
def bulbSwitcherI(n):
    return floor(math.sqrt(n))
```

```python
def bulbSwitcherII(n, m):
    realN = min(3, n)
    if m == 0: return 1
    elif m == 1: return 2 if realN == 1 else 3 if realN == 2 else 4
    elif m == 2: return 2 if realN == 1 else 4 if realN == 2 else 7
    
    return 2 if realN == 1 else 4 if realN == 2 else 8
```

```python
def bulbSwitcherIII(light):
    res = maxLight = 0
    cur = 1
    left = {}
    for l in light:
        left[l] = True
        maxLight = max(maxLight, l)
        if l >= cur:
            while left[cur]:
                cur += 1
            if cur - 1 == maxLight:
                res += 1
    return res
```
