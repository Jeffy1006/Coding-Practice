```python
def pascalTri(n):
    res = []
    if n == 0:
        return res
    
    res.append([1])
    if n == 1:
        return res
    
    res.append([1,1])
    if n == 2:
        return res
    
    for i in range(3,n+1):
        tmp = [1]*i
        for j in range(1,i-1):
            tmp[j] = res[-1][j-1] + res[-1][j]
        res.append(tmp)
    
    return res
```

```python
def generate(numRows):
    res = [[1]]
    
    for i in range(1, numRows):
        temp = [1]
        for j in range(i - 1):
            temp.append(res[-1][j] + res[-1][j + 1])
        temp.append(1)
        res.append(temp)
    
    return res
```

```python
def pascalTriII(n):
    res = [1]
    if n == 0:
        return res
    
    res.append(1)
    if n == 1:
        return res
    
    for i in range(2,n+1):
        tmp = [1,1]
        for j in range(1,i):
            tmp.insert(j,res[j-1] + res[j])
        # res = tmp must be put outside the for loop because res is updated only after tmp is completed
        res = tmp

    
    return res
```
