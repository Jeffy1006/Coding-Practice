```python
from collections import defaultdict
def alertNames(keyName, keyTime):
    hashmap = defaultdict(list)
    times = []
    
    for i in range(len(keyName)):
        hour, minute = map(int, keyTime[i].split(':'))
        hashmap[keyName[i]].append(hour + minute)
    
    res = []
    for name, time in hashmap.items():
        time.sort()
        for i in range(len(time) - 3 + 1):
            if time[i + 2] - time[i] <= 60:
                res.append(name)
                break
    
    return sorted(res)
```

```python
keyName = ["leslie","leslie","leslie","clare","clare","clare","clare"]
keyTime = ["13:00","13:20","14:00","18:00","18:51","19:30","19:49"]
alertNames(keyName, keyTime)
```
