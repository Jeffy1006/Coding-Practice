```python
def combSum(candidates, target):
    res = []
    
    def dfs(i, cur, total):
        if total == target:
            res.append(cur.copy())
            return
        if i >= len(candidates) or total > target:
            return
        
        cur.append(candidates[i])
        dfs(i, cur, total + candidates[i])
        cur.pop()
        dfs(i + 1, cur, total)
        
        return
    
    dfs(0, [], 0)
    
    return res
```

```python
combSum([1,2,3], 4)
```

```python
def combSumII(candidates, target):
    res = []
    candidates.sort()
    
    def dfs(pos, cur, t):
        if t == 0:
            res.append(cur.copy())
            return
        if t < 0:
            return
        
        prev = -1
        for i in range(pos, len(candidates)):
            if candidates[i] == prev:
                continue
            cur.append(candidates[i])
            dfs(i + 1, cur, t - candidates[i])
            cur.pop()
            prev = candidates[i]
    
    dfs(0, [], target)
    return res
```

```python
combSumII([1,2,3], 4)
```

```python
def combSumIII(k, n):
    res = []
    hash_set = set()
    
    def dfs(i, cur, total):
        if total == n:
            for item in cur:
                if item in hash_set:
                    return
            for item in cur:
                hash_set.add(item)
            res.append(cur.copy())
            return
        if len(cur) > k or i > 9 or total > n:
            return
        

        cur.append(i)
        dfs(i + 1, cur, total + i)
        cur.pop()
        dfs(i + 1, cur, total)
        
        return
    
    dfs(1, [], 0)
    
    return res
```

```python
def combSumIII(k, n):
    res = []
    
    def helper(start, k_remain, n_remain, cand):
        if k_remain < 0 or n_remain < 0:
            return
        if k_remain == 0 and n_remain == 0:
            res.append(cand.copy())
            return
        
        for i in range(start, 10):
            cand.append(i)
            helper(i + 1, k_remain - 1, n_remain - i, cand)
            cand.pop()
    
    helper(1, k, n, [])
    return res
```

```python
def combSumIV(nums, target):
    dp = {0:1}
    for total in range(1, target + 1):
        dp [total] = 0
        for n in nums:
            dp[total] += dp.get(total - n, 0)
    return dp[target]
```
