```python
def removeDupArr(arr):
    
    if len(arr) < 2:
        return len(arr)

    index = 0
    for i in range(1,len(arr)):
        if arr[i] != arr[i-1]:            
            index += 1
            arr[index] = arr[i]
#          index+1 not index or index-1 
    return index+1

def comb(res, k, cand, index, arr, start, end):
    
    if index == k:
        res.append([x for x in cand])
        return
    
    if start == end:
        return
    
    cand[index] = arr[start]
    comb(res, k, cand, index+1, arr, start+1, end)
    comb(res, k, cand, index, arr, start+1, end)
    
    return

def combWDup(arr):
    
    if len(arr) < 2:
        return arr
    
    res = [[]]
    arr.sort()
    n = removeDupArr(arr)
#                    n+1 instea of n
    for k in range(1,n+1):
        cand = [0]*k
        comb(res, k, cand, 0, arr, 0, n)
    
    return res
```

```python
a = combWDup([1,4,2,3])
```
