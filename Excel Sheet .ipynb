```python
import string
def convertToNumber(title):
    letter = string.ascii_letters[26:]
    letterToNum = {}
    for i in range(26):
        letterToNum[letter[i]] = i + 1
        
    ans = 0
    factor = 1
    for i in range(len(title) - 1, -1, -1):
        num = letterToNum[title[i]]
        ans = ans + num * factor
        factor *= 26

    return ans
```

```python
def convertToNumber(s):
    res = 0
    
    for c in s:
        d = ord(c) - ord('A') + 1
        res = res * 26 + d
      
    return res
```
