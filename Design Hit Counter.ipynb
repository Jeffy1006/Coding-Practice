```python
class HitCounter:
    def __init__(self):
        self.time = [0] * 300
        self.count = [0] * 300
    
    
    def hit(self, timestamp):
        index = timestamp % 300
        if self.time[index] != timestamp:
            self.time[index] = timestamp
            self.count[index] = 1
        else:
            self.count[index] += 1
    
    def getHits(self, timestamp):
        ans = 0
        for i in range(300):
            if self.time[i] > timestamp - 300:
                ans += self.count[i]
        return ans
```
