```python
def bullsCows(secret, guess):
    hashmap = {}
    
    bulls = 0
    for i in range(len(secret)):
        if secret[i] == guess[i]:
            bulls += 1
        if secret[i] in hashmap:
            hashmap[secret[i]] += 1
        else:
            hashmap[secret[i]] = 1
    
    cows = 0
    for c in guess:
        if c in hashmap:
            cows += 1
            hashmap[c] -= 1
            if hashmap[c] == 0:
                del hashmap[c]
    
    return str(bulls) + "A" + str(cows - bulls) + "B"
```

```python
secret = "1"
guess = "1"
bullsCows(secret, guess)
```
