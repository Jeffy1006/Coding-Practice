```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        
def removeLLElements(head, val):
    if not head:
        return head
    
    temp_head = Node(0)
    temp_head.next = head
    prev, cur = temp_head, head
    
    
    while cur:
        if cur.val == val:
            prev.next = cur.next
            cur = prev.next
        else:
            cur = cur.next
            prev = prev.next
    
    return temp_head.next
```

```python
a = [1,2,6,3,4,5,6]
h = Node(0)
c = h
for x in a:
    c.next = Node(x)
    c = c.next
h = h.next
c = removeLLElements(h, 6)
if c == None:
    print([])
while c:
    print(c.val, end=" ")
    c = c.next
```
