```python
def shuffle(nums, n):
    i, j = n - 1, 2 * n - 1
    while j >= n:
        nums[j] <<= 10
        nums[j] |= nums[i]
        i -= 1
        j -= 1
    
    i, j = 0, n
    while j < 2 * n - 1:
        num1 = nums[j] & 1023
        num2 = nums[j] >> 10
        nums[i], nums[i + 1] = num1, num2
        i += 1
    
    return nums
```

```python
import random
class Solution:
    def __init__(self, nums):
        self.array = nums
        self.original = nums[:]
    
    def reset(self):
        self.array = self.original[:]
        return self.array
    
    def shuffle(self):
        for i in range(len(self.array)):
            swap_idx = random.randrange(i, len(self.array))
            self.array[i], self.array[swap_idx] = self.array[swap_idx], self.array[i]
        return self.array
```
