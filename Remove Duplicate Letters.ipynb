```python
def removeDuplicateLetters(s):
    sLen = len(s)
    lastIndex = {}
    for i in range(sLen):
        lastIndex[s[i]]  = i 
    stack = []
    visited = set()
    for i in range(sLen):
        if s[i] in visited: continue
        while stack and stack[-1] > s[i] and lastIndex[stack[-1]] > i:
            top = stack.pop()
            visited.remove(top)
        stack.append(s[i])
        visited.add(s[i])
    return "".join(stack)
```

```python
removeDuplicateLetters("cbacdbcb")
```
