```python
def containsDup(nums):
    hashSet = set()
    for n in nums:
        if n in hashSet:
            return True
        else:
            hashSet.add(n)
    
    return False
```

```python
def containsDupII(nums, k):
    
    def helper(start, end):
        hashSet = set()
        for i in range(start, end + 1):
            if nums[i] in hashSet:
                return True
            else:
                hashSet.add(nums[i])
                
        return False
    # k inclusive
    for i in range(0,len(nums) - k):
        if helper(i, i + k):
            return True
    
    return False
```

```python
# Space O(1) 
def containsDupII(nums, k):
    
    def helper(start, end):
        for i in range(start + 1, end + 1):
            if i >= len(nums): return False
            if nums[start] == nums[i]:
                return True
                
        return False
    # k inclusive
    for i in range(len(nums)):
        if helper(i, i + k):
            return True
    
    return False
```

```python
def containsDupII(nums, k):
    
    hashMap = {}
    for i in range(len(nums)):
        cur = nums[i]
        if cur in hashMap and abs(i - hashMap[cur]) <= k:
            return True
        else:
            hashMap[cur] = i
    
    return False
```

```python
def containsDupIII(nums, k, t):
    
    def helper(start, end):
        for i in range(start + 1, end + 1):
            if i >= len(nums): return False
            if abs(nums[start] - nums[i]) <= t:
                return True
        return False
    # k inclusive
    for i in range(len(nums)):
        if helper(i, i + k):
            return True
        
    return False
```

```python
def containsDupIII(nums, k, t):
    
    n = len(nums)
    if t == 0 and n == len(set(nums)):
        return False
    
    for i in range(n):
        for j in range(i + 1, i + k + 1):
            if j >= n : break
            if abs(nums[i] - nums[j]) <= t:
                return True
    
    return False
```

```python
def containsDupIII(nums, k, t):
    if t < 0: return False
    n = len(nums)
    d = {}
    w = t + 1
    for i in range(n):
        m = nums[i] // w
        if m in d:
            return True
        if m - 1 in d and abs(nums[i] - d[m - 1]) < w:
            return True
        if m + 1 in d and abs(nums[i] - d[m + 1]) < w:
            return True
        d[m] = nums[i]
        if i >= k: del d[nums[i - k] // w]
    return False
```
