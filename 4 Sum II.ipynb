```python
# TC: O(n^2)
# SC: O(n^2)
def fourSumII(nums1, nums2, nums3, nums4):
    hashmap = {}
    for a in nums1:
        for b in nums2:
            hashmap[a + b] = hashmap.get(a + b, 0) + 1
    ans = 0
    for c in nums3:
        for d in nums4:
            if -(c + d) in hashmap:
                ans += hashmap[-(c + d)]
    return ans
```
