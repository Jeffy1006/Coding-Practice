```python
class TicTacToe(object):
    
    def __init__(self, n):
        self.board = [[0] * n for i in range(n)]
        self.n = n
    
    def move(self, row, col, player):
        self.board[row][col] = player
        
        n = self.n
        matrix = self.board
        
        def checkRow(r):
            for c in range(n):
                if matrix[r][c] != player: return False
            return True
        
        def checkCol(c):
            for r in range(n):
                if matrix[r][c] != player: return False
            return True
        
        def checkDiag():
            for r in range(n):
                if matrix[r][r] != player: return False
        
        def checkAntiDiag():
            for r in range(n):
                if matrix[r][n - r - 1] != player: return False
        
        for r in range(n):
            if checkRow(r): return player
        
        for c in range(n):
            if checkRow(c): return player
        
        if checkDiag(): return player
        if checkAntDiag(): return player
```

```python
class TicTacToe(object):
    
    def __init__(self, n):
        self.size = n
        self.rows = [0] * n
        self.cols = [0] * n
        self.diag = 0
        self.Adiag = 0
    
    def move(self, row, col, player):
        if player == 2:
            player = -1
            
        self.rows[row] += player
        if abs(self.rows[row]) == self.size:
            return player if player == 1 else 2
        
        self.cols[col] += player
        if abs(self.cols[col]) == self.size:
            return player if player == 1 else 2
        
        if row == col:
            self.diag += player
        if abs(self.diag)  == self.size:
            return player if player == 1 else 2
        
        if row + col == self.size - 1:
            self.Adiag += player
        if abs(self.Adiag)  == self.size:
            return player if player == 1 else 2    
```
