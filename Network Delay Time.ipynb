```python
from collections import defaultdict
import heapq
def networkDelayTime(times, n, k):
    edges = defaultdict(list)
    for u, v, w in times:
        edges[u].append((v, w))
    
    minHeap = [(0, k)]
    visited = set()
    res = 0
    
    while minHeap:
        time, node = heapq.heappop(minHeap)
        if node in visited: 
            continue
        visited.add(node)
        res = time
        
        for w, v in edges[node]:
            heapq.heappush(minHeap, (w + time, v))
    
    return res if len(set) == n else -1
```
