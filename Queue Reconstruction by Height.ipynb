```python
def reconstructQueue(people):
    n = len(people)
    people.sort()
    ans = [[-1, -1]] * n
    for i in range(n):
        count = people[i][1]
        for j in range(n):
            if ans[j][0] == -1 and count == 0:
                ans[j][0] = people[i][0]
                ans[j][1] = people[i][1]
                break
            elif ans[j][0] == -1 or ans[j][0] >= people[i][0]:
                count -= 1
    return ans
```
