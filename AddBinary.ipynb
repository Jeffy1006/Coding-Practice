```python
def AddBinary(a,b):
    long = ''
    short = ''
    a = list(a)
    b = list(b)
    result = ''
    
    if len(a) > len(b):
        long = a
        short = b
    else:
        long = b
        short = a
    
    carry = 0
    steps = len(long) - len(short)
    for i in range(len(short)-1,-1,-1):
        tmp = int(long[i+steps]) + int(short[i]) + carry
        if tmp > 1:
            carry = 1
        else:
            carry = 0
        if tmp == 0 or tmp == 2:
            result = '0' + result
        elif tmp == 1 or tmp == 3:
            result = '1' + result
            
     
    for i in range(steps-1, -1, -1):
        if carry == 1: 
            if long[i] == '0':
                result = '1' + result
                carry = 0
            else:
                result = '0' + result
        else:
            result = long[i] + result
    
    if carry == 1:
        result = '1' + result
    
    return result
```

```python
x = '1101'
y = '101110100111'
print(AddBinary(x,y))
```
