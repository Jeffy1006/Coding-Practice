```python
def removeKdigits(num, k):
    n = len(num)
    if n <= k: return "0"
    
    stack = []
    for c in num:
        while stack and k > 0 and stack[-1] > int(c):
            stack.pop()
            k -= 1
        if stack or c != '0':
            stack.append(int(c))
    
    while stack and k:
        stack.pop()
    if not stack:
        return "0"
    
    ans = ''
    while stack:
        ans = str(stack.pop()) + ans
        
    return ans
```
