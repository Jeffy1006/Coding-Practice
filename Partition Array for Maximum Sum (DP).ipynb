```python
def maxSumAfterPartitioning(arr, k):
    def helper(idx):
        if idx >= n: return 0
        if idx in mem: return mem[idx]
        # please use cur, do not change idx
        cur, cur_max, res = idx, 0, -1
        for i in range(1, k + 1):
            # please check boundary
            if cur == n: break
            cur_max = max(cur_max, arr[cur])
            res = max(res, cur_max * i + helper(cur + 1))
            cur += 1
        mem[idx] = res
        return res
    n = len(arr)
    mem = {}
    return helper(0)
```

```python
from functools import lru_cache
def maxSumAfterPartitioning(arr, k):
    @ lru_cache(None)
    def helper(idx):
        if idx >= n: return 0

        cur_max, res = 0, -1
        for i in range(1, k + 1):
            if idx == n: break
            cur_max = max(cur_max, arr[idx])
            res = max(res, cur_max * i + helper(idx + 1))
            idx += 1
        return res
    n = len(arr)
    return helper(0)
```

```python
def maxSumAfterPartitioning(arr, k):
    n = len(arr)
    dp = [0] * (n + 1)
    for i in range(1, n + 1):
        m = float('-inf')
        for j in range(1, min(i, k) + 1):
            m = max(m, arr[i - j])
            dp[i] = max(dp[i], dp[i - j] + m * j)

    return dp[n]
```
