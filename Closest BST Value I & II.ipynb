```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
def cloestBST(root, target):
    
    def dfs(node):
        # declare nonlocal since minvalue and mindiff are not pass by memory
        nonlocal minvalue, mindiff
        if not node: return
        
        tempdiff = abs(target - node.val)
        if tempdiff < mindiff:
            minvalue = node.val
            mindiff = tempdiff

        if node.val > target:
            dfs(node.left)
        else:
            dfs(node.right)
    
    minvalue = None
    mindiff = float("inf")
    dfs(root)
    return minvalue
        
```

```python
def cloestBSTII(root, target, k):
    res = []
    
    def inorder(node):
        if not node: return
        
        inorder(node.left)
        if len(res) == k:
            if abs(target - node.val) < abs(target - res[0]):
                res.pop(0)
            else:
                return
        res.append(node.val)
        inorder(node.right)
    
    inorder(root)
    return res
```

```python
a = Node(4)
b = Node(2)
c = Node(5)
d = Node(1)
e = Node(3)
a.left, a.right = b, c
b.left, b.right = d, e
cloestBSTII(a, 3.214286, 3)
```
