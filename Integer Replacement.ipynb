```python
def integerReplacement(n):
    res = 0
    while n > 1:
        res += 1
        if n & 1:
            # need integer ended with '11' but not '01'
            if n & 2 and n > 3:
                n += 1
                res += 2
                n >>= 2
            else:
                n -= 1
        else:
            n >>= 1
    return res
```

```python
def integerReplacement(n):
    mem = {1:0}
    
    def helper(n):
        if n in mem:
            return mem[n]
        
        if n % 2 == 0:
            mem[n] = 1 + integerReplacement(n / 2)
        else:
            mem[n] = 1 + min(integerReplacement(n - 1), integerReplacement(n + 1))
        return mem[n] 
    
    return helper(n)
```
