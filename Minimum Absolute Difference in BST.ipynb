```python
def getMinimumDifference(root):
    stack = []
    cur = root
    prev = None
    min_diff = float('inf')
    while True:
        if cur:
            stack.append(cur)
            cur = cur.left
        elif stack:
            cur = stack.pop()
            if prev:
                min_diff = min(min_diff, cur.val - prev)
            prev = cur.val
            cur = cur.right
        else: 
            break
    return min_diff
```

```python
def getMinimumDifference(root):
    L = []
    def dfs(node):
        if node.left: dfs(node.left)
        L.append(node)
        if node.right: dfs(node.right)
    dfs(root)
    return min(b - a for a, b in zip(L, L[1:]))
```

```python
def getMinimumDifference(root):
    def fn(node, lo, hi):
        # calculation only happens in leaf node to mimic the difference (gap) between elements in array format
        # min result from each leaf will pass up, and the minimun will reach the root
        if not node: return hi - lo
        left = fn(node.left, lo, node.val)
        right = fn(node.right, node.val, hi)
        return min(left, right)
    return fn(root, float('-inf'), float('inf'))
```
