```python
import string

def decode(res, val, s, start, nlMap):
    
    if len(s) == 0 or (len(s) == 1 and s == '0'):
        return 
    
    if start == len(s):
        res.append(val)
        return
    
    
    for i in range(start,len(s)):
        if s[start] == '0':
            return
        
        val += nlMap[s[start]]
        decode(res, val, s, start+1, nlMap)
        
        if start=1 < len(s) and s[start]+s[start+1] in nlMap:
            val += nlMap[s[start]+s[start+1]]
            decode(res, val, s, start+2, nlMap)
        
        
        
                
    return
    
def decodeDri(s):
    nums = [str(i) for i in range(1,27)]
    letrs = list(string.ascii_letters[26:])
    nlMap = {}
    for i in range(26): nlMap[nums[i]] = letrs[i] 
    res = []
    val = ''
    decode(res, val, s, 0, nlMap)
    
    return res
```

```python
def helper(data, k, memo):
    
    if k == 0:
        return 1
    
    s = len(data) - k
    
    if data[s] == '0':
        return 0
    
    if memo[k] != 0:
        return memo[k]
    
    result = helper(data, k-1, memo)
    if k >= 2 and int(data[s:s+2]) <= 26:
        result += helper(data, k-2, memo)
    memo[k] = result
    #print(str(k), end = ' ')
    #print(memo)
    return result

def num_ways_dp(data):
    memo = [0] * (len(data) + 1)
    return helper(data, len(data), memo)
```
