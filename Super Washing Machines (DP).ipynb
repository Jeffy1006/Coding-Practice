```python
def findMinMoves(machines):
    total = sum(machines)
    n = len(machines)
    if total % n: return -1
    
    target = total // n
    move = [0] * n
    ans = 0
    for i in range(n - 1):
        delta = machines[i] - target
        machines[i + 1] += delta
        
        if delta > 0:
            move[i] += delta
        else: 
            move[i + 1] += -delta
        
        ans = max(ans, move[i])
    ans = max(ans, move[-1])
    return ans
```

```python
findMinMoves([0,6,0,0,6,0])
```
