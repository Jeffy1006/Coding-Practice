```python
def wordPattern(s, pattern):
    s = s.split()
    if len(pattern) != len(s):
        return False
    
    map_p_s, map_s_p = {}, {}
    for i in range(len(s)):
        if pattern[i] not in map_p_s and s[i] not in map_s_p:
            map_p_s[pattern[i]], map_s_p[s[i]] = s[i], pattern[i]
        elif pattern[i] not in map_p_s or s[i] not in map_s_p:
            return False
        elif map_p_s[pattern[i]] != s[i] or map_s_p[s[i]] != pattern[i]:
            return False
    
    return True
```
