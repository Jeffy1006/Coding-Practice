```python
def findLPS(s):
    n = len(s)
    dp = [[0] * n for _ in range(n)]
    for l in range(1, n + 1):
        for i in range(n - l):
            j = i - l + 1
            if i == j:
                dp[i][j] = 1
                continue
            if s[i] == s[j]:
                dp[i][j] = dp[i + 1][j - 1] + 2
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j - 1])
    return dp[0][n - 1]
```

```python
def findLPS(s):
    n = len(s)
    dp0 = [0] * n  # sols of len = l
    dp1 = [0] * n  # sols of len = l - 1
    dp2 = [0] * n  # sols of len = l - 2
    for l in range(1, n + 1):
        for i in range(n - l):
            j = i - l + 1
            if i == j:
                dp0[i] = 1
                continue
            if s[i] == s[j]:
                dp0[i] = dp2[i + 1] + 2
            else:
                dp0[i] = max(dp1[i + 1], dp1[i])
        dp0, dp1 = dp1, dp0
        dp0, dp2 = dp2, dp0
    return dp1[0]
```
