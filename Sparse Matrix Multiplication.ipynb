```python
def multiply(A, B):
    
    list1 = []
    for i in range(len(A)):
        temp = []
        for j in range(len(A[0])):
            if A[i][j] != 0:
                temp.add((j, A[i][j]))
        list1.append(temp)
    
    res = [[0] * len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for a in list1:
            for j in range(len(B[0])):
                col_from_A = a[0]
                value_from_A = a[1]
                res[i][j] += value_from_A * B[col_from_A][j]
    
    return res
```

```python
class SparseMatrix(object):
    def __init__(self, nrow, ncol, S):
        self.nrow = nrow
        self.ncol = ncol
        self.S = S
        
    def to_sparse(M):
        S = dict()
        for r, row in enumerate(M):
            for c, value in enumerate(row):
                if value: S[(r, c)] = value 
        return S
    
    def from_dense(cls, M):
        nrow, ncol = len(M), len(M[0])
        S = cls.to_sparse(M)
        return cls(nrow, ncol, S)
    
    def from_sparse(cls, nrow, ncol, S):
        return cls(nrow, ncol, S)
    
    # overload "@" which is a matrix matrix multiplation in Python, overload to improve speed 
    def __matmul__(self, B):
        C = dict()
        for (a_r, a_c), a_val in self.S.items():
            #for (b_r, b_c), b_val in B.S.items():
            #    if a_c == b_r:
            for b_c in range(B.ncol):
                if (a_c, b_c) in B.S:
                    b_val = B.S[(a_c, b_c)]
                    C[(a_r, b_c)] = C.get((a_r, b_c), 0) + a_val * b_val
        return self.from_sparse(self.nrow, B.ncol, C)
    
    def to_dense(self):
        M = [[0] * self.ncol for _ in range(self.nrow)]
        for (r, c), value in self.S:
            M[r][c] = value
        return M
        
class Solution:
    def multiply(self, A, B):
        A = SparseMatrix.from_dense(A)
        B = SparseMatrix.from_dense(B)
        #C = A.matmul(B)
        C = A @ B
        return C.to_dense()
```
