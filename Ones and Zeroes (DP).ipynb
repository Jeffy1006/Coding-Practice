```python
from collections import Counter
def findMaxForm(strs, m, n):
    mem = {}
    def helper(zero, one, index):
        if index == len(strs) or zero + one == 0: return 0
        if (zero, one, index) in mem: return mem[(zero, one, index)]
        
        count = Counter([strs[index]])
        consider = 0
        if zero >= count['0'] and one >= ['1']:
            consider = 1 + helper(zero - count['0'], one - count['1'], index + 1)
        skip = helper(zero, one, index + 1)
        
        mem[(zero, one, index)] = max(consider, skip)
        return mem[(zero, one, index)] 
    
    return helper(m, n, 0)
```

```python
from collections import Counter
def findMaxForm(strs, m, n):
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for s in strs:
        count = Counter(s)
        for zero in range(m, count['0'] - 1, -1):
            for one in range(n, count['1'] - 1, -1):
                dp[zero][one] = max(dp[zero][one], dp[zero - count('0')][one - count['1']] + 1)
                
    return dp[m][n]
```
