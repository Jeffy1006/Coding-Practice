```python
def getSkyLine(buildings):
    max_co = 0
    for n in buildings:
        max_co = max(max_co, n[1])
    # add 2 to include the last index and one after last index to get the bottom right coordinate
    max_co += 2
    skyline = [0] * (max_co)
    
    for n in buildings:
        for i in range(n[0], n[1] + 1):
            skyline[i] = max(skyline[i], n[2])
    #print(skyline)
    res = []
    change = 0
    for i in range(max_co):
        if skyline[i] > change:
            change = skyline[i]
            res.append([i, skyline[i]])
        elif i + 1 < max_co and skyline[i] > skyline[i + 1]:
            change = skyline[i + 1]
            res.append([i, skyline[i + 1]])
    
    return res
```

```python
from sortContainers import SortedList
def getSkyLine(buildings):
    events = []
    for l, r, h in buildings:
        events.append((l, -h, 0))
        events.append((r, h, 1))
        
    events.sort()
    
    ans = []
    active_heights = SortedList([0])
    n = len(events)
    i = 0
    while i < n:

        x, h, t = events[i]
            
        if t == 0:
            active_heights.add(-h)
        else:
            active_heights.remove(h)
            
        i += 1
        
        if not ans or ans[-1][1] != active_heights[-1]:
            ans.append((x, active_heights[-1]))
    
    return ans
```

```python
buildings = [[2,9,10],[3,7,15],[5,12,12],[15,20,10],[19,24,8]]
x = getSkyLine(buildings)
x
```
