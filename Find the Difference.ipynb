```python
from collections import Counter
def findDifference(s, t):
    return list(Counter(t) - Counter(s))[0]
```

```python
def findDifference(s, t):
    ans = 0
    for c in s:
        ans ^= ord(c)
    for c in t:
        ans ^= ord(c)
    return chr(ans)
```

```python
a = 'abc'
b = 'abcd'
findDifference(a, b)
```
