```python
from collections import deque
def calculateIII(s):
    q = deque([c for c in list(s) if c != " "])
    q.append(" ")
    
    def helper():
        num = prev = curSum = 0
        prevOp = "+"
        
        while queue:
            c = q.popleft()
            if "0" <= c <= "9":
                num = num * 10 + int(c)
            elif c == "(":
                num = helper(q)
            else:
                if prevOP = '+':
                    curSum += prev
                    prev = num
                elif prevOP = '-':
                    curSum += prev
                    prev = -num
                elif prevOP = '*':
                    prev *= num
                elif prevOP = '/':
                    prev /= num
            
                if c == ")": break
                prevOp = c
                num = 0

        return prev + curSum
    
    return helper()
```
