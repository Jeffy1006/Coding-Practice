```python
def largestPalidrome(n):
    if n == 1: return 9
    for z in range(2, 2 * (9 * 10**n) - 1):
        left = 10**n - z
        right = int(str(left)[::-1])
        root1 = root2 = 0
        if z**2 - 4 * right < 0:
            continue
        root1 = 0.5 * (z + (z**2 - 4 * right)**0.5)
        root2 = 0.5 * (z - (z**2 - 4 * right)**0.5)
        if root1 - int(root1) == 0 or root2 - int(root2) == 0:
            return (10**n * left + right)
```

```python
largestPalidrome(2)
```
