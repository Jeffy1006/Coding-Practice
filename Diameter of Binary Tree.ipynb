```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
        
def findDiameter(root):
    diameter = 0
    
    def dfs(node):
        nonlocal diameter
        if not node: return 0
        
        left = dfs(node.left)
        right = dfs(node.right)
        diameter = max(diameter, left + right + 1)
        return left + 1 if left > right else right + 1
    
    dfs(root)
    return diameter - 1
```

```python
class Solution:
    def findDiameter(self, root):
        self.ans = 0
        
        def dfs(p):
            if not p: return 0
            left, right = dfs(p.left), dfs(p.right)
            self.ans = max(self.ans, left + right)
            return 1 + max(left, right)
        
        depth(root)
        return self.ans
```
