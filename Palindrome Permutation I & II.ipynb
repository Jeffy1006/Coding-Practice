```python
def palinPermut(s):
    hashset = set()
    for c in s:
        if c in hashset:
            hashset.remove(c)
        else:
            hashset.add(c)
    
    return True if len(hashset) < 2 else False
```

```python
def palinPermutII(s):
    res = set()
    
    def isPalin(cand):
        start, end = 0, len(cand) - 1
        while start < end:
            if cand[start] != cand[end]:
                return False
            start += 1
            end -= 1
        return True
        
    def dfs(cand, idx):
        if idx == len(cand) and isPalin(cand):
            temp = "".join(cand)
            if temp not in res:
                res.add(temp)
            return
        
        for i in range(idx, len(cand)):
            cand[idx], cand[i] = cand[i], cand[idx]
            dfs(cand, idx + 1)
            cand[idx], cand[i] = cand[i], cand[idx]
            
        return
    
    dfs(list(s),0)
    return list(res)
```

```python
def palinPermutII(s):
    res = []
    hashmap = {}
    mid = ""
    odd = 0
    
    for c in s:
        if c not in hashmap:
            hashmap[c] = 1
        else:
            hashmap[c] += 1
    
    for k in hashmap.keys():
        if hashmap[k] % 2 == 1:
            odd += 1
            mid = k
            
    if odd > 1:
        return res
    if odd == 1:
        del hashmap[mid]
    
    cand = ""
    # hashmap store float type, can only multiply with int, // yield int
    for k in hashmap.keys():
        cand += k * (hashmap[k] // 2)
    hashmap = set()
        
    def dfs(cand, idx):
        if idx == len(cand):
            temp = "".join(cand)
            if temp not in hashmap:
                hashmap.add(temp)
                res.append(temp + mid + temp[::-1])
                return
            
        for i in range(idx, len(cand)):
            cand[idx], cand[i] = cand[i], cand[idx]
            dfs(cand, idx + 1)
            cand[idx], cand[i] = cand[i], cand[idx]
            
        return
    
    dfs(list(cand), 0)
    return list(res)
```
