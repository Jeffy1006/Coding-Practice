```python
def goodNodes(root):
    def helper(node, prev):
        if not node: return 0
        count = 1 if node.val >= prev else 0
        count += helper(node.left, max(prev, node.val))
        count += helper(node.right, max(prev, node.val))
        return count
    return helper(root, float('-inf'))
```

```python
def goodNodes(r, ma=float('-inf')):
    return goodNodes(r.left, max(ma, r.val)) + goodNodes(r.right, max(ma, r.val)) + (r.val >= ma) if r else 0
```

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
d = TreeNode(3)
e = TreeNode(1)
f = TreeNode(5)
c = TreeNode(4, e, f)
b = TreeNode(1, d)
a = TreeNode(3, b, c)
```

```python
goodNodes(a)
```
