```python
def findMaxLength(nums):
    count = 0
    max_len = 0
    table = {0: 0}
    for i, n in enumerate(nums, 1):
        if n == 0:
            count -= 1
        else:
            count += 1
        
        if count in table:
            max_len = max(max_len, i - table[count])
        else:
            table[count] = i
    return max_len        
```
