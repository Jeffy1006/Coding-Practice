```python
def minDistances(word1, word2):
    m, n = len(word1), len(word2)
    if word1 in word2: return n - m
    if word2 in word1: return m - n
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m - 1, -1, -1):
        for j in range(n - 1, -1, -1):
            if word1[i] == word2[j]:
                dp[i][j] = dp[i + 1][j + 1] + 1
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j + 1])
    return m + n - 2 * dp[0][0]
```

```python
def minDistances(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m):
        for j in range(n):
            dp[i + 1][j + 1] = max(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1] + (word1[i] == word2[j]))
    return m + n - 2 * dp[m][n]
```
