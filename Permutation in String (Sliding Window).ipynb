```python
from collections import defaultdict
def checkInclusion(s1, s2):
    m, n = len(s1), len(s2)
    if m > n: return False
    
    count1 = defaultdict(int)
    count2 = defaultdict(int)
    for c in s1: count1[c] += 1
    for c in s2[:m]: count2[c] += 1
    j = 0
    for i in range(m, n):
        if count1 == count2: return True
        count2[s2[j]] -= 1
        if count2[s2[j]] == 0: del count2[s2[j]]
        count2[s2[i]] += 1
        j += 1
    return count1 == count2
```

```python
s1 = 'mn'
s2 = 'eidboaooomn'
checkInclusion(s1, s2)
```
