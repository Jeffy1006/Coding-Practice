```python
import heapq
def maxEvents(events):
    events.sort(reverse = True)
    pq = []
    res = day = 0
    while events or pq:
        if not pq:
            day = events[-1][0]
        while d <= events[-1][0]:
            heapq.heappush(pq, events.pop()[1])
            heapq.heappop(pq)
            res += 1
            d += 1
            while pq and pq[0] < day:
                heapq.heappop()
    return res
```
