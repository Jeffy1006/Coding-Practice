```python
def numSquares(n):
    dp = [n] * (n + 1)
    dp[0] = 0
    
    for target in range(1, n + 1):
        for s in range(1, target + 1):
            square = s * s
            if target - square < 0:
                break
            dp[target] = min(dp[target], dp[target - square] + 1)
    
    return dp[-1]
```
