```python
def outerTrees(trees):
    def cmp(p1, p2, p3):
        x1, y1 = p1
        x2, y2 = p2
        x3, y3 = p3
        
        return (y3 - y2) * (x2 - x1) - (y2 - y1) * (x3 - x2)
    
    points = sorted(trees)
    lower, upper = [], []
    # In my opinion, lower constain upper hull since it remove the point that fit the lower hull, and vice versa
    for point in points:
        while len(lower) >= 2 and cmp(lower[-2], lower[-1], point) > 0:
            lower.pop()
        while len(upper) >= 2 and cmp(upper[-2], upper[-1], point) < 0:
            upper.pop()
        lower.append(tuple(point))
        upper.append(tuple(point))
    print(lower)
    print(upper)
    return list(set(lower + upper))
```

```python
outerTrees([[1,1],[2,2],[2,0],[2,4],[3,3],[4,2]])
```
