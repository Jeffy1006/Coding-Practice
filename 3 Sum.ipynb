```python
def threeSum(nums):
    nums.sort()
    output_arr = []
    
    for i in range(len(nums) - 1):
        if i == 0 or (i > 0 and nums[i] != nums[i-1]):
            low = i + 1
            high = len(nums) - 1
            target = 0 - nums[i]
            
            while low < high:
                if nums[low] + nums[high] == target:
                    output_arr.append([nums[i], nums[low], nums[high]])
                    while low < high and nums[low] == nums[low + 1]: low += 1
                    while low < high and nums[high] == nums[high - 1]: high -= 1
                    low += 1
                    high -= 1
                elif nums[low] + nums[high] > target:
                    high -= 1
                else:
                    low += 1
    
    return output_arr
```

```python
def threeSum(nums):
    nums.sort()
    res = []
    
    for i, a in enumerate(nums):
        if i > 0 and a == nums[i - 1]:
            continue
        
        l, r = i + 1, len(nums) - 1
        while l < r:
            threeSum = a + nums[l] + nums[r]
            if threeSum > 0:
                r -= 1
            elif threeSum < 0:
                l += 1
            else:
                res.append([a, nums[l], nums[r]])
                l += 1
                while nums[l] == nums[l - 1] and l < r:
                    l += 1
    
    return res
```

```python
a = [-1,0,1,2,-1,-4]
threeSum(a)
```

```python
def threeSumSmaller(nums, target):
    count = 0
    nums.sort()
    
    for i in range(len(nums)):
        low = i + 1
        high = len(nums) - 1
            
        while low < high:
            if nums[low] + nums[high] + nums[i] < target:
                count += high - low
                low += 1
            else:
                high -= 1
    
    return count
```
