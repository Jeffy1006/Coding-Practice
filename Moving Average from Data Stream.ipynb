```python
class MovingAverage:
    def __init__(self, size):
        self.queue = []
        self.maxSize = size
        self.sum = 0
    
    def next(val):
        if len(self.queue) == self.maxSize:
            self.sum -= val
        self.queue.append(val)
        self.sum += val
        return self.sum / len(self.queue)
```
