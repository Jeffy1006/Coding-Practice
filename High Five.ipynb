```python
from collections import defaultdict
import heapq
def highFive(items):
    scores = defaultdict(list)
    for i, score in items:
        if len(scores[i]) < 5:
            heapq.heappush(scores[i], score)
        else:
            heapq.heappushpop(scores[i], score)
    res = []
    for i in sorted(scores.keys()):
        res.append([i, sum(scores[i]) // 5])
    return res
```

```python
items = [[1,91],[1,92],[2,93],[2,97],[1,60],[2,77],[1,65],[1,87],[1,100],[2,100],[2,76]]
highFive(items)
```
