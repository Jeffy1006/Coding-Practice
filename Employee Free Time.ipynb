```python
"""
# Definition for an Interval.
class Interval:
    def __init__(self, start: int = None, end: int = None):
        self.start = start
        self.end = end
"""

class Solution:
    def employeeFreeTime(self, schedule: '[[Interval]]') -> '[Interval]':
        hashmap = {}
        for intervals in schedule:
            for interval in intervals:
                hashmap[interval.start] = max(hashmap.get(interval.start, 0), interval.end)
        
        times = sorted([[k, v] for k, v in hashmap.items()])
        res = []
        for i in range(len(times) - 1):
            if times[i][1] < times[i + 1][0]:
                res.append(Interval(times[i][1], times[i + 1][0]))
            else:
                if times[i + 1][1] < times[i][1]:
                    times[i + 1][1] = times[i][1]
        
        return res
```
