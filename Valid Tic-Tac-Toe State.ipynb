```python
class Solution(object):
    def check_win_positions(self, board, player):
        n = len(board)
        
        for i in range(n):
            if board[i][0] == board[i][1] == board[i][2] == player:
                return True
        
        for i in range(n):
            if board[0][i] == board[1][i] == board[2][i] == player:
                return True
        
        if board[0][0] == board[1][1] == board[2][2] == player or \
            board[2][0] == board[1][1] == board[0][2] == player:
            return True
        
        return False
    
    def validTicTacToe(self, board):
        n = len(board)
        x_count = o_count = 0
        
        for i in range(n):
            for j in range(n):
                if board[i][j] == 'X':
                    x_count += 1
                elif board[i][j] == 'O':
                    o_count += 1
        
        if o_count > x_count or x_count - o_count > 1:
            return False
        
        x_win = check_win_positions(board, 'X')
        o_win = check_win_positions(board, 'O')
        
        if x_win and o_win:
            return False
        elif x_win and x_count - o_count != 1:
            return False
        elif o_win and o_count != x_count:
            return False
        
        return True
```
