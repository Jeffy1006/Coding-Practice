```python
def longestIncreasingSubseq(nums):
    LIS = [1] * len(nums)
    
    for i in range(len(nums) - 1, -1, -1):
        for j in range(i + 1, len(nums)):
            if nums[i] < nums[j]:
                LIS[i] = max(LIS[i], 1 + LIS[j]) 
    return max(LIS)
```

```python
def longestIncreasingSubseq(envelopes):
    n = len(envelopes)
    if n == 0: return 0
    envelopes.sort()
    dp = [1] * (n + 1) 
    MAX = 1
    
    for i in range(1, n):
        for j in range(0, i):
            if envelopes[i][0] > envelopes[j][0] and \
                envelopes[i][1] > envelopes[j][1] and \
                1 + dp[j] > dp[i]:
                dp[i] = 1 + dp[j]
            if MAX < dp[i]:
                MAX = dp[i]
    return MAX
```

```python
def largestDivisibleSubset(nums):
    n = len(nums)
    ans = []
    if n == 0: return ans
    
    nums.sort()
    MAX = 1
    dp = [1] * (n + 1)
    for i in range(1, n):
        for j in range(0, i):
            if nums[i] % nums[j] == 0 and 1 + dp[j] > dp[i]:
                dp[i] = 1 + dp[j]
                MAX = max(MAX, dp[i])
    
    prev = -1
    for i in range(n - 1, -1, -1):
        if dp[i] == MAX and (prev % nums[i] == 0 or prev == -1):
            ans.append(nums[i])
            MAX -= 1
            prev = nums[i]
    return ans.reverse()     
```
