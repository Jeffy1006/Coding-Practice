```python
def removeInvalidParentheses(s):
    l = r = 0
    for c in s:
        if c == "(":
            l += 1
        elif c == ')':
            if l == 0:
                r += 1
            else:
                l -= 1
    
    ans = []
    
    def isValid(s):
        count = 0
        for c in s:
            if c == '(': count += 1
            if c == ')': count -= 1
            if count < 0: return False
        
        return count == 0
    
    def dfs(s, start, l, r):
        if l == 0 and r == 0:
            if isValid(s): 
                ans.append(s)
            return
        
        for i in range(start, len(s)):
            # skip duplicates cases
            if i != start and s[i] == s[i - 1]: continue
            # remove ')' first if there is extra ')'
            if s[i] == ")" and r > 0:
                curr = s[:i] + s[i + 1:]
                dfs(curr, i, l, r - 1)
            # remove '(' then only if we've already remove all ')' 
            if  r == 0 and s[i] == "(":
                curr = s[:i] + s[i + 1:]
                dfs(curr, i, l - 1, r)
    
    dfs(s, 0, l, r)
    return ans
```
