```python
class Solution:
    def maxNumber(self, nums1, nums2, k):
        ans = []
        n1, n2 = len(nums1), len(nums2)
        
        def helper(nums, k):
            if k == 0: return []
            ans = []
            to_pop = len(nums) - k
            for num in nums:
                while ans and num > ans[-1] and to_pop > 0:
                    ans.pop()
                ans.append(num)
            return ans[:k]
        
        def merge(nums1, nums2):
            ans = []
            i = j = 0
            while i < len(nums1) and j < len(nums2):
                if nums1[i] > nums2[j]:
                    ans.append(nums1[i])
                    i += 1
                else:
                    ans.append(nums2[j])
                    j += 1
            while i < len(nums1):
                ans.append(nums1[i])
                i += 1
            while j < len(nums2):
                ans.append(nums2[i])
                j += 1
            return ans
                    
        
        for k1 in range(k + 1):
            k2 = k - k1
            if k1 > n1 or k2 > n2: continue
            ans = max(ans, merge(helper(nums1, k1), helper(nums2, k2)))
        return ans
    
```
