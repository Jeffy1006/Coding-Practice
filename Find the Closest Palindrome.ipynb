```python
def nearestPalindromic(n):
    l = len(n)
    candidates = set((str(10**l + 1), str(10 ** (l - 1) - 1)))
    prefix = int(n[:(l + 1) // 2])
    for start in map(str, (prefix - 1, prefix, prefix + 1)):
        candidates.add(start + (start[:-1] if l % 2 else start)[::-1])
        #candidates.add(start + [start, start[:-1]][l & 1][::-1])
    candidates.discard(n)
    return min(candidates, key = lambda x: (abs(int(x) - int(n)), int(x)))
```
