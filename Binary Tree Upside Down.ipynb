```python
def upsideDownBT(root):
    def dfs(cur):
        if not cur.left:
            return cur
        
        newRoot = dfs(cur.left)
        cur.left.left = cur.right
        cur.left.right = cur
        cur.left = None
        cur.right = None
        
        return newRoot
    
    if not root:
        return None
    
    return dfs(root)
```
