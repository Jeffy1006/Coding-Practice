```python
import random
class RandomizedSet:
    def __init(self):
        self.hashmap = {}
        self.v = []
    
    def insert(self, val):
        if self.val in self.hashmap:
            return False
        self.v.append(val)
        self.hashmap[val] = len(self.v) - 1
        return True
    
    def remove(self, val):
        if self.val not in self.hashmap:
            return False
        self.v[self.hashmap[val]] = self.v[-1]
        self.hashmap[self.v[-1]] = self.hashmap[val]
        del self.hashmap[val]
        self.v.pop()
        return True
    
    def getRandom(self):
        return self.v[random.randint(0,(len(self.v) - 1))]
```
