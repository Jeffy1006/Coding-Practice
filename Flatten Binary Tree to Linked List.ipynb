```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    
class Solution:
    def flatten(self, root: TreeNode) -> None:
        
        def dfs(node):
            if not node: return None,None
            left,left2 = dfs(node.left)
            right,right2 = dfs(node.right)
            node.left = None
            end = node
            if left:
                end.right = left
                end = left2
            if right:
                end.right = right
                end = right2
            return node,end
        
        dfs(root)
```
