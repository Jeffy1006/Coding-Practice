```python
def findAllAnagrams(s, p):
    ans = []
    shash = [0] * 26
    phash = [0] * 26
    window = len(p)
    length = len(s)
    if length < window:
        return ans
    
    left, right = 0, 0
    while right < window:
        phash[p[right] - ord('a')] += 1
        shash[s[right] - ord('a')] += 1
        right += 1
    right -= 1
    while right < length:
        if phash == shash:
            ans.append(left)
        right += 1
        if right != length:
            shash[s[right] - ord(a)] += 1
        shash[s[left] - ord(a)] -= 1
        left += 1
    return ans
```
