```python
def singleNumber(nums):
    result = 0
    for i in range(len(nums)):
        result ^= nums[i]
    
    return result
```

```python
def singleNumberII(nums):
    ones = 0
    twos = 0
    
    for n in nums:
        # int in Python must carry sign, 
        # used(& 255) as a bit mask to get rid of the negative bit
        ones = (ones^n) & (~twos & 255)
        twos = (twos^n) & (~ones & 255)
    
    return ones
```

```python
def singleNumberIII(nums):
    xor = 0
    for n in nums:
        xor ^= n
    
    firstbit = xor & (xor - 1) ^ xor
    num1 = 0
    for n in nums:
        if n & firstbit:
            num1 ^= n
    
    return [num1, xor ^ num1]
```

```python
def singleNumberIII(nums):
    xy = 0
    for n in nums:
        xy ^= n
    
    xy ^= -xy
    result = [0, 0]
    for n in nums:
        if n & xy:
            result[1] ^= n
        else:
            result[0] ^= n
    
    return res
```
