```python
def strobogrammaticNumber(num):
    hashmap = {}
    hashmap['6'] = '9'
    hashmap['9'] = '6'
    hashmap['0'] = '0'
    hashmap['1'] = '1'
    hashmap['8'] = '8'
    
    left, right = 0, len(num) - 1
    while left <= right:
        if num[left] not in hashmap:
            return False
        elif hashmap[left] != num[right]:
            return False
        
        left += 1
        right += 1
    
    return True
```

```python
def strobogrammaticNumberII(n):
    hashmap = {}
    hashmap['6'] = '9'
    hashmap['9'] = '6'
    hashmap['0'] = '0'
    hashmap['1'] = '1'
    hashmap['8'] = '8'
    
    res = [""]
    size = 0
    
    if n % 2 == 1:
        res = ["0","1","8"]
        size = 1
        
    while size != n:
        temp = []
        for k in hashmap.keys():
            if size == n - 2 and k == "0": continue 
            for token in res:
                temp.append(k + token + hashmap[k])
        res = temp
        size += 2
    
    return res    
```

```python
def strobogrammaticNumberIIR(n):
    hashmap = {}
    hashmap['6'] = '9'
    hashmap['9'] = '6'
    hashmap['0'] = '0'
    hashmap['1'] = '1'
    hashmap['8'] = '8'
    
    res = []
    
    def helper(sn, remain):
        nonlocal res, hashmap
        if remain == 0:
            res.append(sn)
            return
        
        for k in hashmap.keys():
            if remain == 2 and k == "0": continue 
            helper((k + sn + hashmap[k]), remain - 2)
        
    if n % 2 == 1:
        helper("0", n - 1)
        helper("1", n - 1)
        helper("8", n - 1)
    else:
        helper("", n)

    return res
```

```python
def strobogrammaticNumberIII(low, high):
    
    count = 0
    
    for i in range(len(low),len(high) + 1):
        cand = strobogrammaticNumberII(i)
        for n in cand:
            if int(low) <= int(n) <= int(high):
                count += 1
    
    return count
```
