```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
```

```python
def isSubtree(root, subRoot):
    def helper(node):
        if not node: return 'N'
        return str(node.val) + helper(node.left) + helper(node.right)
    return helper(subRoot) in helper(root)
```

```python
a = Node(3)
b = Node(4)
c = Node(5)
d = Node(1)
e = Node(2)
f = Node(0)
a.left, a.right = b, c
b.left, b.right = d, e
#e.left = f
g = Node(4)
h = Node(1)
i = Node(2)
g.left, g.right = h, i
isSubtree(a, g)
```
