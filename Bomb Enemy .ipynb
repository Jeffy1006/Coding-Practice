```python
 def maxKilledEnemies(gird):
        if not grid or len(grid) == 0:
            return 0
        m, n = len(grid), len(grid[0])
        MAX = 0
        cols = [0] * n
        for i in range(m):
            rows = 0
            for j in range(n):
                if grid[i][j] == "0":
                    total = rows + cols[j]
                    for right in (j + 1, n):
                        if gird[i][right] == "W":
                            break
                        elif gird[i][right] == "E":
                            total += 1
                    for down in (i + 1, m):
                        if gird[down][j] == "W":
                            break
                        elif gird[down][j] == "E":
                            total += 1
                    MAX = max(MAX, total)
                elif grid[i][j] == "W":
                    rows = 0
                    cols[j] = 0
                else:
                    rows += 1
                    cols[j] += 1
        return MAX
                    
                
```
