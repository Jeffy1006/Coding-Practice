```python
from collections import defaultdict
def minAreaRect(points):
    hashmap = defaultdict(set)
    for point in points:
        hashmap[point[0]].add(point[1])
    
    res = float('inf')
    x = list(hashmap.keys())
    for i in range(len(x)):
        if len(hashmap[x[i]]) < 2: continue
        for j in range(i + 1, len(x)):
            if len(hashmap[x[j]]) < 2: continue

            tmp = hashmap[x[i]].intersection(hashmap[x[j]])
            if len(tmp) >= 2:
                tmp = list(sorted(tmp))
                minimum = float('inf')
                for k in range(len(tmp) - 1):
                    minimum = min(minimum, tmp[k + 1] - tmp[k])
            res = min(res, minimum * abs(x[i] - x[j]))
    return res
```

```python
def minAreaRect(points):
    min_area = float('inf')
    points_table = set()
    
    for x, y in points:
        points_table.add((x, y))
    
    for x1, y1 in points:
        for x2, y2 in points:
            if x1 > x2 and y1 > y2:
                if (x1, y2) in points_table and (x2, y1) in points_table:
                    area = abs(x1 - x2) * abs(y1 - y2)
                    if area:
                        min_area = min(area, min_area)
    return 0 if min_area == float('inf') else min_area
```

```python
points = [[1,1],[1,3],[3,1],[3,3],[2,2]]
minAreaRect(points)
```
