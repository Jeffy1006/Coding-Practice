```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
```

```python
def tree2str(root):
    def helper(node):
        if not node: return ''
        res = str(node.val)
        if node.left:
            res += '(' + helper(node.left) + ')'
            return res if not node.right else res + '(' + helper(node.right) + ')'
        else:
            return res if not node.right else res + '()(' + helper(node.right) + ')'
    return helper(root)
```

```python
def tree2str(root):
    if not root: return ''
    left = '({})'.format(tree2str(root.left)) if root.left or root.right else ''
    right = '({})'.format(tree2str(root.right)) if root.right else ''
    return '{}{}{}'.format(root.val, left, right)
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
e = Node(2)
f = Node(0)
a.left, a.right = b, c
b.left = d
tree2str(a)
```
