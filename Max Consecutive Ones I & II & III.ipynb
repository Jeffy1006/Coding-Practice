```python
def longestOnesI(nums):
    ans = 0
    cur = 0
    for j in range(len(nums)):
        if nums[j]:
            cur += 1
            ans = max(ans, cur)
        else:
            cur = 0
    return ans
```

```python
def longestOnesII(nums):
    ans = 0
    k = 1
    i = 0
    for j in range(len(nums)):
        if not nums[j]:
            k -= 1
        while k < 0:
            if not nums[i]:
                k += 1
            i += 1
        ans = max(ans, j - i + 1)
    return ans
```

```python
def longestOnesIII(nums, k):
    ans = 0
    i = 0
    for j in range(len(nums)):
        if not nums[j]:
            k -= 1
        while k < 0:
            if not nums[i]:
                k += 1
            i += 1
        ans = max(ans, j - i + 1)
    return ans
```
