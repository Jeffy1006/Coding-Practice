```python
def complexNumberMultiply(num1, num2):
    i1, i2 = num1.index('+'), num2.index('+') 
    num1_a, num1_b = int(num1[:i1]), int(num1[i1 + 1:-1])
    num2_a, num2_b = int(num2[:i2]), int(num2[i2 + 1:-1])
    return str((num1_a * num2_a) - (num1_b * num2_b)) + '+' + str((num1_a * num2_b) + (num1_b * num2_a)) + 'i' 
```

```python
def complexNumberMultiply(num1, num2):
    a1, a2 = map(int, num1[:-1].split('+'))
    b1, b2 = map(int, num2[:-1].split('+'))
    return '%d+%di' % (a1 * b1 - a2 * b2, a1 * b2 + a2 * b1)
```

```python
complexNumberMultiply('2+5i', '4+-3i')
```
