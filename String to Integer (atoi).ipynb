```python
def myAtoi(s):
    if len(s) == 0: return 0
    s = s.strip()
    sign = -1 if s[0] == '-' else 1
    
    res, i = 0, 0
    if s[i] in ['-', '+']:
        i = 1
    while i < len(s) and s[i].isdigit():
        res = res * 10 + (ord(s[i]) - ord('0'))
        i += 1
    
    return max(-2**-31, min(2**31 - 1, res * sign))
```
