```python
from collections import defaultdict
def subdomainVisits(cpdomains):
    count = defaultdict(int)
    for cd in cpdomains:
        n, s = cd.split
        count[s] += int(n)
        for i in range(len(s)):
            if s[i] == '.':
                count[s[i + 1:]] += int(n)
    return ["%d %s" % (count[k], k) for k in count]
```
