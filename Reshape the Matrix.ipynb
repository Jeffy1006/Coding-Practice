```python
def matrixReshape(mat, r, c):
    m, n = len(mat), len(mat[0])
    if m * n != r * c:
        return mat
    
    ans = [[0] * c for _ in range(r)]
    x = y = 0
    for i in range(m):
        for j in range(n):
            ans[y][x % c] = mat[i][j]
            x += 1
            y = y + 1 if x % c == 0 else y
    return ans
```
