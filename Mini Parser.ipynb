```python
from collections import deque 
def deserialize(s):
    if not s: return s
    if s[0] != "[": return s
    q = deque(s)
    q.popleft()
    
    def helper(q):
        res = []
        while q[0] != "]":
            if q[0] == "[":
                q.popleft()
                res.append(helper(q))
            elif q[0] == ",":
                    q.popleft()
            else:
                tmp = ""
                if q[0] == "-":
                    tmp = "-"
                    q.popleft()
                while q[0] != "," and q[0] != "]":
                    tmp += q[0]
                    q.popleft()
                res.append(int(tmp))
        q.popleft()
        return res
    return helper(q)
```
