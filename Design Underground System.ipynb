```python
from collections import defaultdict
class UndergroundSystem:
    def __init__(self):
        self.ids = {}
        self.pairs = defaultdict(int)
        self.freqs = defaultdict(int)

    def checkIn(self, id: int, stationName: str, t: int) -> None:
        self.ids[id] = (stationName, t)

    def checkOut(self, id: int, stationName: str, t: int) -> None:
        start, startTime = self.ids.pop(id)
        self.pairs[(start, stationName)] += t - startTime
        self.freqs[(start, stationName)] += 1

    def getAverageTime(self, startStation: str, endStation: str) -> float:
        return round(self.pairs[(startStation, endStation)] / self.freqs[(startStation, endStation)], 5)
```
