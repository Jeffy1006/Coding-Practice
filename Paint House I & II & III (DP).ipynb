```python
def paintHouses(costs):
    
    dp = [0] * len(costs[0])
    
    for i in range(len(costs)):
        dp0 = costs[i][0] + min(dp[1], dp[2])
        dp1 = costs[i][1] + min(dp[0], dp[2])
        dp2 = costs[i][2] + min(dp[0], dp[1])         
        dp = [dp0, dp1, dp2]
        
    return min(dp)
```

```python
def paintHousesII(costs):
    m, n = len(costs), len(costs[0])
    dp = [[0] * n for i in range(m)]
    for i in range(n):
        dp[0][i] = costs[0][i]
    
    def helper(i, j):
        nonlocal n
        minValue = float("inf")
        for k in range(n):
            if k == j: continue
            minValue = min(minValue, dp[i][k])
        
        return minValue
    
    for i in range(1, m):
        for j in range(n):
            minValue = helper(i - 1, j)
            dp[i][j] = costs[i][j] + minValue

    return min(dp[-1])
```

```python
def paintHousesII(costs):
    m, n = len(costs), len(costs[0])
    dp1 = costs[0].copy()
    dp2 = [0] * n
    
    def helper(temp, j):
        nonlocal n
        minValue = float("inf")
        for k in range(n):
            if k == j: continue
            minValue = min(minValue, temp[k])
        
        return minValue
    
    count = 1
    for i in range(1, m):
        for j in range(n):
            if i % 2 == 1:
                minValue = helper(dp1, j)
                dp2[j] = costs[i][j] + minValue
            else:
                minValue = helper(dp2, j)
                dp1[j] = costs[i][j] + minValue
            
        count += 1

    return min(dp1) if count % 2 == 1 else min(dp2)
```

```python
costs = [[17,2,17],[16,16,5],[14,3,19]]
paintHousesII(costs)
```

```python
class Solution:
    def minCost(self, houses, cost, m, n, target):
        # maps (i, t, p) -> the minimum cost to paint houses i <= h < m with t neighborhoods, where house i - 1 is color p
        dp = {}
        
        # You can use this functools line instead of dp to make it faster, but I cache 
        # manually because I don't want to abstract the caching away from the reader.
        # @functools.lru_cache(None)
        def dfs(i, t, p):
            key = (i, t, p)
            
            if i == len(houses) or t < 0 or m - i < t:
                # base case - we're trying to color 0 houses. Only i == len(houses) is necessary
                # to check here, but it prunes a bit of the search space to make things faster.
                return 0 if t == 0 and i == len(houses) else float('inf')
            
            if key not in dp:
                if houses[i] == 0:
                    dp[key] = min(dfs(i + 1, t - (nc != p), nc) + cost[i][nc - 1] for nc in range(1, n + 1))
                else:
                    dp[key] = dfs(i + 1, t - (houses[i] != p), houses[i])
                
            return dp[key]
            
        ret = dfs(0, target, -1)
        # if ret is infinity, then we failed every case because there were too many neighborhoods
        # to start. If we could paint over houses that were previously painted, we could remedy that,
        # but the problem doesn't allow that. so, we return -1 in that case.
        return ret if ret < float('inf') else -1
```
