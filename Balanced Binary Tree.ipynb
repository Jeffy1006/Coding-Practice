```python
class Solution:
    def isBalanced(self, n: TreeNode) --> bool:
        
        if balancedHeight(n) > -1: return True
        return False
    
    def balancedHeight(self, n : TreeNode) --> int:
        
        if n == None: return 0
        h1 = balancedHeight(n.left)
        h2 = balancedHeight(n.right)
        
        if h1 == -1 or h2 == -1: return -1
        if (abs(h1-h2) > 1): return -1
        
        return max(h1, h2) + 1
```
