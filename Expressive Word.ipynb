```python
def expressiveWords(s, words):
    def convert(w):
        res = []
        for c in w:
            if res and c == res[-1][0]:
                res[-1][1] += 1
            else:
                res.append([c, 1])
        return res
    s = convert(s)
    n = len(s)
    res = 0
    for w in words:
        w = convert(w)
        if len(w) != n: continue
        flag = True
        for i in range(n):
            # need to check whether this part of w is longer than s since we are only allowed to add but not delete
            if w[i] != s[i] and s[i][1] < max(3, w[i][1]):
                flag = False
                break
        res += flag
    return res            
```

```python
def expressiveWords(s, words):
    def check(S, W):
        i, j, i2, j2, n, m = 0, 0, 0, 0, len(S), len(W)
        while i < n and j < m:
            if S[i] != W[j]: return False
            while i2 < n and S[i2] == S[i]: i2 += 1
            while j2 < m and W[j2] == W[j]: j2 += 1
            if i2 - i != j2 - j and i2 - i < max(3, j2 - j): return False
            i, j = i2, j2
        return i == n and j == m
    
    res = 0
    for w in words:
        res += check(s, w)
    return res        
```
