```python
import time
def removeBoxes(boxes):
    mem = {}
    def helper(b):
        if b == '': return 0
        if b in mem:
            return mem[b]
        res = i = j = 0
        while j < len(b):
            while j < len(b) and b[j] == b[i]: j += 1
            res = max(res, (j - i) * (j - i) + helper(b[:i] + b[j:]))
            i = j
        mem[b] = res
        return res
    return helper(''.join(map(str, boxes)))
```

```python
from functools import lru_cache
def removeBoxes(boxes):
    @lru_cache(None) 
    def dp(l, r, k):
        if l > r: return 0
        while l + 1 <= r and boxes[l] == boxes[l + 1]:
            l += 1
            k += 1
        ans = (k + 1) * (k + 1) + dp(l + 1, r, 0)
        for m in range(l + 1, r + 1):
            if boxes[l] == boxes[m]:
                ans = max(ans, dp(m, r, k + 1) + dp(l + 1, m - 1, 0))
        return ans
    return dp(0, len(boxes) - 1, 0)
```

```python
start_time = time.time()
removeBoxes([1,3,2,2,2,3,4,3,1])
print("--- %s seconds ---" % (time.time() - start_time))
```
