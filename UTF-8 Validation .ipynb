```python
def validUTF8(data):
    left = 0
    for d in data:
        if left == 0:
            if (d >> 3) == 30: left = 3
            elif (d >> 4) == 14: left = 2
            elif (d >> 5) == 6: left = 1
            elif (d >> 7) == 0: left = 0
            else: return 0
        else:
            if (d >> 6) != 2: return False
            left -= 1
    return left == 0
```
