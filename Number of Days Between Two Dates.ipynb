```python
import datetime
def daysBetweenDates(date1, date2):
    d1 = datetime.date(*tuple(int(val) for val in date1.split('-')))
    d2 = datetime.date(*tuple(int(val) for val in date2.split('-')))
    return abs((d1 - d2).days)
```

```python
def daysBetweenDates(date1, date2):
    def helper(date):
        y, m, d = map(int, date.split('-'))
        if m < 3:
            m += 12
            y -= 1
        return 365 * y + y // 4 + y // 400 - y // 100 + d + (153 * m + 8) // 5
    return abs(helper(date1) - helper(date2))
```

```python
daysBetweenDates("2000-01-01", "2001-09-11")
```
