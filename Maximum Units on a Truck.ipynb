```python
def maximumUnits(boxTypes, truckSize):
    boxTypes = sorted(boxTypes, key = lambda x: -x[1])
    res = 0
    for boxes in boxTypes:
        if truckSize >= boxes[0]:
            res += boxes[0] * boxes[1]
            truckSize -= boxes[0]
        else:
            res += truckSize * boxes[1]
            break
    return res
```

```python
boxTypes = [[1,3],[2,2],[3,1]]
truckSize = 4
maximumUnits(boxTypes, truckSize)
```
