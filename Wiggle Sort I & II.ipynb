```python
def wiggleSortI(nums):
    for i in range(0,len(nums),2):
        if i != 0 and nums[i] > nums[i - 1]:
            nums[i - 1], nums[i] = nums[i], nums[i - 1]
        if i != len(nums) - 1 and nums[i] > nums[i + 1]:
            nums[i + 1], nums[i] = nums[i], nums[i + 1]
```

```python
x = [3,5,2,1,6,4,5]
wiggleSortI(x)
x
```

```python
def wiggleSortII(nums):
    nums.sort()
    mid = len(nums) // 2
    res = []
    for i in range(mid):
        res.append(nums[mid - i])
        if len(res) < len(nums):
            res.append(nums[len(nums) - i])
    return res
```
