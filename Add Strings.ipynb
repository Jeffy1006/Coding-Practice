```python
def addStrings(num1, num2):
    ans = ''
    carry = 0
    i, j = len(num1) - 1, len(num2) - 1
    while i >= 0 and j >= 0:
        tmp = (ord(num1[i]) - 48) + (ord(num2[j]) - 48) + carry
        carry, r = divmod(tmp, 10)
        ans = str(r) + ans
        i -= 1
        j -= 1

    while i >= 0 and carry > 0:
        tmp = (ord(num1[i]) - 48) + carry
        carry, r = divmod(tmp, 10)
        ans = str(r) + ans
        i -= 1
    ans = num1[:i + 1] + ans
    while j >= 0 and carry > 0:
        tmp = (ord(num2[j]) - 48) + carry
        carry, r = divmod(tmp, 10)
        ans = str(r) + ans
        j -= 1
    ans = num2[:j + 1] + ans
    
    return '1' + ans if carry else ans
```

```python
def addStrings(num1, num2):
    ans = ''
    carry = 0
    i, j = len(num1) - 1, len(num2) - 1
    while i >= 0 or j >= 0:
        tmp = carry
        if i >= 0:
            tmp += (ord(num1[i]) - 48)
            i -= 1
        if j >= 0:
            tmp += (ord(num2[j]) - 48)
            j -= 1
        carry, r = divmod(tmp, 10)
        ans = str(r) + ans
    return '1' + ans if carry else ans
```
