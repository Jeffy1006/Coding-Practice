```python
class Nodes():
    def __init__(self, data):
        self.val = data
        self.next = None
    
def insertionSort(head):
    if not head or not head.next:
        return head
    
    cur, tail = head.next, head
    while cur:
        if cur.val < head.val:
            tail.next = cur.next
            cur.next = head
            head = cur
        elif cur.val > tail.val:
            tail = tail.next
        else:
            idx = head
            while idx != cur:
                if idx.next.val > cur.val:
                    tail.next = cur.next
                    cur.next = idx.next
                    idx.next = cur
                idx = idx.next
        cur = tail.next
    
    return head
```

```python
a = Nodes(3)
b = Nodes(6)
c = Nodes(8)
d = Nodes(7)
e = Nodes(2)
f = Nodes(4)
g = Nodes(1)
h = Nodes(5)
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
f.next = g
g.next = h
s = insertionSort(a)
```
