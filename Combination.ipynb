```python
def combination(n, k):
    res = []
    
    def backTrack(start, comb):
        if len(comb) == k:
            res.append(comb[:])
            return
        
        for i in range(start, n + 1):
            comb.append(i)
            backTrack(i + 1, comb)
            comb.pop()
    
    backTrack(1, [])
    return res
```
