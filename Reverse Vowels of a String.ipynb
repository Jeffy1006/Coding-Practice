```python
def reverseVowels(s):
    vowelSet = set(['a','e','i','o','u','A','E','I','O','U'])
    s = list(s)
    left, right = 0, len(s) - 1
    while left < right:
        while left < right and s[left] not in vowelSet:
            left += 1
        while right > left and s[right] not in vowelSet:
            right -= 1
        if left <= right:
            s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
    return ''.join(s)
```

```python
reverseVowels('leetcode')
```
