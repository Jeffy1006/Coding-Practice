```python
def arithmeticSlices(nums):
    n = len(nums)
    if n <= 3: return 0
    dp = {}
    for i in range(1, n):
        dp[(i - 1, i)] = nums[i] - nums[i - 1]
    ans = 0
    for i in range(3, n + 1):
        for j in range(n - i + 1):
            if (j, j + i - 2) in dp and dp[(j, j + i - 2)] == nums[j + i - 1] - nums[j + i - 2]:
                ans += 1
                dp[(j, j + i - 1)] = dp[(j, j + i - 2)]
    return ans
```
