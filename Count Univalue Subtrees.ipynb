```python
def countUnivalSubtrees(root):
    res = 0
    
    def helper(root):
        nonlocal res
        if not root: return True
        
        left = helper(root.left)
        right = helper(root.right)
        
        if left and right:
            if root.left and root.left.val != root.val:
                return False
            if root.right and root.right.val != root.val:
                return False
            
            res += 1
            return True
        
        return False
    
    helper(root)
    return res
```
