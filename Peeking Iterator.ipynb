```python
class PeekingIterator:
    def __init__(self, nums):
        self.nums = nums
        self.index = 0
        
    def next(self):
        if hasNext:
            self.index += 1
            return self.index - 1
        else:
            return None
    
    def hasNext(self):
        if self.index < len(self.nums):
            return True
        else:
            return False
    
    def peek(self):
        return self.nums[self.index]
```
