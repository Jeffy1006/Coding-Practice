```python
def shortestDistance(words, word1, word2):
    p1 = p2 = -1
    
    minValue = len(words)
    
    for i in range(len(words)):
        if words[i] == word1:
            p1 = i
        elif words[i] == word2:
            p2 = i
        
        if p1 != -1 and p2 != -1:
            minValue = min(minValue, abs(p1 - p2))
    
    return miValue
```

```python
class ShortestWordDistanceII:
    def __init__(self, words):
        self.hash = {}
        for i in range(len(words)):
            if words[i] not in self.hash:
                self.hash[words[i]] = [i]
            else:
                self.hash[words[i]].append(i)
    
    def shortestDistance(self, word1, word2):
        l1 = self.hash[word1]
        l2 = self.hash[word2]
        
        p1 = p2 = 0
        minValue = float("inf")
        
        while p1 < len(l1) and p2 < len(l2):
            minValue = min(minValue, abs(l1[p1] - l2[p2]))
            if l1[p1] < l2[p2]:
                p1 += 1
            else:
                p2 += 1
            
        return minValue
```
