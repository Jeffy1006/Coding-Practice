```python
def predictTheWinner(nums):
    def helper(i, j, score1, score2):
        if i > j:
            return score1 >= score2
        return not helper(i + 1, j, score2, score1 + nums[i]) or not helper(i, j - 1, score2, score1 + nums[j])
    return helper(0, len(nums) - 1, 0, 0)
```

```python
def predictTheWinner(nums):
    dp = {}
    def find(i, j):
        if (i, j) not in dp:
            if i == j:
                return nums[i]
            dp[i, j] = max(nums[i] - find(i + 1, j), nums[j] - find(i, j - 1))
        return dp[i, j]
    return find(0, len(nums) - 1) >= 0
```
