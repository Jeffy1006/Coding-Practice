```python
def findContestMatch(n):
    res = [str(i) for i in range(1, n + 1)]
    while len(res) != 1:
        n, tmp = len(res), []
        for i in range(len(res) // 2):
            tmp.append('(' + res[i] + ',' + res[n - i - 1] + ')')
        res = tmp
    return res[0]
```

```python
findContestMatch(8)
```
