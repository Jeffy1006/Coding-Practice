```python
def findMode(root):
    mode = []
    count = 0
    def helper(node):
        if not node: return 0
        
        res = 1
        left = helper(node.left)
        right = helper(node.right)
        if left and node.val == node.left.val:
            res += left
        if right and node.val == node.right.val:
            res += right
        
        if res > count:
            count = res
            mode = [node.val]
        elif res == count:
            mode.append(node.val)
        
        return res
    
    helper(root)
    return mode
```
