```python
# TC: O(n^2)
# SC: (O(n))
from itertools import accumulate
class Solution:
    def splitArray(self, nums):
        n = len(nums)
        accum = list(accumulate(nums))
        
        for j in range(3, n -3): # exploring mid point
            targets = set()
            for i in range(1, j - 1): # grouping
                sum_l1 = accum[i - 1]
                sum_l2 = accum[j - 1] - accum[i]
                if sum_l1 == sum_l2:
                    targets.add(sum_l1)                   
            for k in range(j + 2, n - 1): # checking
                sum_r1 = accum[k - 1] - accum[j]
                sum_r2 = accum[n - 1] - accum[k]
                if sum_r1 == sum_r2 and sum_r1 in targets:
                    return True
        return False
```
