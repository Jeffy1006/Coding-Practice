```python
def maxSubArrayLen(nums, k):
    hashmap = {}
    hashmap[0] = -1 
    
    maxLen = curSum = 0
    for i in range(len(nums)):
        curSum += nums[i]
        if curSum - k in hashmap:
            maxLen = max(maxLen, i - hashmap[curSum - k])
        if curSum not in hashmap:
            hashmap[curSum] = i
    return maxLen
```

```python
def checkSubarraySum(nums, k):
    if k == 0:
        return any(nums[i] == 0 and nums[i + 1] == 0 for i in range(len(nums) - 1))
    mods, cum_sum_mod_k = {0: -1}, 0
    for i, n in enumerate(nums):
        cum_sum_mod_k = (cum_sum_mod_k + n) % k
        if cum_sum_mod_k in mods and i - mods[cum_sum_mod_k] > 1:
            return True
        if cum_sum_mod_k not in mods:
            mods[cum_sum_mod_k] = i
    return False
```

```python
def subArraySum(nums, k):
    hashmap = {0: 1}
    
    count = curSum = 0
    for i in range(len(nums)):
        curSum += nums[i]
        count += hashmap.get(curSum - k, 0)
        hashmap[curSum] = hashmap.get(curSum, 0) + 1 
    return count
```
