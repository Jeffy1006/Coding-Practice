```python
def mergeSortedList(arr1, arr2, m, n):
    
    if m == 0:
        for i in range(n):
           arr1[i] = arr2[i]
        
        return
    #if n == 0:
    #    return m
    
    i, j, count = m-1, n-1, m+n-1
    while i >= 0 and j >= 0:
        if arr1[i] > arr2[j]:
            arr1[count] = arr1[i]
            i-=1
        else:
            arr1[count] = arr2[j]
            j-=1
        count-=1
    
    if i >= 0:
        while i >= 0:
            arr1[count] = arr1[i]
            i-=1
            count-=1
    else:
        while j >= 0:
            arr1[count] = arr2[j]
            j-=1
            count-=1
    
```

```python
nums1 = [1,2,3,0,0,0]
nums2 = [2,5,6]
mergeSortedList(nums1,nums2,3,3)
print(nums1)
```

```python
nums1 = [0]
nums2 = [1]
mergeSortedList(nums1,nums2,0,1)
print(nums1)
```
