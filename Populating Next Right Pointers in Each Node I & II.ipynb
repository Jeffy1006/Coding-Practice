```python
class Node:
    def __init__(self, data):
        self.val = data
        self.left = None
        self.right = None
        self.next = None
```

```python
class Solution:
    def connect(root: Node) -> Node:
        leftNode = root
        while leftNode and leftNode.left:
            populateLowerLevel(leftNode)
            leftNode = leftNode.left
        
        return root
    
    def populateLowerLevel(startNode: Node):
        cur = startNode
        while cur:
            cur.left.next = cur.right
            if cur.next:
                cur.right.next = cur.next.left
            cur = cur.next
```

```python
class Solution:
    def connect(self, root):
        
        if not root: return None
        
        leftmostNode = root
        while leftmostNode.left:
            curr = leftmostNode
            while curr:
                curr.left.next = curr.right
                if curr.next:
                    curr.right.next = curr.next.left
                curr = curr.next
            
            leftmostNode = leftmostNode.left
        
        return root
```

```python
class Solution:
    def connect(self, root):
        if not root or not root.left: return
        root.left.next = root.right
        connect(root.left)
        if root.next:
            root.right.next = root.next.left
        connect(root.right)
```

```python
class Solution:
    def connect(self, root):
        if not root: return None
        
        queue = []
        queue.append(root)
        
        while queue:
            curr = None
            nxt = None
            
            for _ in range(len(queue)):
                if not curr:
                    curr = queue.pop(0)
                    if curr.left:
                        queue.append(curr.left)
                    if curr.right:
                        curr.append(curr.right)
                elif not nxt:
                    nxt = queue.pop(0)
                    curr.next = nxt
                    if nxt.left:
                        queue.append(nxt.left)
                    if nxt.right:
                        queue.append(nxt.right)     
                else:
                    curr = queue.pop(0)
                    nxt.next = cur
                    if curr.left:
                        queue.append(curr.left)
                    if curr.right:
                        curr.append(curr.right)
                    nxt = curr
                    
        
        return root
```
