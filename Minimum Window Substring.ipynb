```python
from collections import Counter
def minWindow(s, t):
    if t == "": return ""
    
    counter = Counter(t)
    count = len(t)
    left = 0
    resLeft, resRight = -1, len(s)
    minLen = float('inf')
    
    for i in range(len(s)):
        if s[i] in counter:
            counter[s[i]] -= 1
            if counter[s[i]] >= 0:
                count -= 1
                
            if count == 0:
                while count == 0:
                    if s[left] in counter:
                        counter[s[left]] += 1
                        if counter[s[left]] > 0:
                            count += 1
                    left += 1
                    
                if i - left + 2 < minLen:
                    minLen = i - left + 2
                    resLeft = left - 1
                    resRight = i
    
    return s[resLeft:resRight + 1] if minLen < float('inf') else ""
```

```python
def minWindow(s, t):
    if t == "": return ""
    
    countT, window = {}, {}
    
    for c in t:
        countT[c] = countT.get(c, 0) + 1
    
    have, need = 0, len(countT)
    res, resLen = [-1, -1], float('inf')
    l = 0
    
    for r in range(len(s)):
        c = s[r]
        window[c] = window.get(c, 0) + 1
        
        if c in countT and window[c] == countT[c]:
            have += 1
            
        while have == need:
            if r - l + 1 < resLen:
                res = [l, r]
                resLen = r - l + 1
            
            window[s[l]] -= 1
            if s[l] in countT and window[s[l]] < countT[s[l]]:
                have -= 1
            l += 1
    
    return s[res[0]:res[1] + 1] if resLen != float('inf') else ""
```
