```python
from collections import Counter
from collections import defaultdict
def minDeletions(s):
    count = Counter(s)
    freq_dict, max_freq = defaultdict(int), 0
    for k in count:
        freq_dict[count[k]] += 1
        max_freq = max(max_freq, count[k])
    
    res = 0
    for i in range(max_freq, 0, -1):
        if freq_dict[i] <= 1: continue
        else:
            extras = freq_dict[i] - 1
            res += extras
            freq_dict[i - 1] += extras
    
    return res
```

```python
class Solution:
    def minDeletions(self, s: str) -> int:
        cnt, res, used = collections.Counter(s), 0, set()
        for ch, freq in cnt.items():
            while freq > 0 and freq in used:
                freq -= 1
                res += 1
            used.add(freq)
        return res
```
