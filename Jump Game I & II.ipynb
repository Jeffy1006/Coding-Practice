```python
def canJump(nums):
    goal = len(nums) - 1
    
    for i in range(len(nums) - 1, -1, -1):
        if i + nums[i] >= goal:
            goal = i
    
    return True if goal == 0 else False
```

```python
def canJumpII(nums):
    res = 0
    start, end = 0, 0
    
    
    while end < len(nums) - 1:
        farthest = 0
        for i in range(start, end + 1):
            farthest = max(farthest, nums[i] + i)
        start = end + 1
        end = farthest
        res += 1
    
    return res
```
