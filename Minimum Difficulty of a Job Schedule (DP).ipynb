```python
from functools import lru_cache
def minDifficulty(jobDifficulty, d):
    @ lru_cache(None)
    def helper(idx, d):
        if idx >= n and d == 0: return 0
        if d < 0 or n - idx < d: return float('inf')

        maximum, res = float('-inf'), float('inf')
        for i in range(idx, n):
            maximum = max(maximum, jobDifficulty[i])
            res = min(res, maximum + helper(i + 1, d - 1))
        return res
        
    n = len(jobDifficulty)
    ans = helper(0, d)
    return ans if ans < float('inf') else -1
```

```python
def minDifficulty(jobDifficulty, d):
    n = len(jobDifficulty)
    if d > n: return -1
    dp = [[float('inf')] * (d + 1) for _ in range(n + 1)]
    
    dp[0][0] = 0
    for i in range(1, n + 1):
        for k in range(1, d + 1):
            md = 0
            for j in range(i - 1, k - 2, -1):
                md = max(md, jobDifficulty[j])
                dp[i][k] = min(dp[i][k], dp[j][k - 1] + md)
    return dp[n][d]
```
