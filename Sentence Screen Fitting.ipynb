```python
def wordsTyping(sentence, rows, cols):
    s = ' '.join(sentence) + ' '
    n = len(s)
    total_len = 0
    
    for _ in range(rows):
        total_len += cols
        if s[total_len % n] == ' ':
            total_len += 1
        else:
            while s[(total_len - 1) % n] != ' ' and total_len > 0:
                total_len -= 1
    return total_len // n
```
