```python
def isOneEditDistance(s, t):
    if (abs(len(s) - len(t)) > 1) or s == t:
        return False
    
    if len(s) < len(t):
        s, t = t, s
    
    for i in range(len(t)):
        if s[i] != t[i]:
            if len(s) == len(t):
                return s[i+1:] == t[i+1:]
            else:
                return s[i+1:] == t[i:]
    
    return True
```
