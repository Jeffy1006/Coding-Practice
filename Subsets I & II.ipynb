```python
def subsets(nums):
    res = []
    
    def helper(cur, i):
        if i == len(nums):
            res.append(cur[:])
            return
        
        cur.append(nums[i])
        helper(cur, i + 1)
        cur.pop()
        helper(cur, i + 1)
    
    helper([], 0)
    return res
```

```python
def subsets(nums):
    res = []
    cur = []
    
    def helper(i):
        if i >= len(nums):
            res.append(cur[:])
            return
        
        cur.append(nums[i])
        helper(i + 1)
        cur.pop()
        helper(i + 1)
    
    helper(0)
    return res
```

```python
def subsetsII(nums):
    res = []
    
    def backTrack(i, subset):
        if i == len(nums):
            res.append(subset[:])
            return
        
        subset.append(nums[i])
        backTrack(i + 1, subset)
        subset.pop()
        while i + 1 < len(nums) and nums[i] == nums[i + 1]:
            i += 1
        backTrack(i + 1, subset)
        
    backTrack(0, [])
    return res
```
