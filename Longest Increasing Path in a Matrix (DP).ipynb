```python
def longestIncreasingPath(matrix):
    if not matrix: return 0
    m, n = len(matrix), len(matrix[0])
    dp = [[-1] * n for _ in range(m)]
    
    def dfs(x, y):
        if dp[y][x] != -1: return dp[y][x]
        dp [y][x] = 1
        dirs = [0, -1, 0, 1, 0]
        for i in range(len(dirs)):
            tx = x + dirs[i]
            ty = y + dirs[i + 1]
            if tx < 0 or ty < 0 or tx >= n or ty >= m or matrix[ty][tx] <= matrix[y][x]: continue
            dp[y][x] = max(dp[y][x], 1 + dfs(tx, ty))
        return dp[y][x]
        
    ans = 0
    for y in range(m):
        for x in range(n):
            ans = max(ans, dfs(x, y))
    return ans
```

```python
def longestIncreasingPath(matrix):
    dp = {}
    ROWS, COLS = len(matrix), len(matrix[0])
    
    def dfs(r, c, prev):
        if (r, c) in dp: return dp[(r, c)]
        if (r < 0 or c < 0 or 
            r == ROWS or c == COLS or
            prev >= matrix[r][c]):
            return 0
        
        res = 1
        for i, j in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            res = max(res, 1 + dfs(r + i, c + j, matrix[r][c]))
        
        dp[(r, c)] = res
        return res
    
    res = 0
    for r in range(ROWS):
        for c in range(COLS):
            res = max(res, dfs(r, c, -1))
    
    return res
```
