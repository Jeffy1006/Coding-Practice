```python
def reversePolishNotation(tokens):
    nums = []
    for n in tokens:
        if n in "+-*/":
            num2 = nums.pop()
            num1 = nums.pop()
            if n == "+":
                nums.append(num1 + num2)
            elif n == "-":
                nums.append(num1 - num2)
            elif n == "*":
                nums.append(num1 * num2)
            else:
                nums.append(int(num1 / num2))
        else:
            nums.append(int(n))

    return str(nums.pop())
```

```python
a = ["2","1","+","3","*"]
s = reversePolishNotation(a)
s
```

```python
a = ["4","13","5","/","+"]
s = reversePolishNotation(a)
s
```

```python
a = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
s = reversePolishNotation(a)
s
```
