```python
def verifyPreOrderBST(preorder):
    if len(preorder) == 0:
        return True
    
    stack = []
    minValue = float("-inf")
    
    for num in preorder:
        if num < minValue:
            return False
        
        while stack and stack[-1] < num:
            minValue = stack.pop()
        
        stack.append(num)
    
    return True
```

```python
verifyPreOrderBST([5,2,1,3,6])
```

```python
verifyPreOrderBST([5,2,1,6,3])
```

```python
def verifyPreOrderSerializationBT(preorder):
    preorder = preorder.split(',')
    n = 1
    for num in preorder:
        n -= 1
        if n < 0:
            return False
        if num != "#":
            n += 2
    return True if n == 0 else False
```
