```python
from collections import defaultdict
def isAnagram(s, t):
    dic = defaultdict()
    
    for c in s:
        dic[s] += 1
    
    for c in t:
        if c not in dic:
            return False
        else:
            dic[t] -= 1
    
    return sum(dic.values()) == 0
```

```python
from collections import Counter
def isAnagram(s, t):
    return Counter(s) == Counter(t)
```
