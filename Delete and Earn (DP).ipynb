```python
from collections import Counter
def deleteAndEarn(nums):
    count = Counter(nums)
    keys = sorted(count.keys())
    dp0, dp1 = 0, keys[0] * count[keys[0]]
    for i in range(1, len(keys)):
        c, dp2 = keys[i] * count[keys[i]], 0
        if keys[i] - 1 == keys[i - 1]:
            dp2 = max(dp1, dp0 + c)
        else:
            dp2 = dp1 + c
        dp0, dp1 = dp1, dp2
    return dp1
```

```python
from collections import Counter
def deleteAndEarn(nums):
    count = Counter(nums)
    nums = sorted(count.keys())
    p1, p2 = 0, 0
    
    for i in range(len(nums)):
        cur = count[nums[i]] * nums[i]
        if i != 0 and nums[i] - 1 == nums[i - 1]:
            cur = max(p2, p1 + cur)
        else:
            cur += p2
        p1 = p2
        p2 = cur
    
    return p2
```
