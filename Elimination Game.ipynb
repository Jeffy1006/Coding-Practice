```python
def lastRemaining(n):
    return 1 if n == 1 else 2 * (n // 2 + 1 - lastRemaining(n // 2))
```

```python
def lastRemaining(n):
    head = 1
    remain = n
    left = True
    step = 1
    
    while remain > 1:
        if left or remain % 2 == 1:
            head += step
        remain /= 2
        step *= 2
        left = not left
    return head
```
