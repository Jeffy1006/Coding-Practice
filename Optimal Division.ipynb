```python
def optimalDivision(nums):
    ans, n = '', len(nums)
    if not nums: return ans
    ans = str(nums[0])
    if n == 1: return ans
    if n == 2: return ans + '/' + str(nums[1])
    ans += '/(' + str(nums[1])
    for i in range(2, n):
        ans += '/' + str(nums[i])
    return ans + ')'
```

```python
optimalDivision([1000, 100, 10, 2])
```
