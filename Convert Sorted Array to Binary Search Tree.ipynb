```python
def convert(nums):
    if not nums: return None
    
    def helper(low, high):
        if low > high: return None
        
        mid = (low + high) // 2
        root = TreeNode(nums[mid])
        root.left = (low, mid - 1)
        root.right = (mid + 1, high)
        return root
    
    return helper(0, len(nums) - 1)
```
