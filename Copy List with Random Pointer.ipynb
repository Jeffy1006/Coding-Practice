```python
class Node:
    def __init__(self, data):
        self.val = data
        self.next = None
        self.random = None
        
def copyRandomList(head):
    oldToCopy = {None : None}
    
    cur = head
    while cur:
        copy = Node(cur.val)
        oldToCopy[cur] = copy
        cur = cur.next
    
    cur = head
    while cur:
        # remember to set the copy pointer
        copy = oldToCopy[cur]
        # if cur.next: is not necessary after inserting None : None in the dict
        copy.next = oldToCopy[cur.next]
        # if cur.random: is not necessary after inserting None : None in the dict
        copy.random = oldToCopy[cur.random]
        # Please don't forget to update the pointer
        cur = cur.next
    
    return oldToCopy[head]
```
