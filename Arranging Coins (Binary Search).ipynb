```python
def arrangeCoins(n):
    start, end = 1, n
    while start <= end:
        mid = (start + end) // 2
        total = (mid * (mid + 1)) // 2
        if total == n:
            return mid
        elif total < n:
            start = mid + 1
        else:
            end = mid - 1
    return end  
```
