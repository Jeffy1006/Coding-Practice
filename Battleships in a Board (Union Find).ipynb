```python
def countBattleships(board):
    parents = {}
    ranks = {}
    m, n = len(board), len(board[0])
    size = 0
    
    def find(key):
        if parents[key] != key:
            parents[key] = find(parents[key])
        return parents[key]
    
    def union(k1, k2):
        nonlocal size
        p1, p2 = find(k1), find(k2)
        if p1 == p2: return
        if ranks[p1] > ranks[p2]:
            parents[p2] = p1
            ranks[p1] += ranks[p2]
        else:
            parents[p1] = p2
            ranks[p2] += ranks[p1]
        size -= 1
    
    
    for i in range(m):
        for j in range(n):
            if board[i][j] == 'X':
                parents[(i, j)] = (i, j)
                ranks[(i, j)] = 1
                size += 1
                if (i - 1, j) in parents:
                    union((i, j), (i - 1, j))
                if (i, j - 1) in parents:
                    union((i, j), (i, j - 1)) 
    
    return size
```

```python
board = [['X','.','.','X'],
        ['.','.','.','X'],
        ['X','X','.','X']]
countBattleships(board)
```
