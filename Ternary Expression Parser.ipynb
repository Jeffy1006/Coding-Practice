```python
def parseTernary(expression):
    
    def evaluate(s):
        if s[0] == 'T':
            return s[2]
        else:
            return s[4]
    
    stack = []
    curStr = ''
    for i in range(len(expression)):
        if expression[i + 1] == '?':
            stack.append(curStr)
            curStr = expression[i]
        else:
            curStr += expression[i]
            while len(curStr) == 5:
                curStr = evaluate(curStr)
                if stack:
                    curStr = stack[-1] + curStr
                    stack.pop()
    return curStr
```
