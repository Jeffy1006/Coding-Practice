```python
def minArea(image, x, y):
    if len(image) == 0 or len(image[0]) == 0: return 0
    minX = maxX = x
    minY = maxY = y
    
    def dfs(i, j):
        nonlocal minX, minY, maxX, maxY
        if 0 <= i < len(image) and 0 <= j < len(image[0]) and image[i][j] == "1":
            image[i] = image[i][:j] + "0" + image[i][j + 1:]
            minX = min(minX, i)
            minY = min(minY, j)
            maxX = max(maxX, i)
            maxY = max(maxY, j)
            dfs(i + 1, j)
            dfs(i, j + 1)
            dfs(i - 1, j)
            dfs(i, j - 1)
            
    dfs(x, y)
    #print("{} {} {} {}".format(minX, maxX, minY, maxY))   
    return ((maxX - minX) + 1) * ((maxY - minY) + 1)
```

```python
image = ["0010","0110","0111"]
x, y = 0, 2
minArea(image, x, y)
```
