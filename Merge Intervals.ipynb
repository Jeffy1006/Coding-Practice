```python
def merge(intervals):
    if not intervals: return None
    
    intervals.sort()
    res = []
    begin, end = intervals[0][0], intervals[0][1]
    for i in range(1, len(intervals)):
        if end >= intervals[i][0]:
            if end < intervals[i][1]:
                end = intervals[i][1]
        else:
            res.append([begin, end])
            begin = intervals[i][0]
            end = intervals[i][1]
    res.append([begin, end])
    return res
```

```python
def merge(intervals):
    if not intervals: return None
    
    intervals.sort()
    merged = []
    for interval in intervals:
        if not merged or merged[-1][1] < intervals[0]:
            merged.append(interval)
        else:
            merged[-1][1] = max(merge[-1][1], interval[1])
    return merged
```

```python
intervals = [[1,3],[2,6],[8,10],[15,18]]
merge(intervals)
```
