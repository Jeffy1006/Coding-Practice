```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
class Solution:
    #DFS
    def minDepthR(self, root: TreeNode) -> int:
    
        if not root: return 0
        # without self, delcare nonlocal inside dfs()
        self.small = float('inf')
    
        def dfs(node,num):
            if not node: return
            if not node.left and not node.right:
                self.small = min(self.small,num)
            
            dfs(node.left,num+1)
            dfs(node.right,num+1)
        
        dfs(root,1)
    
        return self.small
    
    #BFS
    def minDepthNR(self, root: TreeNode) -> int:
    
        if not root: return 0
    
        q = [(root,1)]
        
        while q:
            cand,num = q.pop(0)
            if not cand.left and not cand.right: return num
            if cand.left: q.append((cand.left,num+1))
            if cand.right: q.append((cand.right,num+1))
    
        return -1
    
    def maxDepthR(self, root: TreeNode) -> int:
    
        if not root: return 0
    
        self.large = float('-inf')
    
        def dfs(node,num):
            if not node: return
            if not node.left and not node.right:
                self.large = max(self.large,num)
            
            dfs(node.left,num+1)
            dfs(node.right,num+1)
        
        dfs(root,1)
    
        return self.large
    
    def maxDepthNR(self, root: TreeNode) -> int:
    
        if not root: return 0
    
        q = [(root,1)]
        depth = -1
    
        while q:
            cand,num = q.pop(0)
            depth = num
            if cand.left: q.append((cand.left,num+1))
            if cand.right: q.append((cand.right,num+1))
    
        return depth
```
