```python
def inorderSuccessor(root, p):
    
    def findLeftmostChild(node):
        while node.left:
            node = node.left      
        return node
    
    # check leftmost right child first because it must be the successor if it's presented
    if p.right:
        return findLeftmostChild(p.right)
    
    
    successor = None
    while root:
        if p.val < root.val:
            successor = root
            root = root.left
        elif p.val > root.val:
            root = root.right
        else:
            break
    
    return successor
```
