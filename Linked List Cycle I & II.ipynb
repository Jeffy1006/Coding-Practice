```python
def hasCycle(head):
    
    seen = set()
    cur = head
    while cur:
        if cur in seen:
            return True
        seen.add(cur)
        cur = cur.next
    
    return False
```

```python
# Space O(1)
def hasCycle(head):
    
    slow = fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        # must be at the bottom because slow and fast are in the same positions in the beginning
        if slow == fast: return True
    
    return False        
```

```python
def hasCycleII(head):
    
    seen = set()
    cur = head
    while cur:
        if cur in seen:
            return cur
        seen.add(cur)
        cur = cur.next
    
    return None   
```

```python
# Space O(1)
def hasCycleII(head):
    
    slow = fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        # must be at the bottom because slow and fast are in the same positions in the beginning
        if slow == fast:
            slow = head
            while slow != fast:
                slow = slow.next
                fast = fast.next
            return slow
    
    return None
```
