```python
from functools import lru_cache
def mincostTickets(days, costs):
    N = len(days)
    durations = [1, 7, 30]

    @lru_cache(None)
    def dp(i): # How much money to do days[i]+
        if i >= N: return 0

        ans = float('inf')
        j = i
        for c, d in zip(costs, durations):
            while j < N and days[j] < days[i] + d:
                j += 1
            ans = min(ans, dp(j) + c)

        return ans

    return dp(0)
```
