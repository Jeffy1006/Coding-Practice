```python
from heapq import *
from collections import defaultdict
class TrieNode:
    def __init__(self):
        self.children = defaultdict(TrieNode)
        self.words = defaultdict(int)
        
class AutocompleteSystem:

    def __init__(self, sentences, times):
        self.root = TrieNode()
        self.cur = self.root
        self.prefix = ''
        for s, i in zip(sentences, times):
            self.add(s, i)
            
    def add(self, s, i):
        cur = self.root
        for c in s:
            cur = cur.children[c]
            cur.words[s] += i
    
    
    def input(self, c):
        res = []
        if c == '#':
            self.add(self.prefix, 1)
            self.cur = self.root
            return res
    
        self.prefix += c
        self.cur = self.cur.children[c]
        records, res = [], []
        for s, i in self.cur.words.items():
            heappush(records, (-i, s))
        for i in range(3):
            if records:
                _, s = heappop(records)
                res.append(s)
        return res
```
