```python
def reverseString(s, k):
    i, ans = 0, ""
    while i < len(s):
        ans += s[i:i + k][::-1] + s[i + k: i + 2 * k]
        i += 2 * k
    return ans
```

```python
def reverseString(s, k):
    s = list(s)
    for i in range(0, len(s), 2 * k):
        s[i:i + k] = reversed(s[i:i + k])
    return ''.join(s)
```

```python
reverseString('abcdefghijklmn', 2)
```
