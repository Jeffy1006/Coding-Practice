```python
import string
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False
    
    hashMap = {}
    alpha = set(list(string.ascii_letters[:26]))
    for i in range(len(s)):
        if s[i] in hashMap:
            if hashMap[s[i]] != t[i]:
                return False
        else:
            if t[i] in alpha:
                hashMap[s[i]] = t[i]
                alpha.remove(t[i])
            else:
                return False
    
    return True 
```

```python
def isIsomorphic(s, t):
    mapping_s_t = {}
    mapping_t_s = {}
    
    for c1, c2 in zip(s,t):
        if (c1 not in mapping_s_t) and (c2 not in mapping_t_s):
            mapping_s_t[c1], mapping_t_s[c2] = c2, c1
            # dict.get() return None if index not exist, dict[index] will throw Exception if index not exist
        elif (mapping_s_t.get(c1) != c2) or (mapping_t_s.get(c2) != c1):
            return False
    
    return True
```
