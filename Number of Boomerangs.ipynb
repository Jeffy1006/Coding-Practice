```python
def boomerangs(points):
    if len(points) < 3: return 0
    res = 0
    for p in points:
        cmap = {}
        for q in points:
            f = p[0] - q[0]
            s = p[1] - q[1]
            # distance function assumed to me Euclidean distance, not Manhattan distance
            cmap[f * f + s * s] = ! + cmap.get(f * f + s * s, 0)
        for k in cmap:
            res += cmap[k] * (cmap[k] - 1)
    return res
```
