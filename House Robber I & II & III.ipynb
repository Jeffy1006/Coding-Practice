```python
def rob(nums):
    rob1, rob2 = 0, 0
    
    for n in nums:
        temp = max(n + rob1, rob2)
        rob1 = rob2
        rob2 = temp
    
    return rob2
```

```python
def robii(nums):
    
    def rob(nums):
        rob1, rob2 = 0, 0

        for n in nums:
            temp = max(n + rob1, rob2)
            rob1 = rob2
            rob2 = temp

        return rob2
    # nums[0] is edge case where nums only have one element
    return max(nums[0], rob(nums[1:]), rob(nums[:-1]))
```

```python
def robiii(root):
    
    def dfs(root):
        if not root:
            return [0, 0]
        
        leftPair = dfs(root.left)
        rightPair = dfs(root.right)
        withRoot = root.val + leftPair[1] + rightPair[1]
        withoutRoot = max(leftPair) + max(rightPair)
        
        return [withRoot, withoutRoot]
    
    return max(dfs(root))
```

```python
nums = [1,2,3,2,1]
robii(nums)
```
