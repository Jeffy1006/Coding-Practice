```python
from itertools import accumulate
def leastBrick(wall):
    n = sum(wall[0])
    cut = [len(wall)] * (n)
    for w in wall:
        tmp = list(accumulate(w))[:-1]
        for i in tmp:
            cut[i] -= 1
    return min(cut[1:n])
```
