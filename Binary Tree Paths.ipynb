```python
class Node:
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None
        
def allPaths(root):
    res = []
    if not root:
        return res
    
    def helper(path, node):
        if not node.left and not node.right:
            path += str(node.val)
            res.append(path)
            return
        
        path = path + str(node.val) + "->"
        if node.left:
            helper(path, node.left)
        if node.right:
            helper(path, node.right)
        
        return
    
    helper("", root)
    return res
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(5)
a.left, a.right = b, c
b.right = d
allPaths(a)
```
