```python
def groupShiftedStrings(words):
    
    hashmap = {}
    res = []
    
    def helper(word):
        distance = ord(word[0]) - ord('a')
        k = "a"
        for i in range(1,len(word)):
            c = ord(word[i]) - distance
            if c < ord('a'):
                c += 26
            k += chr(c)
        
        return k
        
    for token in words:
        key = helper(token)
        if key in hashmap:
            hashmap[key].append(token)
        else:
            hashmap[key] = [token]
    
    for l in hashmap.values():
        res.append(l)
    
    return res
```
