```python
from collections import deque
def courseScheduleI(numCourses, prerequisites):
    if numCourses == 1:
        return [0]
    
    adjList = [[] for i in range(numCourses)]
    set_pre = list(set([(n[1], n[0]) for n in prerequisites]))
    in_degree = [0] * numCourses
    
    for n in set_pre:
        adjList[n[0]].append(n[1])
        in_degree[n[1]] += 1
    
    queue = deque()
    count = 0
    
    for i in range(numCourses):
        if in_degree[i] == 0:
            queue.append(i)
        
    while queue:
        n = queue.popleft()
        
        for c in adjList[n]:
            in_degree[c] -= 1
            if in_degree[c] == 0:
                queue.append(c)
            count += 1
            
    return count == numCourses
```
