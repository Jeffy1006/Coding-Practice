```python
def findLonelyPixel(picture):
    index = [row.index('B') for row in picture if row.count('B') == 1]
    t_picture = list(zip(*picture))
    return len([i for i in index if t_picture[i].count('B') == 1])
```

```python
def findLonelyPixel(picture):
    return sum(col.count('B') == 1 == picture[col.index('B')].count('B') for col in zip(*picture))
```

```python
picture = [
    ['W','W','B'],
    ['W','B','W'],
    ['B','W','W'],
]
findLonelyPixel(picture)
```

```python
def findLonelyPixelII(picture, N):
    row, col = len(picture), len(picture[0])
    ans = 0
    if row < 1: return ans
    
    checkRow = [r.count('B') for r in picture]
    checkCol = [c.count('B') for c in zip(*picture)]
    hashmap = {}
    for s in picture:
        tmp = ''.join(s)
        hashmap[tmp] = hashmap.get(tmp, 0) + 1
    
    for i in range(row):
        if checkRow[i] != N: continue
        for j in range(col):
            if checkCol[j] != N: continue
            if picture[i][j] == 'B':
                tmp = ''.join(picture[i])
                if hashmap[tmp] == N:
                    ans += 1
    return ans
```

```python
picture = [
    ['W','B','W','B','B','W'],
    ['W','B','W','B','B','W'],
    ['W','B','W','B','B','W'],
    ['W','W','B','W','B','W']
]
findLonelyPixelII(picture, 3)
```
