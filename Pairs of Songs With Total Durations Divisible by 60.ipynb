```python
from collections import defaultdict
import bisect
def numPairsDivisibleBy60(time):
    hashmap = defaultdict(list)
    for i, t in enumerate(time):
        hashmap[t % 60] .append(i)
    
    pairs = 0
    for i, t in enumerate(time):
        pair_t = - t % 60
        if pair_t in hashmap:
            n = len(hashmap[pair_t])
            index = bisect.bisect(hashmap[pair_t], i)
            pairs += n - index
    return pairs
```

```python
from collections import defaultdict
def numPairsDivisibleBy60(time):
    ans, cnt = 0, defaultdict(int) # or 60 buckets with index 0 - 59
    for t in time:
        theOther = -t % 60
        ans += cnt[theOther]
        cnt[t % 60] += 1
    return ans
```
