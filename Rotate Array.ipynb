```python
def rotateArray(nums, k):
    # k can be larger than length of nums
    k = k % len(nums)
    
    def reverse(i, j):
        while i < j:
            nums[i], nums[j] = nums[j], nums[i]
            # Please update counter
            i += 1
            j -= 1
    
    reverse(0, len(nums) - 1)
    reverse(0, k - 1)
    reverse(k, len(nums) - 1)

    return    
```

```python
def rotate(nums, k):
    n = len(nums)
    k = k % n
    
    start = count = 0
    while count < n:
        prev_idx, prev = start, nums[start]
        while True:
            next_idx = (prev_idx + k) % n
            nums[next_idx], prev = prev, nums[next_idx]
            prev_idx = next_idx
            count += 1
            #print(nums)
            if start == prev_idx:
                break
        #print()
        start += 1
```

```python
a = [1,2,3,4,5,6,7,8,9]
rotateArray(a, 6)
a
```
