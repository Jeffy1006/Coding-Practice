```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    
class Solution:
    def buildTree(arr):
        
        if len(arr) == 0:
            return None
        
        if len(arr) == 1:
            return TreeNode(arr[0])
        
        def Helper(InStart, InEnd):
            
            if InStart > InEnd:
                return None
            
            if InStart == InEnd:
                return TreeNode(arr[InStart])
            
            mid = (InStart + InEnd) // 2
            #print(mid)
            root = TreeNode(arr[mid])
            root.left = Helper(InStart, mid-1)
            root.right = Helper(mid+1, InEnd)
            
            return root
        
        res = Helper(0, len(arr)-1)
        
        return res
```

```python
def BTLevelOrderT(root):
    
    res = []
    if root == None:
        return
    
    queue1 = []
    queue1.append(root)
    queue2 = []
    cur = None
    
    while len(queue1) != 0 or len(queue2) != 0:
        
        tmp = []
        # Must check if len(queue1) != 0
        if len(queue1) != 0:
            while len(queue1) != 0:
                cur = queue1.pop(0)
                tmp.append(cur.data)
                if cur.left != None:
                    queue2.append(cur.left)
                if cur.right != None:
                    queue2.append(cur.right)
        else:
            while len(queue2) != 0:
                cur = queue2.pop(0)
                tmp.append(cur.data)
                if cur.left != None:
                    queue1.append(cur.left)
                if cur.right != None:
                    queue1.append(cur.right)
        
        res.append(tmp)
    
    return res
```

```python
nums = [-10,-3,0,5,9]
a = Solution.buildTree(nums)
BTLevelOrderT(a)
```
