```python
def licenseKeyFormatting(S, K):
    S = list(S.upper().replace('-', ''))
    length = len(S)
    head = K if length % K == 0 else length % K
    idx = head
    res = S[:head]
    hyp = ['-']
    while idx < length:
        res += hyp + S[idx:idx + K]
        idx += K
    return ''.join(res)
```

```python
licenseKeyFormatting("5f3Z-2e-9-w", 4)
```
