```python
def patchArray(nums, n):
    SUM = 0
    nextpatch = 1
    idx = 0
    res = 0
    while SUM < n:
        while idx < len(nums) and nextpatch >= nums[idx]:
            SUM += nums[idx]
            idx += 1
        if nextpatch > SUM:
            SUM += nextpatch
            res += 1
        nextpatch = SUM + 1
    return res
```
