```python
# TC: O(n)
# SC: O(k)
from collections import Counter
def subarraysDivByK(nums, k):
    cul_sum, prefix_sum = 0, [0]
    for num in nums:
        cul_sum += num
        prefix_sum.append(cul_sum % k)
    prefix_sum = Counter(prefix_sum)
    return sum( (value * (value - 1)) // 2 for value in prefix_sum.values())
```
