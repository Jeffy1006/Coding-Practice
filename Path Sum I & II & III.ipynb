```python
class TreeNode:
    def __init__(self, data):
        self.val = data
        self.left = None
        self.right = None
        
def pathSum(root,t):
    if not root: return False
    
    q = [(root,root.val)]
    while q:
        node,num = q.pop(0)
        if not node.left and not node.right:
            if num == t: return True
        if node.left: q.append((node.left,num+node.left.val))
        if node.right: q.append((node.right,num+node.right.val))
        
    return False

def pathSumR(root,t):
    if not root: return False

    def dfs(node,accuSum):
        if not node: return False
        if not node.left and not node.right:
            if (accuSum+node.val) == t:
                return True
        leftPath = dfs(node.left, accuSum+node.val)
        rightPath = dfs(node.right, accuSum+node.val)
        
        return (leftPath or rightPath)
    # Start with 0 instead of root.val
    return dfs(root,0)

def pathSumII(root,t):
    res = []
    if not root: return res
    
    def helper(root,t):
        q = [(root,str(root.val),root.val)]
        while q:
            node,path,num = q.pop(0)
            if not node.left and not node.right:
                #if num == t: res.append(path.split())
                if num == t: res.append(list(map(int, path.split())))
            if node.left:
                q.append((node.left,path + " " + str(node.left.val),num+node.left.val))
            if node.right:
                q.append((node.right,path + " " + str(node.right.val),num+node.right.val))
        
        return res
    
    return helper(root,t)

def pathSumRII(root,t):
    if not root: return False
    res = []
    def dfs(node,accuSum,path):
        if not node.left and not node.right:
            if (accuSum+node.val) == t:
                # Do not do append(path), path is a holder of the list's memory, it will keep changing
                # Instead, create a new list with the same content
                res.append([n for n in path])
        if node.left:
            path.append(node.left.val)
            dfs(node.left, accuSum+node.val, path)
            path.pop()
        if node.right:
            path.append(node.right.val)
            dfs(node.right, accuSum+node.val, path)
            path.pop()
        
        return
    
    dfs(root,0,[root.val])
    return res

def pathSumRIII(root,t):
    if not root: return 0
    
    def dfs(node,accuSum):
        if not node: return 0
        res = 0
        if node.val + accuSum == t: res += 1
        
        res += dfs(node.left, accuSum + node.val)
        res += dfs(node.right, accuSum + node.val)
        res += dfs(node.left,0)
        res += dfs(node.right,0)
        
        return res

    return dfs(root,0)

def pathSumRRIII(root,SUM):
    if not root: return 0
    
    def dfs(node,SUM):
        if not node: return 0
        res = 0
        if node.val == SUM: res += 1
        
        res += dfs(node.left, SUM - node.val)
        res += dfs(node.right, SUM - node.val)
        
        return res

    return pathSumRRIII(root.left, SUM) + dfs(root, SUM) + pathSumRRIII(root.right, SUM)
```
