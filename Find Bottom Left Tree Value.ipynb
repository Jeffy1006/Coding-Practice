```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
def findLeftMostNode(root):
    queue = [root]
    for node in queue:
        queue += filter(None, (node.right, node.left))
    return node.val    
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
a.left, a.right = b, c
c.left = d
findLeftMostNode(a)
```
