```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    
def SameTree(root1, root2):
        
    if root1 == None and root2 == None:
        return True
        
    if root1 != None and root2 != None:
        return (root1.data == root2.data) and SameTree(root1.left, root2.left) and SameTree(root1.right, root2.right)
        
    return False

def SymTreeHelper(root1, root2):
        
    if root1 == None and root2 == None:
        return True
        
    if root1 != None and root2 != None:
        return (root1.data == root2.data) and SymTreeHelper(root1.left, root2.right) and SymTreeHelper(root1.right, root2.left)
        
    return False
    
def SymTree(root):
        
    if root == None:
        return True
        
    return SymTreeHelper(root.left, root.right)
```

```python
from collections import deque
def isSym(root):
    if not root:
        return True
    
    queueL = deque([root.left])
    queueR = deque([root.right])
    
    while queueL and queueR:
        if len(queueL) != len(queueR):
            return False
        
        for i in range(len(queueL)):
            left = queueL.popleft()
            right = queueR.popleft()
            
            if not left and not right:
                continue
            elif left and right and left.val == right.val:
                queueL.append(left.left)
                queueL.append(left.right)
                queueR.append(right.right)
                queueR.append(right.left)
            else:
                return False
    
    return True
```
