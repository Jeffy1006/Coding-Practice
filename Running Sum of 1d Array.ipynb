```python
def runningSum(nums):
    cul_sum, res = 0, []
    for num in nums:
        cul_sum += num
        res.append(cul_sum)
    return res
```

```python
def runningSum(nums):
    i = 1
    while i<len(nums):
        nums[i]+=nums[i-1]
        i+=1
    return nums
```

```python
def runningSum(nums):
    res = [nums[0]] + [0] * (len(nums) - 1)
    for i, num in enumerate(nums[1:]):
        res[i + 1] += res[i] + num
    return res
```

```python
nums = [1,2,3,4]
runningSum(nums)
```
