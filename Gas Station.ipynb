```python
def gasStation(gas, cost):
    zipLists = zip(gas, cost)
    diff = [(x - y) for x, y in zipLists]
    if sum(diff) < 0:
        return -1
    else:
        start = 0
        maxStart = 0
        maxSum = diff[0]
        curSum = 0
        
        for i in range(0,len(diff)):
            if curSum < 0:
                curSum = 0
                start = i
            curSum += diff[i]
            if maxSum < curSum:
                maxSum = curSum
                maxStart = start
        
        return maxStart
```

```python
def canCompleteCircuit(gas, cost):
    if sum(gas) < sum(cost):
        return -1
    
    total = 0
    res = 0
    
    for i in range(len(gas)):
        total += (gas[i] - cost[i])
        
        if total < 0: 
            total = 0
            res = i + 1
    
    return res
```

```python
gas = [1,2,3,4,5]
cost = [3,4,5,1,2]
gasStation(gas, cost)
```
