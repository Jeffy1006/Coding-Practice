```python
class FenwickTree:
    def __init__(self, n):
        self.tree = [0] * (n + 1)
    
    def update(self, i, delta):
        while i < len(self.tree):
            self.tree[i] += delta
            i += (i & -i)
    
    def query(self, i):
        total = 0
        while i > 0:
            total += self.tree[i]
            i -= (i & -i)
        return total

class Solution:
    def countSamller(self, nums):
        sortedNums = sorted(set(nums))
        rank, ranks = {}, 0
        for num in sortedNums:
            rank += 1
            ranks[num] = rank
        
        ans = []
        tree = FenwickTree(rank)
        for i in range(len(nums) - 1, -1, -1):
            ans.append(tree.query(ranks[num[i]] - 1))
            tree.update(ranks[nums[i]], 1)
        
        return ans.reverse()
```
