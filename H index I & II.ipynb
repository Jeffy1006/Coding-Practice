```python
def hIndexII(citations):
    n = len(citations)
    low, high, mind = 0, n - 1, None
    
    while low <= high:
        mid = (high + low) // 2
        if citations[mid] == (n - mid):
            return citations[mid]
        elif citations[mid] > (n - mid):
            high = mid - 1
        else:
            low = mid + 1
    
    return n - low
```
