```python
import heapq
def kthSmallest(matrix, k):
    n = len(matrix)
    pq = []
    for j in range(min(k, n)):
        heapq.heappush(pq, (matrix[0][j], 0, j))
    while k > 1:
        num, i, j = heapq.heappop(pq)
        if i < n - 1:
            heapq.heappush(pq, (matrix[i + 1][j], i + 1, j))
        k -= 1
    return pq[0]
```
