```python
def canPlaceFlowers(flowerbed, n):
    flowerbed = [0] + flowerbed + [0]
    def helper(idx, remain):
        if remain == 0: return True
        while idx < len(flowerbed) - 1 and \
            not (flowerbed[idx - 1] == 0 and flowerbed[idx] == 0 and flowerbed[idx + 1] == 0):
            idx += 1
        if idx == len(flowerbed) - 1: return False
        flowerbed[idx] = 1
        return helper(idx + 2, remain - 1)
    return helper(1, n)
```

```python
def canPlaceFlowers(flowerbed, n):
    flowerbed = [0] + flowerbed + [0]
    for i in range(1, len(flowerbed)):
        if not flowerbed[i - 1] and not flowerbed[i] and not flowerbed[i + 1]:
            n -= 1
            flowerbed[i] = 1
    return n <= 0
```
