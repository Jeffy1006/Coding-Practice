```python
from collections import Counter
import math
import sys
from decimal import *
class Solution:
    def maxPoints(self, points):
        
        def getSlope(p1, p2):
            yDiff = p2[1] - p1[1]
            xDiff = p2[0] - p1[0]
            
            if xDiff == 0:
                return sys.maxsize
            

            # not enough precision, e.g. [[0,0],[94911150,94911151],[94911151,94911152]]
            #angles = math.atan2(yDiff, xDiff) / math.pi * 180
            
            #if angles < 0:
            #    angles += 360
            
            return Decimal(yDiff) / Decimal(xDiff)
        
        ans = 0
        n = len(points)
        
        for i in range(n):
            p1 = points[i]
            same = 0
            slopes = []
            for j in range(n):
                if i != j:
                    p2 = points[j]
                    if p1 == p2:
                        same += 1
                    else:
                        slopes.append(getSlope(p1, p2))
            
            counter = Counter(slopes)
            max1 = 0
            if counter.values():
                max1 = max(counter.values())
            # add 1 for p1 itself
            ans = max(ans, max1 + same + 1)
        
        return ans
```
