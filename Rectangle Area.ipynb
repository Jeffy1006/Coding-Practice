```python
def computeArea(ax1, ay1, ax2, ay2, bx1, by1, bx2, by2):
    if not (ax2 < bx1 or bx2 < ax1 or ay2 < by1 or by2 < ay1):
        cx1, cy1 = max(ax1, bx1), max(ay1, by1)
        cx2, cy2 = min(ax2, bx2), min(ay2, by2)
    else:
        cx1 = cx2 = cy1 = cy2 = 0
    print("{} {} {} {}".format(cx1, cy1, cx2, cy2))
    return ((ax2 - ax1) * (ay2 - ay1)) + ((bx2 - bx1) * (by2 - by1)) - ((cx2 - cx1) * (cy2 - cy1))
```

```python
computeArea(-3,0,3,4,0,-1,9,2)
```
