```python
def findMinDifference(timePoints):
    n = len(timePoints)
    if len(set(timePoints)) != n: return 0
    times = [int(time[:2]) * 60 + int(time[3:]) for time in timePoints]
    times.sort()
    min_diff = 1440 - times[-1] + times[0]
    for i in range(1, n):
        min_diff = min(min_diff, times[i] - times[i - 1])
    return min_diff
```

```python
def findMinDifference(timePoints):
    def convert(time):
        return int(time[:2]) * 60 + int(time[3:]) 
    minutes = list(map(convert, timePoints))
    minutes.sort()
    return min((y - x) % (24 * 60) for x, y in zip(minutes, minutes[1:] + minutes[:1]))
```
