```python
def findRadius(houses, heaters):
    heaters = sorted(heaters) + [float('inf')]
    i = radius = 0
    for house in sorted(houses):
        while house >= sum(heaters[i:i + 2]) / 2:
            i += 1
        radius = max(radius, abs(heaters[i] - house))
    return radius
```

```python
findRadius([1,5], [2])
```
