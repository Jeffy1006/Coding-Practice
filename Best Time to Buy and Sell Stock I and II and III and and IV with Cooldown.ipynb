```python
# Space O(n)
def bestTime(arr):
    dp = [0]*len(arr)
    futureMax = 0
    
    for i in range(len(arr)-2,-1,-1):
        # update futureMax first before update dp
        futureMax = max(futureMax,arr[i+1])
        dp[i] = futureMax
    maxGain = 0
    for i in range(len(arr)-1):
        maxGain = max(maxGain, dp[i]-arr[i])
    
    return maxGain
```

```python
# DP - Memeorization
def bestTimeIV(prices, k):
    mem = {}
    
    def recursion(pos, k, bought):
        if pos >= len(prices) or k == 0:
            return 0
        if (bought, k, pos) in mem:
            return mem[(bought, k, pos)]
        
        res = recursion(pos + 1, k, bought)
        if bought:
            res = max(res, recursion(pos + 1, k - 1, 0) + prices[pos])
        else:
            res = max(res, recursion(pos + 1, k, 1) - prices[pos])
        
        return res
    
    return recursion(0, k, 0)
```

```python
prices = [2,3,4]
bestTimeIII(prices, 2)
```

```python
# DP - Memeorization
def bestTimeCooldown(prices):
    mem = {}
    n = len(prices)
    
    def findMax(curr, n):
        if curr >= n:
            return 0
        if curr in mem:
            return mem[curr]
        
        maxVal = 0
        # This for loop assumed we bought the stock on currth day
        for i in range(curr + 1, n):
            if prices[curr] < prices[i]:
                maxVal = max(maxVal, prices[i] - prices[curr] + findMax(i + 2, n))
        # This line assumed we didn't buy the stock on currth day (exclude curr day)
        maxVal = max(maxVal, findMax(curr + 1, n))
        mem[curr] = maxVal
        
        return maxVal
    
    return findMax(0, n)
```

```python
# State Machine
def bestTimeCooldown(prices):
    n = len(prices)
    if n <= 1: return 0
    
    noStock1 = noStock2 = 0
    inHand1 = inHand2 = -prices[0]
    Sold1 = Sold2 = 0
    
    for i in range(1, n):
        noStock2 = max(noStock1, Sold1)
        inHand2 = max(inHand1, noStock1 - prices[i])
        Sold2 = inHand1 + prices[i]
        noStock1 = noStock2
        inHand1 = inHand2
        Sold1 = Sold2
    
    return max(noStock2, Sold2)
```

```python
prices = [9,1,3,5,7,6,9]
bestTimeCooldown(prices)
```
