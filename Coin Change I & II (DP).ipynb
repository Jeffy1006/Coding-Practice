```python
def coinChange(coins, amount):
    dp = [amount + 1] * (amount + 1)
    dp[0] = 0
    
    for a in range(1, amount + 1):
        for c in coins:
            if a - c >= 0:
                dp[a] = min(dp[a], 1 + dp[a - c])
    
    return dp[amount] if dp[amount] != amount + 1 else -1
```

```python
def coinChange(coins, amount):
    ans = float("inf")
    coins.sort(reverse = True)
    
    def helper(s, amount, count):
        nonlocal ans
        coin = coins[s]
        if s == len(coins) -1:
            if amount % coin == 0:
                ans = min(ans, count + amount / coin)
            else:
                k = amount / coin
                while k >= 0 and count + k < ans:
                    helper(s + 1, amount - k * coin, count + k)
                    k -= 1
    
    helper(0, amount, 0)
    return ans if ans != float("inf") else -1
```

```python
def coinChangeII(coins, amount):
    mem = {}
    res = 0
    def helper(idx, remain):
        if idx == len(coins) and remain == 0: return 1
        if idx == len(coins) or remain < 0: return 0
        if (idx, remain) in mem:
            return mem[(idx, remain)]
        
        res = helper(idx, remain - coins[idx]) + helper(idx + 1, remain)
        mem[(idx, remain)] = res
        return res
    return helper(0, amount)
```

```python
from functools import lru_cache
def coinChangeII(coins, amount):
    @lru_cache(maxsize = None)
    def change_(amount, coins):
        if amount == 0: return 1
        if amount < 0 or len(coins) == 0: return 0
        return change_(amount - coins[-1], coins) + change_(amount, coins[:-1])
    return change_(amount, tuple(coins))
```
