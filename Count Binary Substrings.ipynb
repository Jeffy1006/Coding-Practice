```python
 def countBinarySubstrings(s):
        zeros, ones, prev = 0, 0, None
        res = 0
        s += '0' if s[-1] == '1' else '1'
        for c in s:
            if c == '0':
                if prev == '0': zeros += 1
                else:
                    res += min(zeros, ones)
                    zeros = 1
            else:
                if prev == '1': ones += 1
                else: 
                    res += min(zeros, ones)
                    ones = 1
            prev = c
        return res
```

```python
def countBinarySubstrings(s):
    s = map(len, s.replace('01', '0 1').replace('10', '1 0').split())
    return sum(min(a, b) for a, b in zip(s, s[1:]))
```
