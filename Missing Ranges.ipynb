```python
def findMissingRanges(nums, lower, upper):
    def formatRange(lower, upper):
        if lower == upper:
            return str(lower)
        return str(lower) + "->" + str(upper)
    
    res = []
    prev = lower - 1
    for i in range(len(nums) + 1):
        curr = nums[i] if i < len(nums) else upper + 1
        if prev + 1 <= curr - 1:
            res.append(formatRange(prev + 1, cur - 1))
        prev = curr
    
    return res
```
