```python
def deleteNode(root, key):
    if not root: return root
    if root.val > key:
        root.left = deleteNode(root.left, key)
    elif root.val < key:
        root.right = deleteNode(root.right. key)
    else:
        if not root.right:
            return root.left
        if not root.left:
            return root.right
        temp = root.right
        mini = temp.val
        while temp.left:
            temp = temp.left
            mini = temp.val
        root.val = mini
        root.right = deleteNode(root.right, mini)
    return root
```
