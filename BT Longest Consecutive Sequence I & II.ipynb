```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
def btLongestConsecutiveSeq(root):
    res = []
    maxLen = 0
    
    def dfs(node,count):
        nonlocal maxLen
        if not node:
            return None
        
        res.append(node.val)
        if len(res) == 1 or res[-1] == res[-2] + 1:
            count += 1
        else:
            count = 1
        maxLen = max(maxLen, count)
        dfs(node.left, count)
        dfs(node.right, count)
        res.pop()
    
    dfs(root, 0)
    return maxLen
```

```python
def btLongestConsecutiveSeq(root):
    maxLen = 0
    
    def dfs(node, count, target):
        nonlocal maxLen
        if not node:
            return None
        elif node.val == target:
            count += 1
        else:
            count = 1
        
        maxLen = max(maxLen, count)
        dfs(node.left, count, node.val + 1)
        dfs(node.right, count, node.val + 1)    
    
    dfs(root, 0, 0)
    return maxLen
```

```python
def btLongestConsecutiveSeqII(root):
    def helper(node, parent):
        nonlocal max_length
        if not node: 
            return 0, 0
        l_inc, l_dec = helper(node.left, node)
        r_inc, r_dec = helper(node.right, node)
        max_length = max(max_length, l_inc + r_dec + 1, l_dec + r_inc + 1)
        if parent and node.val + 1 == parent.val:
            return max(l_inc, r_inc) + 1, 0
        elif parent and node.val - 1 == parent.val:
            return 0, max(l_dec, r_dec) + 1
        return 0, 0

    max_length = 0
    helper(root, None)
    return max_length
```

```python
a = Node(1)
b = Node(3)
c = Node(2)
d = Node(4)
e = Node(5)
a.right = b
b.left, b.right = c, d
d.right = e
btLongestConsecutiveSeq(a)
```
