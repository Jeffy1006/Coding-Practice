```python
def findCheapestPrice(n, flights, src, dst, K):
    graph = collections.defaultdict(dict)
    for s, d, w in flights:
        graph[s][d] = w
    pq = [(0, src, K+1)]
    vis = [0] * n
    while pq:
        w, x, k = heapq.heappop(pq)
        if x == dst:
            return w
        if vis[x] >= k:
            continue
        vis[x] = k
        for y, dw in graph[x].items():
            heapq.heappush(pq, (w+dw, y, k-1))
    return -1
```
