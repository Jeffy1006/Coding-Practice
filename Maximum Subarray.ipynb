```python
def maximumSumSubarray(nums):
    maxSub = nums[0]
    curSum = 0
    
    for n in nums:
        if curSum < 0:
            curSum = 0
        curSum += n
        maxSub = max(maxSub, curSum)
    return maxSub
```

```python
def maximumProductSubarray(nums):
    maxSub = nums[0]
    curPro = 1
    
    for n in nums:
        if curPro < 1:
            curPro = 1
        curPro *= n
        maxSub = max(maxSub, curPro)
    return maxSub
```

```python
# return index o
def maximumSumSubarray(nums):
    maxSub = nums[0]
    curSum = 0
    maxStart = 0
    start = 0
    
    for i in range(0,len(nums)):
        if curSum < 0:
            curSum = 0
            start = i
        curSum += nums[i]
        if curSum > maxSub:
            maxSub = curSum
            maxStart = start
            
    return maxSub, maxStart
```
