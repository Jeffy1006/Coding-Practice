```python
def sortTransformedArray(arr, a, b, c):
    
    def helper(x):
        return (a * x * x + b * x + c)
    
    n = len(arr)
    start, end = 0, n - 1
    ans = []
    if a >= 0:
        for _ in range(n):
            x1 = helper(arr[start])
            x2 = helper(arr[end])
            if x1 > x2:
                ans.append(x1)
                start += 1
            else:
                ans.append(x2)
                end -= 1
        ans.reverse()
    else:
        for _ in range(n):
            x1 = helper(arr[start])
            x2 = helper(arr[end])
            if x1 < x2:
                ans.append(x1)
                start += 1
            else:
                ans.append(x2)
                end -= 1
    return ans
```

```python
sortTransformedArray([-1, 0, 1, 2, 3, 4], 1, 2, -1)
```
