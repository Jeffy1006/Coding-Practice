```python
def minDifference(nums):
    if len(nums) < 5: return 0
    max1 = max2 = max3 = max4 = float('-inf')
    min1 = min2 = min3 = min4 = float('inf')
    
    for num in nums:
        if num > max1:
            max4 = max3
            max3 = max2
            max2 = max1
            max1 = num
        elif num > max2:
            max4 = max3
            max3 = max2
            max2 = num
        elif num > max3:
            max4 = max3
            max3 = num
        elif num > max4:
            max4 = num
        
        if num < min1:
            min4 = min3
            min3 = min2
            min2 = min1
            min1 = num
        elif num < min2:
            min4 = min3
            min3 = min2
            min2 = num
        elif num < min3:
            min4 = min3
            min3 = num
        elif num < min4:
            min4 = num
    # largest one can be very large and smallest one can be very small
    return min(max1 - min4, max2 - min3, max3 - min2, max4 - min1)
```

```python
def minDifference(nums):
    nums.sort()
    return min(b - a for a, b in zip(nums[:4], nums[-4:]))
```

```python
import heapq
def minDifference(nums):
    return min(a - b for a,b in zip(heapq.nlargest(4, nums), heapq.nsmallest(4, nums)[::-1]))
```

```python
nums = [5,3,2,4]
minDifference(nums)
```
