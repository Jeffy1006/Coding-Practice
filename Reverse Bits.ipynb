```python
def reverseBits(n):
    res = 0
    
    for i in range(32):
        bit = (n >> i) & 1
        res = res | (bit << (31 - i))
    
    return res
```

```python
def reverseBits(n):
    res, power = 0, 31
    
    while n:
        res += (n & 1) << power
        n >>= 1
        power -= 1
    
    return res
```
