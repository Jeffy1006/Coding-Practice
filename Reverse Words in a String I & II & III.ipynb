```python
def reverseWords(s):
    if not s:
        return None
    
    result, i, n = "", 0, len(s)
    
    while i < n:
        while i < n and s[i] == " ":
            i += 1
        if i >= n: break
        j = i + 1
        while j < n and s[j] != " ":
            j += 1
        w = s[i:j]
        if len(result) == 0:
            result = w
        else:
            result = w + " " + result
        i = j + 1

    return result
```

```python
def reverseWordsII(s):
    if not s: return s
    # string is immutable in Python
    s_arr = list(s)
    
    def reverse(i, j):
        nonlocal s_arr
        while i < j:
            s_arr[i], s_arr[j] = s_arr[j], s_arr[i]
            i += 1
            j -= 1
    
    reverse(0, len(s) - 1) 
    i = j = 0
    while j <= len(s_arr):
        # Always first check if reach the end or not
        if j == len(s_arr) or s_arr[j] == " " :
            reverse(i, j - 1)
            i = j + 1
        j += 1
    
    return "".join(s_arr)
```

```python
def reverseWordsII(s):
    return ' '.join(s.split()[::-1])
```
