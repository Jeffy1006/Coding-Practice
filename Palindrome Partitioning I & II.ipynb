```python
def palindromePartition(s):
    res, mem = [], {}
    if len(s) == 0: return res
    
    def isPal(lo, hi):
        while lo <= hi:
            if s[lo] != s[hi]:
                return False
            lo, hi = lo + 1, hi - 1
        return True
        
    def helper(cur, lo):
        if lo == len(s):
            res.append(cur[:])
            return
        
        for hi in range(lo, len(s)):
            if isPal(lo, hi):
                cur.append(s[lo:hi+1])
                helper(cur, hi + 1)
                cur.pop()
    
    helper(res, [], 0)
    return res
```

```python
def palindromePartitionII(s):
    
    isPal = [[False for j in range(len(s) + 1)] for i in range(len(s) + 1)]
    
    def computeAllPals(A):
        nonlocal isPal
        for i in range(len(A),0,-1):
            isPal[i][i - 1] = True
            isPal[i][i] = True
            for j in range(i +1 ,len(A) + 1):
                if A[i - 1] == A[j - 1]:
                    isPal[i][j] = isPal[i + 1][j - 1]
                else:
                    isPal[i][j] = False
            
    
    computeAllPals(s)
    minPals = [float("inf")] * (len(s) + 2)
    minPals[len(s) + 1] = 0
    for k in range(len(s), 0, -1):
        for l in range(k, len(s) + 1):
            if isPal[k][l]:
                minPals[k] = min(minPals[k], 1 + minPals[l + 1])
    print(minPals)
    return minPals[1] - 1
    
```
