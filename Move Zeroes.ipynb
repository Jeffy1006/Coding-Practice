```python
def moveZeroes(nums):
    zeroIdx = nonzeroIdx = 0
    while zeroIdx < len(nums) and nonzeroIdx < len(nums):
        while zeroIdx < len(nums) and nums[zeroIdx] != 0:
            zeroIdx += 1
        if nonzeroIdx == 0:
            nonzeroIdx = zeroIdx
        while nonzeroIdx < len(nums) and nums[nonzeroIdx] == 0:
            nonzeroIdx += 1
        if zeroIdx == len(nums) or nonzeroIdx == len(nums): break
        nums[zeroIdx], nums[nonzeroIdx] = nums[nonzeroIdx], nums[zeroIdx]
```
