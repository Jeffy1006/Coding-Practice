```python
def lexicographicalNumbers(n):
    res = []
    if n == 0: return res
    
    def helper(num):
        for i in range(0, 10):
            tmp = num * 10 + i
            if tmp > n:
                return
            res.append(tmp)
            helper(tmp)
    
    for i in range(1, 10):
        if i > n:
            break
        res.append(i)
        helper(i)
    return res
```

```python
def findKthNumber(n, k):
    cur = 1
    k -= 1
    
    def calSteps(n, cur):
        steps = 0
        n1, n2 = cur, cur + 1
        while n1 <= n:
            steps += min(n + 1, n2) - n1
            n1 *= 10
            n2 *= 10
        return steps
        
    while k > 0:
        steps = calSteps(n, cur)
        if steps <= k:
            cur += 1
            k -= steps
        else:
            cur *= 10
            k -= 1
    return cur
```
