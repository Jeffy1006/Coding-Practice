```python
# TC: O(n)
# SC: O(n)
def circularArrayLoop(nums):
    done, n =set(), len(nums)
    if n <= 1: return False
    for i in range(n):
        if i in done: continue
        j, cur = i, set()
        while j not in cur and j not in done and nums[i] * nums[j] > 0:
            cur.add(j)
            j = (j + nums[j]) % n
        if j in cur and nums[j] % n != 0:
            return True
        done = done.union(cur)
    return False
```

```python
circularArrayLoop([1,4,1,-1])
```
