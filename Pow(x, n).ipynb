```python
def myPow(x, n):
    if x == 0: return 0
    if n == 0: return 1
    
    def helper(n):
        if n == 1:
            return x
        
        val = helper(n // 2)
        val *= val
        return val * x if n % 2 else val
    
    res = helper(n)
    return res if n > 0 else 1 / res
```
