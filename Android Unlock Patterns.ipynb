```python
# TC : O(n * n!)
# SC : O(n)
def numberOfPatterns(m, n):    
    def dfs(cur_num, remain):
        if remain == 0:
            return 1
        
        ans = 0
        for next_num in range(1,10):
            if next_num not in seen and ((cur_num, next_num) not in skip or skip[(cur_num, next_num)] in seen):
                seen.add(next_num)
                ans += dfs(next_num, remain - 1)
                seen.remove(next_num)
        return ans
    
    skip = {
        (1,3):2
        ,(1,3):2
        ,(1,7):4
        ,(7,1):4
        ,(3,9):6
        ,(9,3):6
        ,(7,9):8
        ,(9,7):8
        ,(1,9):5
        ,(9,1):5
        ,(3,7):5
        ,(7,3):5
        ,(2,8):5
        ,(8,2):5
        ,(4,6):5
        ,(6,4):5
    }
    
    ans = 0
    for i in range(m, n + 1):
        for start in range([1,2,5]):
            seen = set([start])
            ans += dfs(start, 1 - 1)
            if start <= 2:
                ans *= 4
    return ans
```
