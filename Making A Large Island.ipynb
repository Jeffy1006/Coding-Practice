```python
def largestIsland(grid):
    def dfs(i, j, id):
        if 0 <= i < m and 0 <= j < n and grid[i][j] == 1:
            grid[i][j] = id
            res = 1
            for d in directions:
                res += dfs(i + d[0], j + d[1], id)
            return res
        return 0
    
    m, n = len(grid), len(grid[0])
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    sizes, id = {}, 1
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 1:
                id += 1
                sizes[id] = dfs(i, j, id)
    
    if id == 1: return 0
    if id == 2: return min(m * n, sizes[id] + 1)
    
    res, bridges = max(sizes.values()) + 1, {}
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 0:
                # can be more than two islands
                seen = set()
                for d in directions:
                    ni, nj = i + d[0], j + d[1]
                    if 0 <= ni < m and 0 <= nj < n and grid[ni][nj] != 0:
                        seen.add(grid[ni][nj])
                res = max(res, 1 + sum(sizes[i] for i in seen))            
    return res
```

```python
class Solution(object):
    def largestIsland(self, grid):
        N = len(grid)

        def neighbors(r, c):
            for nr, nc in ((r-1, c), (r+1, c), (r, c-1), (r, c+1)):
                if 0 <= nr < N and 0 <= nc < N:
                    yield nr, nc

        def dfs(r, c, index):
            ans = 1
            grid[r][c] = index
            for nr, nc in neighbors(r, c):
                if grid[nr][nc] == 1:
                    ans += dfs(nr, nc, index)
            return ans

        area = {}
        index = 2
        for r in xrange(N):
            for c in xrange(N):
                if grid[r][c] == 1:
                    area[index] = dfs(r, c, index)
                    index += 1

        ans = max(area.values() or [0])
        for r in xrange(N):
            for c in xrange(N):
                if grid[r][c] == 0:
                    seen = {grid[nr][nc] for nr, nc in neighbors(r, c) if grid[nr][nc] > 1}
                    ans = max(ans, 1 + sum(area[i] for i in seen))
        return ans
```
