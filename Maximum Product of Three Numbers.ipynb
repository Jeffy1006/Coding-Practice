```python
def maximumProduct(nums):
    max1, max2, max3 = None, None, None
    min1, min2 = None, None
    for num in nums:
        if not min1 or min1 > num:
            min2 = min1
            min1 = num
        elif not min2 or min2 > num:
            min2 = num
    
        if not max1 or max1 < num:
            max3 = max2
            max2 = max1
            max1 = num
        elif not max2 or max2 < num:
            max3 = max2
            max2 = num
        elif not max3 or max3 < num:
            max3 = num
    return max(max1 * max2 * max3, min1 * min2 * max1)
```

```python
nums = [-1,-2,-3,1]
maximumProduct(nums)
```
