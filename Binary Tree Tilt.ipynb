```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
```

```python
def findTilt(root):
    tilt = 0
    def helper(node):
        nonlocal tilt
        if not node: return 0
        tmp = node.val
        left, right = helper(node.left), helper(node.right)
        node.val = abs(left - right)
        tilt += node.val
        return tmp + left + right
    helper(root)
    return tilt
```
