```python
def fractionToDecimal(numerator, denominator):
    if numerator == 0: return '0'
    
    ans = ''
    if (numerator < 0 and denominator > 0) or (numerator > 0 and denominator < 0):
        ans += "-"
        
    dividend = abs(numerator)
    divisor = abs(denominator)
    remainder = dividend % divisor
    ans += str(dividend // divisor)
    
    if remainder == 0:
        return ans
    
    ans += "."
    numMap = {}
    while remainder != 0:
        if remainder in numMap:
            ans = ans[:numMap[remainder]] + "(" + ans[numMap[remainder]:] + ")" 
            break
        else:
            numMap[remainder] = len(ans)
            remainder *= 10
            ans += str((remainder // divisor))
            remainder %= divisor
    
    return ans
```

```python
numerator = 4
denominator = 333
fractionToDecimal(numerator, denominator)
```
