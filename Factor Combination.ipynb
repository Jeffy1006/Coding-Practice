```python
def getFactors(n):
    res = []
    # start is used to prevent duplicate
    def helper(temp, num, start):
        nonlocal res
        if num == 1:
            if len(temp) > 1:
                # create hard coy of list
                res.append(temp.copy())
                return
            
        for i in range(start, num + 1):
            if num % i == 0:
                temp.append(i)
                helper(temp, num // i, i)
                temp.pop()
            
    
    helper([], n, 2)
    return res
```

```python
getFactors(12)
```
