```python
def rankTeams(votes):
    n = len(votes[0])
    ranks = {}
    for vote in votes:
        for i in range(n):
            if vote[i] not in ranks:
                ranks[vote[i]] = [0] * n
            ranks[vote[i]][i] += 1
    ranks = sorted([(v, k) for k, v in ranks.items()], key = lambda x: x[1])
    ranks = sorted(ranks, key = lambda x: x[0], reverse = True)
    return ''.join([word for _, word in ranks])
```

```python
def rankTeams(votes):
    count = {v: [0] * len(votes[0]) + [v] for v in votes[0]}
    for vote in votes:
        for i, v in enumerate(vote):
            count[v][i] -= 1
    return ''.join(sorted(votes[0], key=count.get))
```
