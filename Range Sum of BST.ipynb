```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def rangeSumBST(root, low, high):
    def dfs(node):
        if not node: return 0
        
        if low <= node.val <= high:
            return node.val + dfs(node.left) + dfs(node.right)
        elif node.val < low:
            return dfs(node.right)
        else: 
            return dfs(node.left)
    
    return dfs(root)
```

```python
a = TreeNode(10)
b = TreeNode(5)
c = TreeNode(15)
d = TreeNode(3)
e = TreeNode(7)
f = TreeNode(13)
g = TreeNode(18)
h = TreeNode(1)
i = TreeNode(6)
a.left, a.right = b, c
b.left, b.right = d, e
c.left, c.right = f, g
d.left = h
e.left = i
```

```python
rangeSumBST(a, 6, 10)
```
