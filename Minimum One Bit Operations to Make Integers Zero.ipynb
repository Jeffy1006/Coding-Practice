```python
def minimumOneBitOperations(n):
    if n == 0: return 0
    if n == 1: return 1
    res = [0, 1]
    num = 2
    while num <= n:
        tmp = num * 2 - 1
        for i in range(len(res)):
            res.append(tmp - res[i])
            if num == n:
                return res[-1]
            num += 1
```

```python
def minimumOneBitOperations(n):
    if n <= 1: return n
    bit = 0
    while (1 << bit) <= n:
        bit += 1
    return ((1 << bit) - 1) - minimumOneBitOperations(n - (1 << (bit - 1)))
```
