```python
from heapq import *
from collections import Counter
def reorganizeString(s):
    res, c = [], Counter(s)
    pq = [(-v, k) for k, v in c.items()]
    heaplify(pq)
    p_cnt, p_ch = 0, ''
    while pq:
        cnt, ch = heappop(pq)
        res.append(ch)
        if p_cnt < 0:
            heappush((p_cnt, p_ch))
        cnt += 1
        p_cnt, p_ch = cnt, ch
    res = ''.join(res)
    if len(res) != len(s): return ''
    return res
```
