```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def convertBST(root):
    def helper(node):
        nonlocal prev
        if node.right: helper(node.right)
        node.val += prev
        prev = node.val
        if node.left: helper(node.left)
    prev = 0
    helper(root)
```

```python
a = Node(4)
b = Node(1)
c = Node(6)
d = Node(0)
e = Node(2)
f = Node(5)
g = Node(7)
h = Node(3)
i = Node(8)
a.left, a.right = b, c
b.left, b.right = d, e
c.left, c.right = f, g
e.right = h
g.right = i
convertBST(a)
```
