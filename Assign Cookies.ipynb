```python
# TC: O(nlgn)
# SC: O(1)
def assignCookies(s, g):
    g.sort()
    s.sort()
    
    ans = 0
    j = 0
    for i in range(len(g)):
        while j < len(s) and s[j] < g[i]:
            j += 1
        if j < len(s):
            ans += 1
            j += 1
    return ans
```
