```python
def consecutiveNumbersSum(self, N: int) -> int:
    i, ans = 1, 0
    while N > i * (i - 1) // 2:
        if (N - i * (i - 1) // 2) % i == 0:
            ans += 1
        i += 1
    return ans
```

```python
def consecutiveNumbersSum(self, N):
    res = 1
    i = 3
    while N % 2 == 0:
        N /= 2
    while i * i <= N:
        count = 0
        while N % i == 0:
            N /= i
            count += 1
        res *= count + 1
        i += 2
    return res if N == 1 else res * 2
```
