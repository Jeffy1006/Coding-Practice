```python
#from collections import defaultdict
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word_list = []

class Solution:
    def suggestedProducts(self, products, searchWord):
        root = TrieNode()
        
        for word in products:
            cur = root
            for c in word:
                if c not in cur.children:
                    cur.children[c] = TrieNode()
                cur = cur.children[c]
                cur.word_list.append(word)
                if len(cur.word_list) > 3:
                    cur.word_list.sort()
                    cur.word_list.pop()
        
        cur = root
        res = []
        for i, c in enumerate(searchWord):
            if c not in cur.children:
                return res.extend([[] * (len(searchWord) - i)])
            cur = cur.children[c]
            res.append(cur.word_list)
        
        return res
```
