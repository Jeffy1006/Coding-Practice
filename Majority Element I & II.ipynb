```python
def majorityElement(nums):
    moore = nums[0]
    count = 1
    for i in range(len(nums)):
        if nums[i] == moore:
            count += 1
        else:
            count -= 1
            if count == 0:
                moore = nums[i]
                count = 1
     
    return moore
```

```python
def majorityElementII(nums):
    
    number1 = number2 = -1
    count1 = count2 = 0
    for i in range(len(nums)):
        if nums[i] == number1: 
            count1 += 1
        elif nums[i] == number2: 
            count2 += 1
        elif count1 == 0:
            number1, count1 = nums[i], 1
        elif count2 == 0:
            number2, count2 = nums[i], 1
        else:
            count1, count2 = count1 - 1, count2 - 1
     
    ans = []
    if nums.count(number1) > (len(nums) / 3):
        ans.append(number1)
    if nums.count(number2) > (len(nums) / 3):
        ans.append(number2)
    
    return ans
```
