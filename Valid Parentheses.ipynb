```python
def isValid(s):
    stack = []
    closeToOpen = {")" : "(", "]" : "[", "}" : "{"}
    
    for c in s:
        if c in closeToOpen:
            if stack and closeToOpen[c] == stack[-1]: 
                stack.pop()
            else:
                return False
        else:
            stack.append(c)
    
    return True if not stack else False
```
