```python
def findSubstringInWraproundString(p):
    res = 1
    count = 1
    start = 0
    hashset = {p[0]}
    for i in range(len(p) - 1):
        if (ord(p[i]) + 1 - 97) % 26 == ord(p[i + 1]) - 97:
            if p[start:i + 2] not in hashset:
                count += 1
                res += count
                hashset.add(p[start:i + 2])
        else:
            count = 1
            start = i + 1
            if p[start] not in hashset:
                res += count
                hashset.add(p[start])

    return res    
```

```python
def findSubstringInWraproundString(p):
    res = {i : 1 for i in p}
    l = 1
    for i, j in zip(p, p[1:]):
        l = l + 1 if (ord(j) - ord(i)) % 26 == 1 else 1
        res[j] = max(res[j], l)
    return sum(res.values())
```
