```python
from sortedcontainers import SortedList
from collections import defaultdict
def findItinerary(tickets):
    adj = defaultdict(SortedList)
    ans = []
    n = len(tickets)
    
    for i in range(n):
        adj[tickets[i][0]].add(tickets[i][1])
    
    stack = []
    stack.append('JFK')
    while stack:
        src = stack[-1]
        if not adj[src]:
            ans.append(src)
            stack.pop()
        else:
            dst = adj[src][0]
            stack.append(dst)
            adj[src].remove[dst]
    return ans.reverse()
```
