```python
def twoSumII(nums, target):
    start, end = 0, len(nums) - 1
    while start < end:
        if nums[start] + nums[end] == target:
            return [start + 1, end + 1]
        elif nums[start] + nums[end] > target:
            end -= 1
        else:
            start += 1
    
    return []
```

```python
twoSumII([-4,-1,0,3,6,10], 5)
```

```python
# Two Sum III
class Solution:
    def __init__(self):
        self.arr = []
        
        
    def add(self, number):
        self.arr.append(number)
        
    
    def find(self, value):
        hashMap = {}
        for num in self.arr:
            if num in hashMap:
                return True
            else:
                hashMap[value - num] = num
        return False
```
