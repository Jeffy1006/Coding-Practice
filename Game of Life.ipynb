```python
def gameOfLife(board):
    ROW, COL = len(board), len(board[0])
    
    def helper(i, j):
        count = 0
        for r in range(i - 1, i + 2):
            for c in range(j - 1, j + 1):
                if r == i and c == j: continue
                if 0 <= r < ROW and 0 <= c < COL and abs(board[r][c]) == 1:
                    count += 1
        return count
    
    for i in range(ROW):
        for j in range(COL):
            count = helper(i, j)
            
            if board[i][j] == 1 and (count < 2 or count > 3):
                board[i][j] = -1
            elif board[i][j] == 0 and count == 3:
                board[i][j] = 2
    
    for i in range(ROW):
        for j in range(COL):
            if board[i][j] == -1:
                board[i][j] = 0
            elif board[i][j] == 2:
                board[i][j] = 1
```

```python
def gameOfLife(board):
    # Original | New | State
    #    0     |  0  |   0
    #    1     |  0  |   1
    #    0     |  1  |   2
    #    1     |  1  |   3
    ROW, COL = len(board), len(board[0])
    
    def countNeighbors(r, c):
        nei = 0
        for i in range(r -1, r + 2):
            for j in range(c - 1, c + 2):
                if ((i == r and j == c) or i < 0 or j < 0 or i == ROW or j == COL):
                    continue
                if board[i][j] in [1,3]:
                    nei += 1
        
        return nei  

    for r in range(ROW):
        for c in range(COL):
            nei = countNeighbors(r , c)
            
            if board[r][c]:
                if nei in [2,3]:
                    board[r][c] = 3
            elif nei == 3:
                    board[r][c] = 2
    
    for r in range(ROW):
        for c in range(COL):
            if board[r][c] == 1:
                board [r][c] = 0
            elif board[r][c] in [2,3]:
                board[r][c] = 1
```
