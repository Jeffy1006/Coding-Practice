```python
def CrossProduct(A):
    X1 = (A[1][0] - A[0][0])
    Y1 = (A[1][1] - A[0][1])
    X2 = (A[2][0] - A[1][0])
    Y2 = (A[2][1] - A[1][1])
    return (X1 * Y2 - Y1 * X2)

def isConvex(points):
    N = len(points)
    prev = 0
    curr = 0
    for i in range(N):
        temp = [points[i], points[(i + 1) % N], points[(i + 2) % N]]
        curr = CrossProduct(temp)
        if (curr != 0):
            if (curr * prev < 0):
                return False
            else:
                prev = curr
    return True
```

```python
isConvex([[1,1],[3,1],[3,3],[1,3]])
```
