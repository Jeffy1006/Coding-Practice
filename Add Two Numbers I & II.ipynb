```python
def addNumbers(l1, l2):
    carry = 0
    head = Node(None)
    cur = head
    cur1, cur2 = l1, l2
    while cur1 or cur2 or carry:
        tmp = carry
        if cur1:
            tmp += cur1.val
            cur1 = cur1.next
        if cur2:
            tmp += cur2.val
            cur2 = cur2.next
        carry, r = divmod(tmp, 10)
        cur.next = Node(r)
        cur = cur.next
    return head.next
```

```python
def addNumbersII(l1, l2):
    
    def reverseList(head):
        prev = None
        cur = head
        while cur:
            tmp = cur.next
            cur.next = prev
            prev = cur
            cur = tmp
        return prev
        
    carry = 0
    head = tail = None
    cur1, cur2 = reverseList(l1), reverseList(l2)
    while cur1 or cur2 or carry:
        tmp = carry
        if cur1:
            tmp += cur1.val
            cur1 = cur1.next
        if cur2:
            tmp += cur2.val
            cur2 = cur2.next
        carry, r = divmod(tmp, 10)
        head = Node(tmp)
        head.next = tail
        tail = head
    
    return head
```
