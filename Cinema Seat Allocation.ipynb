```python
from collections import defaultdict
def maxNumberOfFamilies(n, reservedSeats):
    seats = defaultdict(set)
    for i, j in reservedSeats:
        seats[i].add(j)

    case1 = set([2,3,4,5,6,7,8,9])
    case2 = set([4,5,6,7])
    case3 = set([2,3,4,5])
    case4 = set([6,7,8,9])
    res = 2 * (n - len(seats))
    for value in seats.values():
        if not value & case1:
            res += 2
        else:
            if not value & case2:
                res += 1
            elif not value & case3:
                res += 1
            elif not value & case4:
                res += 1
    return res
```

```python
n = 3
reservedSeats = [[1,2],[1,3],[1,8],[2,6],[3,1],[3,10]]
maxNumberOfFamilies(n, reservedSeats)
```
