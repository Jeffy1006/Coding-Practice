```python
def decodeString(s):
    ans = num = mini = ''
    rep = False
    for c in s:
        if c in '0123456789':
            num += c
        elif c =='[':
            rep = True
        elif c == ']':
            ans += (mini * int(num))
            num = ''
            mini = ''
            rep = False
        else:
            if rep:
                mini += c
            else:
                ans += c
    return ans
```

```python
decodeString('2[abc]3[cd]ef')
```

```python
decodeString('abc15[h]xyz')
```
