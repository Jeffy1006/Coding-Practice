```python
def numTeams(rating):
    n = len(rating)
    ans = 0
    for j in range(n):
        l, r = 0, 0
        for i in range(j):
            if rating[i] < rating[j]:
                l += 1
        for k in range(j + 1, n):
            if rating[j] < rating[k]:
                r += 1
        ans = (l * r) + (j - l) * (n - j - 1 - r)
    return ans 
```
