```python
def isPowerOfTwo(num):
    count = 0
    mask = 1
    for i in range(32):
        #print(mask)
        if (num & mask) == mask:
            count += 1
        mask *= 2

    return True if count == 1 else False
```

```python
def isPowerOfTwo(n):
    if n == 0: return False
    
    while n > 1 and (n & 1) == 0:
        n >>= 1
        
    return True if n == 1 else False
```

```python
def isPowerOfFour(n):
    if n <= 0: return False
    return (n & (n - 1) == 0) and (n % 3 == 1)
```
