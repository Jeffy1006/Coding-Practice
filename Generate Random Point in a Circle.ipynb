```python
import random
import math
class Solution:
    def __init__(self, radius, x_center, y_center):
        self.radius = radius
        self.x = x_center
        self.y = y_center
    
    def randPoint(self):
        hyp = self.radius * math.sqrt(random.uniform(0,1))
        theta = random.uniform(0, 2 * math.pi)
        return [self.x + (hyp * math.cos(theta))] + [self.y + (hyp * math.sin(theta))] 
```

```python
a = Solution(10, 2, 2)
```
