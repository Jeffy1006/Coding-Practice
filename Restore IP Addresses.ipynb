```python
class Solution:
    def resoreIPAdresses(self, s):
        
        if not 4 <= len(s) <= 12:
            return []
        
        def helper(idx, curr):
            nonlocal result
            if len(curr) == 4:
                if idx == len(s):
                    result.append('.'join(curr))
                return
            
            num = ''
            for i in range(idx, min(idx+3, len(s))):
                num += s[i]
                if  len(num) > 1 and num[0] == '0':
                    break
                if 0 <= int(num) <= 255:
                    curr.append(num)
                    helper(i+1,curr)
                    curr.pop()
                    
        
        result = []
        helper(0,[])
        return result
        
            
```
