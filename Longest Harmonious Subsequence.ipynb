```python
from collections import Counter
def findLHS(nums):
    count = Counter(nums)
    max_len = 0
    for k in count:
        if k + 1 in count:
            max_len = max(max_len, count[k] + count[k + 1])
    return max_len
```

```python
from collections import Counter
def findLHS(nums):
    count = Counter(nums)
    return max([count[k] + count[k + 1] for k in count if k + 1 in count], default = 0)
```
