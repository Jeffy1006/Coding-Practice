```python
class Node:
    def __init__(self, val):
        self.val = val
        self.children = []
```

```python
def preorder(root):
    res = []
    def helper(node):
        res.append(node.val)
        for child in node.children:
            helper(child)
    helper(root)
    return res
```

```python
def preorder(root):
    if not root: return []
    stack, res = [root], []
    while stack:
        node = stack.pop()
        res.append(node.val)
        stack.extend(node.children[::-1])
    return res
```

```python
def postorder(root):
    res = []
    def helper(node):
        for child in node.children:
            helper(child)
        res.append(node.val)
    helper(root)
    return res
```

```python
def postorder(root):
    if not root: return []
    stack, res = [root], []
    while stack:
        node = stack.pop()
        res.append(node.val)
        stack.extend(node.children)
    return res[::-1]
```
