```python
class FenwickTree:
    def __init__(self, n):
        self.n = n + 1
        self.tree = [0] * (n + 1)
    
    def update(self, i, delta):
        while i < self.n:
            self.tree[i] += delta
            i += (i & -i)
    
    def query(self, i):
        total = 0
        while i > 0:
            total += self.tree[i]
            i -= (i & -i)
        return total

class Solution:
    def reversePairs(self, nums):
        new_nums = nums + [x * 2 for x in nums]
        sorted_set = sorted(list(set(new_nums)))
        tree = FenwickTree(len(sorted_set))
        res = 0
        ranks = {}
        for i, n in enumerate(sorted_set):
            ranks[n] = i + 1
        for n in nums[::-1]:
            res += tree.query(ranks[n] - 1)
            tree.update(ranks[n * 2], 1)
        return res
```

```python
def reversePairs(nums):
    def reversePairsSub(l, r):
        if l >= r: return 0
        m = l + ((r - l) >> 1)
        res = reversePairsSub(l, m) + reversePairsSub(m + 1, r)
        
        i, j, k, p = l, m + 1, 0, m + 1
        merge = [0] * (r - l + 1)
        
        while i <= m:
            while p <= r and nums[i] > 2 * nums[p]:
                p += 1
            res += p - (m + 1)
            
            while j <= r and nums[i] >= nums[j]: 
                merge[k] = nums[j]
                k += 1
                j += 1
            merge[k] = nums[i]
            k += 1
            i += 1
        
        while j <= r:
            merge[k] = nums[j]
            k += 1
            j += 1
        
        nums[l:l + len(merge)] = merge
        return res
            
        
    return reversePairsSub(0, len(nums) - 1)
```
