```python
import math
def maximumGap(nums):
    maxGap = 0
    n = len(nums)
    if n < 2:
        return 0
    
    minNum = float("inf")
    maxNum = float("-inf")
    
    for num in nums:
        minNum = min(minNum, num)
        maxNum = max(maxNum, num)
    
    bucketSize = math.ceil((maxNum - minNum) / (n - 1))
    minOfBucket = [float("inf")] * (n - 1)
    maxOfBucket = [float("-inf")] * (n - 1)
    
    for i in range(n):
        if nums[i] == minNum or nums[i] == maxNum:
            continue
        bucketIdx = (nums[i] - minNum) // bucketSize
        minOfBucket[bucketIdx] = min(minOfBucket[bucketIdx], nums[i])
        maxOfBucket[bucketIdx] = max(maxOfBucket[bucketIdx], nums[i])
    print(minOfBucket)
    print(maxOfBucket)
    for i in range(n - 1):
        if maxOfBucket[i] == float("-inf"):
            continue
        maxGap = max(maxGap, (minOfBucket[i] - minNum))
        minNum = maxOfBucket[i]
    # when there is two elements in the dataset
    maxGap = max(maxGap, (maxNum - minNum))
    
    return maxGap
```

```python
maximumGap([21,9,25,3,37,43,49,29])
```

```python
maximumGap([1,100])
```
