```python
def nestedSum(List):
    
    def dfs(l, depth):
        SUM = 0
        for i in l:
            if type(i) == int:
                SUM += depth * i
            else:
                SUM += dfs(i, depth + 1)
        return SUM
    return dfs(List, 1)
```

```python
def nestedSumII(List):
    def helper(List, SUM):
        total = SUM
        tmp = []
        for e in List:
            if e.isInteger(): 
                total += e.getInteger
            else:
                tmp.extend(e.getList())
        if tmp:
            total += helper(tmp, total)
        return total
    
    return helper(List, 0)
```
