```python
def rightViewOfBST(root):
    
    res = []
    if not root:
        return res
    
    queue = []
    queue.append(root)
    
    while queue:
        for i in range(len(queue)):
            node == queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.qppen(node.right)
            if i == len(queue) - 1:
                res.append(node.val)
    
    return res
```
