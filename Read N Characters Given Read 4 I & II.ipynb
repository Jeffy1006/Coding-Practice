```python
def read(buf, n):
    
    buf4 = [''] * 4
    num_read = 0
    EOF = False
    while num_read < n and not EOF:
        curr_read = read4(buf4)
        delta = min(curr_read, n - num_read)
        buf[num_read:num_read + delta] = buf4[:delta]
        num_read += delta
        if cur_read < 4: EOF = True
    
    return num_read
```

```python
class Solution:
    def __init__(self):
        self.buf4 = [''] * 4
        self.curr_used = 0
        self.curr_read = 0
        self.EOF = False
        
    def read(buf, n):
        
        num_read = 0
        while num_read < n and not self.EOF:
            if self.curr_used == self.curr_read:
                self.curr_read = read4(self.buf4)
                self.curr_used = 0
                if self.curr_read == 0: self.EOF = True
            
            else:
                delta = min(self.curr_read - self.curr_used, n - num_read)
                buf[num_read:num_read + delta] = buf4[self.curr_used:self.curr_used + delta]
                num_read += delta
                self.curr_used += delta
    
        return num_read
    
```
