```python
def isValid(p1, p2, p3, p4):
    def dist(p1, p2):
        return (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2
    ls = set([dist(p1, p2), dist(p1, p3), dist(p1, p4), dist(p2, p3), dist(p2, p4), dist(p3, p4)])
    return 0 not in ls and len(ls) == 2
```
